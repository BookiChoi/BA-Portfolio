# 🧠 개념 (Concept)

## NumPy란 무엇인가? (What is it?)

**NumPy(Numerical Python)** 는 Python에서 수치 계산을 위한 **핵심 라이브러리**이다.  
핵심 자료구조인 **ndarray(n-dimensional array)** 를 제공하며,  
대용량 데이터를 빠르게 처리할 수 있도록 **C언어로 구현된 연산 엔진**을 내장하고 있다.

**NumPy (Numerical Python)** is the core library for numerical computation in Python.  
It provides **ndarray (n-dimensional array)** as its central data structure,  
with a **C-implemented computation engine** for processing large datasets at high speed.

| 구분 (Item) | Python list | NumPy ndarray |
|---|---|---|
| 타입 (Types) | 혼합 가능 (Mixed OK) | 단일 타입만 (Single type only) |
| 메모리 (Memory) | 비연속적 (Non-contiguous) | 연속적 (Contiguous) |
| 연산 속도 (Speed) | 느림 (Slow — loop needed) | 빠름 (Fast — vectorized) |
| 배열 연산 (Array ops) | 루프 필요 (Loop required) | 한 줄로 가능 (One line) |
| 결측치 (Missing values) | 없음 (Not supported) | `np.nan` 사용 가능 |

## 왜 사용하는가? (Why do we use it?)

Python list는 범용 컨테이너로 유연하지만, 수백만 개의 숫자를 처리할 때는 너무 느리다.  
ndarray는 타입이 고정되어 메모리가 연속 배치되므로, **루프 없이 C 수준의 속도**로 전체 배열을 한 번에 연산할 수 있다.

Python list is flexible but too slow when processing millions of numbers.  
ndarray's fixed type allows contiguous memory layout — enabling **C-level speed on the entire array at once, without loops**.

## 비즈니스 애널리틱스에서 언제 쓰이나? (When is it used in Business Analytics?)

- **Pandas 내부 기반** / Internal foundation of Pandas  
  → Pandas DataFrame의 각 컬럼은 내부적으로 NumPy 배열로 저장된다  
  → Each column of a Pandas DataFrame is stored internally as a NumPy array

- **대용량 연산** / Large-scale computation  
  → 수천만 건의 매출/고객 데이터를 루프 없이 한 줄로 계산할 때  
  → Process millions of rows of sales/customer data in a single line

- **머신러닝 전처리** / ML preprocessing  
  → 모델에 데이터를 넣기 전 정규화(normalization), 변환(transformation) 시  
  → Before feeding data into ML models: normalization, scaling, transformation

- **통계 지표 계산** / Statistical metrics  
  → 평균(mean), 표준편차(std), 분위수(percentile) 등  
  → Mean, standard deviation, percentiles, etc.

## 주요 변형 (Common Variations)

````python
import numpy as np

# ── Python list 생성 (Create a Python list) ──────────────────────────
py_list = [1, 2, 3, 4, 5]
print("list type :", type(py_list))   # <class 'list'>

# ── NumPy ndarray 생성 (Create a NumPy ndarray) ──────────────────────
np_array = np.array([1, 2, 3, 4, 5])
print("ndarray type:", type(np_array))  # <class 'numpy.ndarray'>
print("dtype       :", np_array.dtype)  # int64 (자동 추론 / auto-inferred)

# ── 타입 혼합 비교 (Mixed-type behavior) ─────────────────────────────
mixed_list  = [1, "hello", True, 3.14]    # list: 혼합 허용 (mixed OK)
mixed_array = np.array([1, "hello", True, 3.14])  # ndarray: 모두 str로 변환!
print("mixed list :", mixed_list)
print("mixed array:", mixed_array)         # ['1' 'hello' 'True' '3.14']
print("dtype      :", mixed_array.dtype)   # <U32 (유니코드 문자열 / unicode string)

# ── 연산 비교 (Operation comparison) ────────────────────────────────
py_list_2  = [1, 2, 3, 4, 5]
np_array_2 = np.array([1, 2, 3, 4, 5])

print([x * 2 for x in py_list_2])   # list:    루프 필요 (loop required)
print(np_array_2 * 2)               # ndarray: 한 줄로 가능 (one line — vectorized)

# list + list vs ndarray + ndarray — 결과가 다르다! (different behavior!)
print([1,2,3] + [4,5,6])             # [1,2,3,4,5,6]  리스트 연결 (concatenation)
print(np.array([1,2,3]) + np.array([4,5,6]))  # [5 7 9]  요소별 덧셈 (element-wise add)
````

## Example 1 — NumPy 임포트 & 버전 확인 (Import & Version Check)

## Example 2 — list vs ndarray: 타입 & 연산 비교 (Type & Operation Comparison)

````python
import numpy as np

# ── 같은 데이터, 다른 자료구조 (Same data, different structures) ──────
sales_list  = [150000, 320000, 80000, 450000, 210000]   # Python list
sales_array = np.array([150000, 320000, 80000, 450000, 210000])  # ndarray

# 부가세 10% 적용 (Apply 10% VAT)
# list: 루프 또는 리스트 컴프리헨션 필요 (loop or list comprehension needed)
sales_with_tax_list = [x * 1.1 for x in sales_list]
print("list   (VAT applied):", sales_with_tax_list)

# ndarray: 한 줄로 전체 배열에 적용 (one line applies to every element)
sales_with_tax_array = sales_array * 1.1
print("ndarray(VAT applied):", sales_with_tax_array)

# ── 통계 계산 비교 (Statistics comparison) ──────────────────────────
print("\n평균 (Mean):", np.mean(sales_array))     # NumPy: 한 줄
print("최대 (Max) :", np.max(sales_array))
print("최소 (Min) :", np.min(sales_array))

# ── 조건 필터링 (Boolean filtering) ────────────────────────────────
# 30만원 이상인 고객 (customers who spent 300,000+ won)
is_vip = sales_array >= 300000
print("\nVIP 여부 (Is VIP):", is_vip)              # [False  True False  True False]
print("VIP 금액 (VIP sales):", sales_array[is_vip]) # [320000 450000]
````

## Example 3 — 성능 비교: 100만 건 데이터 (Performance: 1 Million Records)

````python
import numpy as np
import time

# 실무 상황: 100만 건의 매출 데이터 합산
# Business scenario: sum up 1 million sales records
sales_data = np.random.randint(1000, 100000, size=1_000_000)

# ── NumPy 방식 (NumPy approach) ──────────────────────────────────────
start = time.time()
total_numpy = np.sum(sales_data)
numpy_time  = time.time() - start

# ── Python 루프 방식 (Python loop approach) ──────────────────────────
start = time.time()
total_loop = sum([x for x in sales_data])
loop_time  = time.time() - start

print(f"NumPy  소요 시간 (time): {numpy_time:.4f}s  →  합계: {total_numpy:,}")
print(f"Loop   소요 시간 (time): {loop_time:.4f}s  →  합계: {total_loop:,}")
print(f"NumPy가 약 {loop_time / numpy_time:.0f}배 빠름 (NumPy is ~{loop_time / numpy_time:.0f}x faster)")

# 실무 설명 (Business insight):
# 일별 리포트 배치 작업에서 100만 건 이상의 데이터를 반복 처리할 때
# NumPy의 속도 차이는 몇 초 → 수십 ms 수준으로 매우 큼
# In daily report batch jobs processing 1M+ records,
# the speed difference is dramatic: seconds (loop) vs milliseconds (NumPy)
````

# ⚠️ 자주 하는 실수 (Common Mistakes)

**Mistake 1 — `as np` 없이 임포트 (Importing without `as np`)**  
`import numpy` → 이후 모든 코드에서 `numpy.array()` 처럼 풀네임 사용해야 함  
`import numpy` → must write `numpy.array()` everywhere afterward  
✅ 항상 / Always: `import numpy as np`

**Mistake 2 — 다른 별칭 사용 (Using a different alias)**  
`import numpy as num` → 팀 협업 시 혼란 발생. np는 전 세계 표준 관례  
`import numpy as num` → confuses teammates. `np` is the universal convention.  
✅ 반드시 / Must use: `import numpy as np`

**Mistake 3 — 타입 혼합 시 자동 변환 (Auto-conversion with mixed types)**  
`np.array([1, 2, "3"])` → 모두 문자열로 변환됨 → 수치 연산 불가  
`np.array([1, 2, "3"])` → all converted to string → arithmetic operations fail  
✅ ndarray는 단일 타입만 / ndarray holds only one dtype — keep them consistent

**Mistake 4 — list `+` vs ndarray `+` 혼동 (Confusing list `+` and ndarray `+`)**  
`[1,2,3] + [4,5,6]` → `[1,2,3,4,5,6]` (리스트 연결 / concatenation)  
`np.array([1,2,3]) + np.array([4,5,6])` → `[5 7 9]` (요소별 덧셈 / element-wise add)

# 💡 팁 (Tips)

- **임포트 관례** / Import convention: `import numpy as np` — 전 세계 표준, 예외 없이 사용 / universal standard, no exceptions
- **버전 확인** / Version check: `np.__version__` — 환경 세팅 시 항상 확인 / always check when setting up an environment
- **dtype 먼저 확인** / Check dtype first: ndarray는 단일 타입 → 연산 전에 `.dtype` 으로 타입 확인  
  / ndarray is single-typed → verify `.dtype` before any operation
- **Pandas 연동** / Pandas connection: `df['column'].values` 또는 `.to_numpy()` → ndarray 반환  
  / `df['column'].values` or `.to_numpy()` returns an ndarray
- **list → ndarray 변환** / Converting: `np.array(my_list)` — 가장 기본적인 변환  
  / `np.array(my_list)` — the most basic conversion

# 💼 비즈니스 예시 (Business Example)

**시나리오 (Scenario):**  
한 이커머스 회사의 BA가 **고객 5명의 월 구매금액** 데이터를 분석한다.  
A BA at an e-commerce company analyzes **monthly purchase amounts** for 5 customers.

**할 일 (Tasks):**  
1. 부가세(10%) 포함 금액 계산 / Calculate VAT-inclusive amounts (10%)  
2. 평균 구매금액 계산 / Compute average purchase amount  
3. VIP 고객 분류 (30만원 이상) / Classify VIP customers (≥ ₩300,000)

````python
import numpy as np

# 고객 5명의 월 구매금액 (단위: 원)
# Monthly purchase amounts for 5 customers (unit: KRW)
purchases = np.array([150000, 320000, 80000, 450000, 210000])

# 1. 부가세 10% 추가 (Add 10% VAT)
#    ndarray는 스칼라 연산을 모든 원소에 자동 적용 (broadcasting)
#    ndarray applies scalar operation to every element automatically (broadcasting)
purchases_with_tax = purchases * 1.1
print("부가세 포함 금액 (VAT-inclusive):", purchases_with_tax)

# 2. 평균 구매금액 (Average purchase amount)
print("평균 구매금액 (Average):", np.mean(purchases), "원")

# 3. VIP 분류: 30만원 이상 (VIP flag: 300,000+ won)
#    Boolean 배열 반환 (returns a Boolean array)
is_vip = purchases >= 300000
print("VIP 여부 (Is VIP) :", is_vip)
print("VIP 구매금액 (VIP purchases):", purchases[is_vip])
print("VIP 고객 수 (VIP count):", np.sum(is_vip))
````

# 📝 요약 (Summary)

NumPy는 Python의 핵심 수치 계산 라이브러리로, `ndarray`(n-dimensional array)라는 단일 타입의 고속 배열을 제공한다.  
Python list와 달리 ndarray는 타입이 고정되어 메모리가 연속 배치되므로, 루프 없이 C 수준의 속도로 전체 배열을 한 번에 연산할 수 있다.  
`list + list`는 연결(concatenation)이지만 `ndarray + ndarray`는 요소별 덧셈(element-wise)이라는 점에 주의해야 한다.  
모든 데이터 분석 라이브러리(Pandas, Scikit-learn 등)의 기반이 되므로 `import numpy as np`는 모든 분석 코드의 시작점이다.

NumPy is Python's core numerical computing library, providing `ndarray` — a single-type, high-speed array.  
Unlike Python list, ndarray's fixed type allows contiguous memory layout, enabling C-level speed across the entire array without loops.  
Note: `list + list` means concatenation, while `ndarray + ndarray` means element-wise addition.  
Since it underlies all data analysis libraries (Pandas, Scikit-learn, etc.), `import numpy as np` is the starting point of every analysis script.

> NumPy replaces Python list with a **C-speed numerical array (ndarray)**, forming the foundation of all data analysis including Pandas.
