# 🧠 Concept (개념)

## What is it? (이게 뭔가요?)
*(Explain it in your own words. 자신의 언어로 설명해보세요.)*

A comprehension builds a new `list`, `dict`, or `set` in a single expression: `[x**2 for x in range(10)]` reads as "give me `x**2` for every `x` in `range(10)`." A generator expression looks almost identical but uses `()` instead of `[]`, and produces values lazily, one at a time, instead of building the whole collection in memory upfront. A generator *function* does the same lazy production, but as a full function using the `yield` keyword instead of `return`.

컴프리헨션은 하나의 표현식으로 새로운 `list`, `dict`, `set`을 만든다: `[x**2 for x in range(10)]`는 "`range(10)`의 모든 `x`에 대해 `x**2`를 달라"로 읽힌다. 제너레이터 표현식은 거의 똑같이 생겼지만 `[]` 대신 `()`를 쓰며, 전체 컬렉션을 미리 메모리에 만드는 대신 값을 하나씩 게으르게 생성한다. 제너레이터 *함수*는 같은 게으른 생성을 하지만, `return` 대신 `yield` 키워드를 쓰는 완전한 함수 형태다.

| | JS/TS | Python |
|---|---|---|
| Transform a list 리스트 변환 | `arr.map(x => x**2)` | `[x**2 for x in arr]` |
| Filter a list 리스트 필터링 | `arr.filter(x => x % 2 === 0)` | `[x for x in arr if x % 2 == 0]` |
| Transform + filter 변환 + 필터링 | `arr.filter(...).map(...)` | `[x**2 for x in arr if x % 2 == 0]` (one pass, one line) |
| Build an object 객체 만들기 | `Object.fromEntries(arr.map(...))` | `{k: v for k, v in ...}` |
| Lazy, memory-efficient sequence 게으른, 메모리 효율적 시퀀스 | generator function (`function*`) | generator expression `(...)` or `yield` |
| Pause and resume a function 함수 일시정지·재개 | `function*` + `yield` | `def` + `yield` (same idea!) |

The one thing with no real JS/TS equivalent: Python comprehensions do transform-and-filter *in one pass, one line*, while JS/TS typically chains `.filter()` then `.map()` as two separate array traversals.

JS/TS에 딱히 대응물이 없는 것 하나: Python 컴프리헨션은 변환과 필터링을 *한 번의 순회, 한 줄*로 처리하지만, JS/TS는 보통 `.filter()`와 `.map()`을 체이닝해서 배열을 두 번 순회한다.

## Why do we use it? (왜 사용하나요?)
*(When is it useful? 언제 유용한가요?)*

Comprehensions read as a single, declarative statement of intent — "the squares of the even numbers" — instead of a multi-line loop that builds up a result through explicit `.append()` calls; both do the same work, but the comprehension is usually faster and states the *what* without the *how*.

컴프리헨션은 명시적인 `.append()` 호출을 통해 결과를 쌓아가는 여러 줄짜리 반복문 대신, "짝수들의 제곱"처럼 단일하고 선언적인 의도 표현으로 읽힌다 — 둘 다 같은 일을 하지만, 컴프리헨션이 보통 더 빠르고 *어떻게*가 아니라 *무엇을*을 표현한다.

Generators solve a different problem: memory. `sum(x**2 for x in range(10_000_000))` never holds ten million numbers in memory at once — it computes and discards one at a time — while the list-comprehension version would build the entire list first, just to sum it and throw it away.

제너레이터는 다른 문제, 메모리를 해결한다. `sum(x**2 for x in range(10_000_000))`는 천만 개의 숫자를 한 번에 메모리에 담지 않는다 — 하나씩 계산하고 버린다 — 반면 리스트 컴프리헨션 버전은 합산하고 버리기 위해서만 전체 리스트를 먼저 만들어야 한다.

## When is it used in Business Analytics? (Business Analytics에서는 언제 쓰이나요?)
*(Real-world use case 실제 사용 사례)*

Reshaping raw records into the exact structure needed for a report — extracting one field from every dict in a list, filtering out invalid rows, or flipping a code-to-name lookup into a name-to-code one — is what comprehensions do dozens of times a day in a typical cleaning script.

리포트에 필요한 정확한 구조로 원본 레코드를 재구성하는 일 — 딕셔너리 리스트에서 필드 하나를 뽑아내거나, 유효하지 않은 행을 걸러내거나, 코드→이름 조회를 이름→코드로 뒤집는 일 — 은 일반적인 정리 스크립트에서 컴프리헨션이 하루에도 수십 번씩 하는 일이다.

Generator expressions matter the moment a dataset gets large enough that "build the whole list first" becomes slow or memory-hungry — reading a huge CSV line by line, or summing a computed value across millions of rows without pandas, are exactly the situations where a generator keeps a script responsive instead of freezing.

제너레이터 표현식은 데이터셋이 "전체 리스트를 먼저 만들기"가 느려지거나 메모리를 많이 잡아먹을 만큼 커지는 순간부터 중요해진다 — 거대한 CSV를 한 줄씩 읽거나 pandas 없이 수백만 행에 걸쳐 계산된 값을 합산하는 것은 정확히 제너레이터가 스크립트를 멈추지 않고 반응하게 유지해주는 상황이다.

# 📝 Syntax (문법)

## Basic Syntax (기본 문법)

````python
numbers = [1, 2, 3, 4, 5, 6]

# List comprehension — transform every item
# 리스트 컴프리헨션 — 모든 항목을 변환
squares = [x**2 for x in numbers]
print(squares)

# ...with a filter
# ...필터를 추가
evens = [x for x in numbers if x % 2 == 0]
print(evens)

# Dict comprehension — build key-value pairs
# 딕셔너리 컴프리헨션 — 키-값 쌍 만들기
square_map = {x: x**2 for x in numbers}
print(square_map)

# Set comprehension — unique results only
# 셋 컴프리헨션 — 고유한 결과만
remainders = {x % 3 for x in numbers}
print(remainders)

# Generator expression — same syntax as a list comprehension, but with ()
# 제너레이터 표현식 — 리스트 컴프리헨션과 같은 문법, 다만 ()를 사용
total = sum(x**2 for x in numbers)
print(total)
````

## Common Variations (자주 쓰는 변형)

````python
# Conditional expression INSIDE a comprehension — not a filter, a value choice
# 컴프리헨션 안의 조건 표현식 — 필터가 아니라 값을 고르는 것
labels = ["even" if x % 2 == 0 else "odd" for x in range(5)]
print(labels)

# Nested comprehension — flatten a 2D structure in one line
# 중첩 컴프리헨션 — 2D 구조를 한 줄로 평탄화
matrix = [[1, 2, 3], [4, 5, 6]]
flat = [x for row in matrix for x in row]
print(flat)                # read order: for row -> for x -> expression (읽는 순서: for row -> for x -> 표현식)

# Dict comprehension to invert a mapping
# 매핑을 뒤집는 딕셔너리 컴프리헨션
code_to_name = {"A": "Alpha", "B": "Beta"}
name_to_code = {v: k for k, v in code_to_name.items()}
print(name_to_code)

# Generator expressions work directly as function arguments — no extra parentheses needed
# 제너레이터 표현식은 함수 인자로 바로 사용 가능 — 추가 괄호 불필요
values = [1, 5, 200]
print(any(x > 100 for x in values))
print(max(x for x in values))
````

# 🧪 Small Examples (예제)

## Example 1 — List Comprehensions (리스트 컴프리헨션)

````python
# Basic transform
# 기본 변환
print([x**2 for x in range(10)])

# Filter only (no transform)
# 필터만 (변환 없음)
print([x for x in range(10) if x % 2 == 0])

# Transform AND filter together, in one pass
# 변환과 필터를 한 번에, 한 번의 순회로
print([x**2 for x in range(10) if x % 2 == 0])

# Conditional expression — chooses a VALUE, not a filter
# 조건 표현식 — 필터가 아니라 값을 고름
print(["odd" if x % 2 else "even" for x in range(5)])

# Nested comprehension — flattening a matrix
# 중첩 컴프리헨션 — 행렬 평탄화
matrix = [[1, 2, 3], [4, 5, 6]]
print([x for row in matrix for x in row])

# Equivalent written as a traditional loop, for comparison
# 비교를 위해 전통적인 반복문으로 작성한 동등한 코드
result = []
for row in matrix:
    for x in row:
        result.append(x)
print(result)
````

## Example 2 — Dict Comprehensions (딕셔너리 컴프리헨션)

````python
# Basic key-value construction from two parallel sequences
# 두 병렬 시퀀스로부터 기본적인 키-값 구성
print({k: v for k, v in zip("abc", [1, 2, 3])})

# Transforming values while keeping the keys
# 키는 유지하면서 값을 변환
prices = {"coffee": 4.5, "tea": 3.5}
doubled = {k: v * 2 for k, v in prices.items()}
print(doubled)

# Filtering — only keep entries that pass a condition
# 필터링 — 조건을 통과하는 항목만 유지
scores = {"Bob": 70, "Ann": 95, "Kim": 60}
passing = {name: score for name, score in scores.items() if score >= 70}
print(passing)

# Inverting a dict — swap keys and values
# 딕셔너리 역전 — 키와 값을 교환
status_codes = {200: "OK", 404: "Not Found"}
name_to_code = {name: code for code, name in status_codes.items()}
print(name_to_code)

# Building a dict directly from two lists
# 두 리스트로부터 직접 딕셔너리 만들기
names = ["Bob", "Ann", "Kim"]
ages = [30, 25, 40]
print({name: age for name, age in zip(names, ages)})
````

## Example 3 — Set Comprehensions (셋 컴프리헨션)

````python
# Deduplicate while transforming — a set comprehension does both at once
# 변환하면서 중복 제거 — 셋 컴프리헨션은 둘 다 한 번에 처리
print({x % 3 for x in range(10)})

# Unique word lengths in a list
# 리스트에 있는 고유한 단어 길이
words = ["hi", "hello", "hey", "yo", "sup"]
print({len(w) for w in words})

# A set comprehension is the natural way to collect unique values while filtering
# 필터링하면서 고유한 값을 모으는 자연스러운 방법
numbers = [1, -2, 3, -4, 3, 1, 5]
print({abs(n) for n in numbers})

# Comparing to a list comprehension on the same data — note the different result shape
# 같은 데이터에 대한 리스트 컴프리헨션과 비교 — 결과 형태가 다름을 확인
print([abs(n) for n in numbers])       # keeps every value, in order, with duplicates (모든 값을 순서대로, 중복 포함)
print({abs(n) for n in numbers})         # keeps only unique values, unordered (고유한 값만, 순서 없음)
````

## Example 4 — Generator Expressions (제너레이터 표현식)

````python
words = ["apple", "banana", "kiwi", "fig"]

# sum() / max() with a generator expression — no intermediate list is ever built
# 제너레이터 표현식을 쓰는 sum()/max() — 중간 리스트가 전혀 만들어지지 않음
print(sum(len(w) for w in words))
print(max(len(w) for w in words))

# any() short-circuits — stops at the first True instead of checking everything
# any()는 단락 평가 — 모든 것을 확인하는 대신 첫 True에서 멈춤
print(any(len(w) > 5 for w in words))

# next() pulls just the first matching value, without generating the rest
# next()는 나머지를 생성하지 않고 첫 번째로 일치하는 값만 꺼냄
first_long_word = next(w for w in words if len(w) > 4)
print(first_long_word)

# A generator expression can be assigned to a variable and reused — but only ONCE
# 제너레이터 표현식은 변수에 할당해 재사용할 수 있지만 — 딱 한 번만
lengths_gen = (len(w) for w in words)
print(list(lengths_gen))          # consumes the generator (제너레이터를 소진시킴)
print(list(lengths_gen))            # already exhausted — empty! (이미 소진됨 — 비어 있음!)
````

````python
# ============================================================
# Practice — Generator Functions: yield (연습 — 제너레이터 함수)
# ============================================================

# 1) A generator function that counts up to n — fill in the yield target
#    n까지 세는 제너레이터 함수 — yield 대상 채우기
def count_up(n):
    i = 0
    while i < n:
        yield i
        i += 1

print("1)", list(count_up(5)))            # TODO: which variable should each yield produce? Expected: [0, 1, 2, 3, 4]

# 2) A generator function that yields only even numbers up to n
#    n까지의 짝수만 yield하는 제너레이터 함수
def even_numbers(n):
    for i in range(n):
        if i % 2 == 0:
            yield i

print("2)", list(even_numbers(10)))        # Expected: [0, 2, 4, 6, 8]

# 3) Pulling values one at a time with next()
#    next()로 값을 하나씩 꺼내기
gen = count_up(3)
print("3)", next(gen), next(gen))      # TODO: call next() on which generator object?

# 4) A generator expression (no def/yield needed) — the compact one-line form
#    제너레이터 표현식 (def/yield 불필요) — 한 줄짜리 압축 형태
squares_gen = (x**2 for x in range(5))
print("4)", list(squares_gen))               # TODO: fill in the range so this produces [0, 1, 4, 9, 16]

# 5) yield from — delegate to another iterable instead of writing a nested loop
#    yield from — 중첩 반복문 대신 다른 이터러블에 위임
def flatten(nested):
    for inner_list in nested:
        yield from inner_list

print("5)", list(flatten([[1, 2], [3, 4], [5]])))       # TODO: which variable holds each inner list?

# 6) Generators are exhausted after one full pass — this is what makes them memory-efficient
#    제너레이터는 한 번 다 쓰면 소진됨 — 이것이 메모리 효율의 이유
gen2 = (x for x in range(3))
first_pass = list(gen2)
second_pass = list(gen2)                  # TODO: which generator object is now exhausted?
print("6)", first_pass, second_pass)           # Expected: [0, 1, 2] []
````

# ⚠️ Common Mistakes (자주 하는 실수)

### Mistake 1 — Getting the read order of nested comprehensions backward
`[x for row in matrix for x in row]` reads left to right in the SAME order as writing nested `for` loops normally: `for row in matrix:` is the outer loop, `for x in row:` is the inner loop. Writing them in the wrong order is a very common early mistake.

`[x for row in matrix for x in row]`는 일반적인 중첩 `for` 반복문을 작성하는 것과 같은 순서로 왼쪽에서 오른쪽으로 읽힌다: `for row in matrix:`가 바깥 반복문, `for x in row:`가 안쪽 반복문이다. 순서를 거꾸로 쓰는 것은 초반에 매우 흔한 실수다.

✅ **Fix / 해결법:**  
Mentally unroll it into a normal nested loop, in the same left-to-right order, to check it is correct.  
왼쪽에서 오른쪽 순서 그대로 일반적인 중첩 반복문으로 머릿속에서 풀어보며 확인하자.
```python
matrix = [[1, 2], [3, 4]]
# [x for row in matrix for x in row]  reads as:
for row in matrix:            # outer — matches the FIRST for
    for x in row:                # inner — matches the SECOND for
        pass                        # x  — matches the expression
```

### Mistake 2 — Reusing an already-exhausted generator
A generator can only be iterated ONCE. After a `for` loop or `list()` call consumes it fully, iterating it again silently produces nothing — no error, just an empty result, which is a confusing thing to debug.

제너레이터는 한 번만 순회할 수 있다. `for` 반복문이나 `list()` 호출이 완전히 소진시킨 뒤 다시 순회하면 조용히 아무것도 만들어내지 않는다 — 에러 없이 그냥 빈 결과이며, 디버깅하기 혼란스럽다.

✅ **Fix / 해결법:**  
If the data needs to be used more than once, store it in a `list` (or create a fresh generator each time it is needed).  
데이터를 두 번 이상 써야 한다면 `list`에 저장하자 (또는 필요할 때마다 새 제너레이터를 만들자).
```python
gen = (x for x in range(3))
first = list(gen)            # [0, 1, 2]
second = list(gen)             # [] — already exhausted!
# fix: values = list(x for x in range(3)); reuse `values` as many times as needed
```

### Mistake 3 — Assuming a function with `yield` runs like a normal function
The moment a function body contains `yield` anywhere, calling that function does NOT execute the body at all — it immediately returns a generator object. The code only runs when the generator is actually iterated (via a `for` loop, `list()`, or `next()`).

함수 본문 어딘가에 `yield`가 있는 순간, 그 함수를 호출해도 본문이 전혀 실행되지 않는다 — 즉시 제너레이터 객체를 반환할 뿐이다. 코드는 제너레이터가 실제로 순회될 때(`for` 반복문, `list()`, `next()`를 통해)만 실행된다.

✅ **Fix / 해결법:**  
Remember that calling a generator function only creates the generator; wrap it in `list(...)`, loop over it, or call `next()` to actually run any of its code.  
제너레이터 함수를 호출하면 제너레이터만 생성된다는 것을 기억하자; 실제로 코드를 실행하려면 `list(...)`로 감싸거나, 반복문을 돌리거나, `next()`를 호출하자.
```python
def greet():
    print("this only prints when iterated")     # (순회될 때만 출력됨)
    yield "hello"

g = greet()             # nothing printed yet — just a generator object created (아직 아무것도 출력되지 않음)
next(g)                    # ✅ NOW "this only prints when iterated" appears
```

# 💡 Tips (팁)
Useful tips or shortcuts (유용한 팁이나 단축법)

- A comprehension that needs more than one `if` or gets hard to read in one line is a signal to switch back to a regular `for` loop — comprehensions should stay readable, not become a code-golf challenge.  
컴프리헨션에 `if`가 여러 개 필요하거나 한 줄로 읽기 어려워지면 일반 `for` 반복문으로 돌아가라는 신호다 — 컴프리헨션은 읽기 쉬워야지 코드 골프 도전이 되면 안 된다.
- Use a generator expression (`(...)`) instead of a list comprehension (`[...]`) whenever the result is only going to be consumed once, right away, by something like `sum()`, `max()`, or `any()` — there is no reason to build a throwaway list.  
결과가 `sum()`, `max()`, `any()` 같은 것에 의해 한 번만, 즉시 소비될 거라면 리스트 컴프리헨션(`[...]`) 대신 제너레이터 표현식(`(...)`)을 사용하자 — 버려질 리스트를 만들 이유가 없다.
- `{v: k for k, v in d.items()}` is the standard one-liner for inverting a dictionary — but it silently drops duplicate values, keeping only the last key seen for each.  
`{v: k for k, v in d.items()}`는 딕셔너리를 뒤집는 표준적인 한 줄 코드다 — 하지만 중복된 값은 조용히 버려지고, 각 값에 대해 마지막으로 본 키만 남는다.
- `next(x for x in data if condition)` finds the first matching item without scanning or storing the rest — much more efficient than filtering into a list and taking `[0]`.  
`next(x for x in data if condition)`는 나머지를 훑거나 저장하지 않고 첫 번째로 일치하는 항목을 찾는다 — 리스트로 필터링한 뒤 `[0]`을 가져오는 것보다 훨씬 효율적이다.
- `yield from other_iterable` is almost always better than writing `for item in other_iterable: yield item` manually — it does the exact same thing in one line.  
`yield from other_iterable`는 `for item in other_iterable: yield item`을 직접 쓰는 것보다 거의 항상 낫다 — 정확히 같은 일을 한 줄로 한다.

# 💼 Business Example (비즈니스 예제)
*How would a Business Analyst use this? BA는 이것을 실제로 어떻게 사용할까요?*

**Scenario (시나리오):** A raw list of transaction dicts needs to be reshaped into three different views for a report: a filtered list, a lookup dict, and a memory-safe running total — all using comprehensions and generators, no pandas.

**시나리오:** 원본 거래 딕셔너리 리스트를 리포트를 위한 세 가지 다른 형태로 재구성해야 한다: 필터링된 리스트, 조회 딕셔너리, 메모리 안전한 누적 합계 — 전부 컴프리헨션과 제너레이터만으로, pandas 없이.

**To do (할 일 목록):**
- [x] Use a list comprehension to keep only completed transactions.  
리스트 컴프리헨션으로 완료된 거래만 남긴다.
- [x] Use a dict comprehension to build a transaction-ID-to-amount lookup.  
딕셔너리 컴프리헨션으로 거래ID→금액 조회 테이블을 만든다.
- [x] Use a generator expression with `sum()` so the total is never held as an intermediate list.  
총합이 중간 리스트로 저장되지 않도록 `sum()`과 함께 제너레이터 표현식을 사용한다.
- [x] Write a small generator function that yields only the high-value transactions, one at a time.  
고액 거래만 하나씩 yield하는 작은 제너레이터 함수를 작성한다.

````python
transactions = [
    {"id": "T1", "amount": 250.0, "status": "completed"},
    {"id": "T2", "amount": 80.0,  "status": "pending"},
    {"id": "T3", "amount": 500.0, "status": "completed"},
    {"id": "T4", "amount": 30.0,  "status": "failed"},
    {"id": "T5", "amount": 120.0, "status": "completed"},
]

# List comprehension — keep only completed transactions
# 리스트 컴프리헨션 — 완료된 거래만 유지
completed = [t for t in transactions if t["status"] == "completed"]
print("Completed (완료됨):", [t["id"] for t in completed])

# Dict comprehension — a fast id -> amount lookup
# 딕셔너리 컴프리헨션 — 빠른 id -> 금액 조회
amount_by_id = {t["id"]: t["amount"] for t in completed}
print("Lookup (조회 테이블):", amount_by_id)

# Generator expression — sum without ever building an intermediate list
# 제너레이터 표현식 — 중간 리스트 없이 합산
total = sum(t["amount"] for t in completed)
print(f"Total completed (완료된 총액): ${total:,.2f}")

# Generator function — yields high-value transactions one at a time
# 제너레이터 함수 — 고액 거래를 하나씩 yield
def high_value_transactions(records, threshold=150):
    for t in records:
        if t["status"] == "completed" and t["amount"] >= threshold:
            yield t

print("High-value (고액 거래):")
for t in high_value_transactions(transactions):
    print(f"  {t['id']}: ${t['amount']:,.2f}")
````

# 📝 Summary (요약)
*Write today's concept in 3~5 sentences. 오늘 배운 개념을 3~5문장으로 정리해보세요.*

Comprehensions build a `list`, `dict`, or `set` in a single declarative expression — `[expr for item in iterable if condition]` — that reads as a statement of intent rather than a step-by-step loop, and nested comprehensions read left-to-right in the same order as writing nested `for` loops normally. Generator expressions use the same syntax with `()` instead of `[]`, but produce values lazily, one at a time, which matters enormously for memory once a dataset gets large — and a generator can only be iterated once, so reusing an exhausted one silently returns nothing. A generator *function* achieves the same laziness using `yield` instead of `return`; calling it does not run the function body at all, it just creates a generator object, and the body only executes as values are actually pulled out via `next()` or a loop. `yield from` delegates to another iterable in one line instead of writing a manual inner loop. These patterns are the direct ancestors of pandas' `.apply()`, boolean masking, and chunked file reading.

컴프리헨션은 `list`, `dict`, `set`을 단일 선언적 표현식 — `[expr for item in iterable if condition]` — 으로 만든다. 이는 단계별 반복문이 아니라 의도를 나타내는 문장으로 읽히며, 중첩 컴프리헨션은 일반적인 중첩 `for` 반복문을 작성하는 것과 같은 순서로 왼쪽에서 오른쪽으로 읽힌다. 제너레이터 표현식은 `[]` 대신 `()`를 쓰는 같은 문법이지만 값을 하나씩 게으르게 생성한다. 이는 데이터셋이 커질수록 메모리 측면에서 대단히 중요해진다 — 그리고 제너레이터는 한 번만 순회할 수 있어서, 소진된 것을 재사용하면 조용히 아무것도 반환하지 않는다. 제너레이터 *함수*는 `return` 대신 `yield`를 사용해 같은 게으름을 달성한다; 호출해도 함수 본문이 전혀 실행되지 않고 제너레이터 객체만 생성되며, 본문은 `next()`나 반복문을 통해 실제로 값을 꺼낼 때만 실행된다. `yield from`은 수동 내부 반복문을 작성하는 대신 다른 이터러블에 한 줄로 위임한다. 이 패턴들은 pandas의 `.apply()`, 불리언 마스킹, 청크 단위 파일 읽기의 직계 조상이다.

> Comprehensions (`[...]`, `{...}`) build a whole collection in one declarative line, while generator expressions (`(...)`) and generator functions (`yield`) produce the same values lazily, one at a time, which saves memory but means they can only be consumed once.

> 컴프리헨션(`[...]`, `{...}`)은 한 줄의 선언적 표현으로 전체 컬렉션을 만들고, 제너레이터 표현식(`(...)`)과 제너레이터 함수(`yield`)는 같은 값을 하나씩 게으르게 생성한다 — 메모리는 절약되지만 한 번만 소비할 수 있다는 뜻이다.
