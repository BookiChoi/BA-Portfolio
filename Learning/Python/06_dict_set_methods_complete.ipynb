# 🧠 Concept (개념)

## What is it? (이게 뭔가요?)
*(Explain it in your own words. 자신의 언어로 설명해보세요.)*

`dict` methods handle safe lookup, insertion, and removal for key-value data (`.get()`, `.update()`, `.pop()`), while `set` methods provide mathematical set operations — union, intersection, difference — as both named methods and operators. JS/TS's `Map` and `Set` cover similar ground, but far less of a JS/TS developer's daily code actually uses them, since plain `Object` and `Array` usually do the job instead — in Python, `dict` and `set` are the default, first-choice tools.

`dict` 메서드는 키-값 데이터의 안전한 조회, 삽입, 제거를 다루고(`.get()`, `.update()`, `.pop()`), `set` 메서드는 합집합, 교집합, 차집합 같은 수학적 집합 연산을 메서드와 연산자 둘 다로 제공한다. JS/TS의 `Map`과 `Set`도 비슷한 영역을 다루지만, JS/TS 개발자의 일상 코드에서는 보통 평범한 `Object`와 `Array`가 그 역할을 대신하기 때문에 훨씬 덜 쓰인다 — Python에서는 `dict`와 `set`이 기본이자 첫 번째 선택 도구다.

| | JS/TS | Python |
|---|---|---|
| Safe lookup with fallback 대체값을 갖는 안전한 조회 | `obj.x ?? default` | `d.get("x", default)` |
| Add/update multiple keys 여러 키 추가·수정 | `Object.assign(obj, other)` | `d.update(other)` |
| Set only if missing 없을 때만 설정 | `obj.x ??= default` | `d.setdefault("x", default)` |
| Remove a key 키 제거 | `delete obj.x` | `del d["x"]` or `d.pop("x")` |
| Get all keys/values 모든 키·값 가져오기 | `Object.keys(obj)` / `Object.values(obj)` | `d.keys()` / `d.values()` |
| Set union 합집합 | `new Set([...a, ...b])` | `a` \| `b` or `a.union(b)` |
| Set intersection 교집합 | `new Set([...a].filter(x => b.has(x)))` | `a & b` or `a.intersection(b)` |
| Membership check 멤버십 확인 | `set.has(x)` | `x in s` |

The single most valuable habit here: `d[key]` and `d.pop(key)` both raise an exception if the key is missing, while `d.get(key, default)` and `d.pop(key, default)` never do — reaching for the `default` versions by default (pun intended) prevents an entire category of `KeyError` crashes.

여기서 가장 값진 습관 하나: `d[key]`와 `d.pop(key)`는 키가 없으면 둘 다 예외를 던지지만, `d.get(key, default)`와 `d.pop(key, default)`는 절대 그러지 않는다 — 기본적으로 `default` 버전을 사용하는 습관을 들이면 `KeyError` 크래시라는 카테고리 전체를 예방할 수 있다.

## Why do we use it? (왜 사용하나요?)
*(When is it useful? 언제 유용한가요?)*

Real data is rarely guaranteed to have every key present — a config dict might be missing an optional setting, a record from an API might omit a field. `.get()` with a default turns "is this key even there?" from a crash risk into a one-line non-issue, which is why it is reached for far more often than direct `d[key]` access in production code.

실제 데이터는 모든 키가 반드시 있다고 보장되지 않는다 — 설정 딕셔너리에 선택적 설정이 빠져 있을 수도 있고, API에서 온 레코드가 필드를 생략할 수도 있다. 기본값을 갖는 `.get()`은 "이 키가 있긴 한가?"라는 질문을 크래시 위험에서 한 줄짜리 사소한 일로 바꿔준다. 그래서 실무 코드에서는 직접 `d[key]`에 접근하는 것보다 훨씬 자주 쓰인다.

Set operations turn what would be nested `for` loops with `if` checks into single expressions: comparing two collections for overlap, uniqueness, or containment is a well-known mathematical operation, and Python exposes it directly instead of making you reimplement it by hand every time.

셋 연산은 `if` 검사가 있는 중첩 `for` 반복문이 될 뻔한 것을 단일 표현식으로 바꿔준다: 두 컬렉션의 겹침, 고유성, 포함 관계를 비교하는 것은 잘 알려진 수학적 연산이며, Python은 매번 직접 재구현하게 만드는 대신 이를 직접 노출한다.

## When is it used in Business Analytics? (Business Analytics에서는 언제 쓰이나요?)
*(Real-world use case 실제 사용 사례)*

Reconciling two data sources — "which IDs are in system A but not system B", "which categories overlap between two campaigns" — is exactly what `.difference()` and `.intersection()` are for, and answering it takes one line instead of a nested loop with manual bookkeeping.

두 데이터 소스를 대조하는 일 — "시스템 A에는 있지만 시스템 B에는 없는 ID는?", "두 캠페인 사이에 겹치는 카테고리는?" — 은 정확히 `.difference()`와 `.intersection()`이 하는 일이며, 수동으로 관리하는 중첩 반복문 대신 한 줄로 답할 수 있다.

Dictionaries are also the natural shape for lookup tables built during data cleaning — mapping a raw category code to a readable label with `.get(code, "Unknown")` is a pattern that appears in almost every real cleaning script, and `dict.setdefault()` is the classic tool for grouping records by key without a `collections.defaultdict` import.

딕셔너리는 데이터 정리 중에 만드는 조회 테이블에도 자연스러운 형태다 — `.get(code, "Unknown")`으로 원본 카테고리 코드를 읽기 좋은 라벨에 매핑하는 패턴은 거의 모든 실제 정리 스크립트에 등장하며, `dict.setdefault()`는 `collections.defaultdict` import 없이 레코드를 키별로 그룹화하는 고전적인 도구다.

# 📝 Syntax (문법)

## Basic Syntax (기본 문법)

````python
d = {"name": "booki", "role": "BA-in-training"}

# Safe lookup — .get() never raises, unlike d[key]
# 안전한 조회 — .get()은 d[key]와 달리 절대 예외를 던지지 않음
print(d.get("name"))
print(d.get("missing", "N/A"))

# Add/update, then remove safely
# 추가·수정 후 안전하게 제거
d["team"] = "Analytics"
d.pop("role", None)
print(d)

s = {1, 2, 3}

# Set algebra reads like math
# 셋 연산은 수학처럼 읽힘
other = {2, 3, 4}
print(s & other)       # intersection
print(s | other)         # union
print(s - other)           # difference
````

## Common Variations (자주 쓰는 변형)

````python
# .update() accepts a dict OR keyword arguments
# .update()는 딕셔너리 또는 키워드 인자를 받을 수 있음
config = {"theme": "light"}
config.update({"lang": "en"})
config.update(version=2)
print(config)

# .setdefault() — add a key only if it is missing, otherwise leave it alone
# .setdefault() — 없을 때만 키를 추가, 있으면 그대로 둠
counts = {"a": 5}
counts.setdefault("a", 0)      # "a" already exists, stays 5
counts.setdefault("b", 0)       # "b" is new, becomes 0
print(counts)

# .items() is the most common way to loop over a dict
# .items()는 딕셔너리를 순회하는 가장 흔한 방법
for key, value in config.items():
    print(key, "->", value)

# operator vs method for set operations — both do the same thing
# 셋 연산의 연산자 vs 메서드 — 둘 다 동일하게 동작
a, b = {1, 2, 3}, {2, 3, 4}
print(a.union(b) == (a | b))
print(a.intersection(b) == (a & b))
````

# 🧪 Small Examples (예제)

## Example 1 — Dict Access & Modify (딕셔너리 조회·수정)

````python
d = {"a": 1, "b": 2, "c": 3}

# Direct access — raises KeyError if the key is missing
# 직접 접근 — 키가 없으면 KeyError
print(d["a"])

# .get() — returns None instead of raising
# .get() — 예외 대신 None 반환
print(d.get("a"))
print(d.get("z"))                # None, no error

# .get(key, default) — a custom fallback instead of None
# .get(key, default) — None 대신 커스텀 대체값
print(d.get("z", 0))

# Assignment — adds a new key or overwrites an existing one
# 대입 — 새 키를 추가하거나 기존 키를 덮어씀
d["d"] = 4
print(d)

# .update() — merge many keys at once, from a dict...
# .update() — 딕셔너리로부터 여러 키를 한 번에 병합...
d.update({"e": 5, "f": 6})
print(d)

# ...or from keyword arguments
# ...또는 키워드 인자로부터
d.update(g=7, h=8)
print(d)

# .setdefault() — set a key ONLY if it is missing; existing keys are untouched
# .setdefault() — 없을 때만 키 설정; 기존 키는 건드리지 않음
d.setdefault("a", 99)          # "a" exists (=1), stays 1
d.setdefault("z", 0)             # "z" is new, becomes 0
print(d)
````

## Example 2 — Dict Remove & Iterate (딕셔너리 제거·순회)

````python
d = {"a": 1, "b": 2, "c": 3}

# .pop(key) — removes and returns the value (raises KeyError if missing)
# .pop(key) — 값을 제거하고 반환 (없으면 KeyError)
removed = d.pop("a")
print(removed, d)

# .pop(key, default) — the safe version, like list's version
# .pop(key, default) — list의 버전처럼 안전한 버전
print(d.pop("z", None))          # None, no error

# .popitem() — removes and returns the LAST inserted (key, value) pair
# .popitem() — 마지막으로 삽입된 (key, value) 쌍을 제거하고 반환
d["d"] = 4
print(d.popitem())

# del — removes a key without returning it
# del — 반환 없이 키를 제거
d = {"a": 1, "b": 2, "c": 3}
del d["b"]
print(d)

# .clear() — remove everything
# .clear() — 전체 제거
temp = {"x": 1}
temp.clear()
print(temp)

# Iterating — .keys() / .values() / .items()
# 순회 — .keys() / .values() / .items()
d = {"a": 1, "b": 2, "c": 3}
print(list(d.keys()))
print(list(d.values()))
print(list(d.items()))

# Membership checks — "in" checks KEYS by default, not values
# 멤버십 확인 — "in"은 기본적으로 값이 아니라 키를 확인
print("a" in d)                 # checks keys (키 확인)
print(1 in d.values())            # checks values explicitly (값을 명시적으로 확인)

# .copy() — a shallow copy, same one-level-deep caveat as list.copy()
# .copy() — 얕은 복사, list.copy()와 같은 한 단계까지만 복사되는 주의사항
d_copy = d.copy()
d_copy["a"] = 999
print(d, d_copy)                  # original untouched (원본은 그대로)
````

## Example 3 — Set Basics (셋 기초)

````python
s = {1, 2, 3}

# add() — insert one element (no effect if it is already present)
# add() — 요소 하나 추가 (이미 있으면 아무 효과 없음)
s.add(4)
s.add(2)                # already present, set stays the same (이미 있으므로 셋은 그대로)
print(s)

# remove() — delete an element (raises KeyError if missing)
# remove() — 요소 삭제 (없으면 KeyError)
s.remove(1)
print(s)

# discard() — same as remove(), but never raises if the value is missing
# discard() — remove()와 같지만 값이 없어도 절대 예외를 던지지 않음
s.discard(99)             # 99 was never there — no error (99는 애초에 없었음 — 에러 없음)
print(s)

# pop() — removes and returns an ARBITRARY element (sets have no order)
# pop() — 임의의 요소를 제거하고 반환 (셋은 순서가 없음)
arbitrary = s.pop()
print(arbitrary, s)

# copy() — a shallow copy, independent from the original
# copy() — 얕은 복사, 원본과 독립적
original = {1, 2, 3}
duplicate = original.copy()
duplicate.add(99)
print(original, duplicate)

# clear() — remove everything
# clear() — 전체 제거
s.clear()
print(s)
````

## Example 4 — Set Algebra (셋 집합 연산)

````python
a = {1, 2, 3}
b = {2, 3, 4}

# union — everything in either set (method and operator are equivalent)
# union — 둘 중 하나에라도 있는 모든 것 (메서드와 연산자는 동등)
print(a.union(b))
print(a | b)

# intersection — only what is in BOTH sets
# intersection — 둘 다에 있는 것만
print(a.intersection(b))
print(a & b)

# difference — in `a` but NOT in `b` (order matters!)
# difference — `a`에는 있지만 `b`에는 없는 것 (순서가 중요!)
print(a.difference(b))
print(a - b)
print(b - a)               # different result — the other direction (다른 결과 — 반대 방향)

# symmetric_difference — in exactly ONE of the two sets, never both
# symmetric_difference — 둘 중 정확히 하나에만 있는 것, 둘 다는 아님
print(a.symmetric_difference(b))
print(a ^ b)

# A practical data-quality question, answered with set algebra
# 셋 연산으로 답하는 실전 데이터 품질 질문
file_a_ids = {"C001", "C002", "C003"}
file_b_ids = {"C002", "C003", "C004"}
print("only in file A:", file_a_ids - file_b_ids)
````

# ⚠️ Common Mistakes (자주 하는 실수)

### Mistake 1 — Using `d[key]` when the key might not exist
`d[key]` raises `KeyError` immediately if the key is missing — a crash that often surfaces far from where the dict was actually built, making it confusing to debug.

`d[key]`는 키가 없으면 즉시 `KeyError`를 던진다 — 이 크래시는 딕셔너리가 실제로 만들어진 곳과 멀리 떨어진 곳에서 나타나는 경우가 많아 디버깅하기 혼란스럽다.

✅ **Fix / 해결법:**  
Use `.get(key, default)` whenever the key's presence is not guaranteed.  
키의 존재가 보장되지 않을 때는 항상 `.get(key, default)`를 사용하자.
```python
d = {"a": 1}
print(d.get("missing", "N/A"))   # ✅ "N/A", no crash
```

### Mistake 2 — Assuming `x in d` checks values
`in` on a dict checks **keys** by default, not values. `5 in {"a": 5}` is `False`, because `"a"` is the key and `5` is only a value — a common surprise for anyone expecting `in` to search the "content" of the dict.

딕셔너리에서 `in`은 기본적으로 값이 아니라 **키**를 확인한다. `5 in {"a": 5}`는 `False`다, `"a"`가 키이고 `5`는 값일 뿐이기 때문이다 — `in`이 딕셔너리의 "내용물"을 검색할 거라 기대하는 사람에게 흔한 놀라움이다.

✅ **Fix / 해결법:**  
Use `x in d` to check keys (the default), and `x in d.values()` explicitly when you mean values.  
키를 확인하려면 `x in d`(기본값)를, 값을 확인하려면 명시적으로 `x in d.values()`를 사용하자.
```python
d = {"a": 5}
print(5 in d)                # False — checks keys
print(5 in d.values())         # ✅ True — checks values
```

### Mistake 3 — Using `.remove()` on a set when the value might be missing
Just like lists, a set's `.remove()` raises `KeyError` if the value is not present. `.discard()` does the exact same job but never raises — it simply does nothing if the value was never there.

리스트와 마찬가지로, 셋의 `.remove()`는 값이 없으면 `KeyError`를 던진다. `.discard()`는 정확히 같은 일을 하지만 절대 예외를 던지지 않는다 — 값이 애초에 없었다면 그냥 아무 일도 하지 않는다.

✅ **Fix / 해결법:**  
Use `.discard()` when you are not sure the value is present, and `.remove()` only when a missing value should be treated as an error.  
값이 있는지 확신이 없다면 `.discard()`를, 값이 없는 게 에러여야 할 때만 `.remove()`를 사용하자.
```python
s = {1, 2, 3}
s.discard(99)   # ✅ no error, even though 99 was never in the set
```

# 💡 Tips (팁)
Useful tips or shortcuts (유용한 팁이나 단축법)

- `d.get(key, default)` and `d.setdefault(key, default)` look similar but do different things: `.get()` never modifies the dict, while `.setdefault()` actually ADDS the key with the default value if it was missing.  
`d.get(key, default)`와 `d.setdefault(key, default)`는 비슷해 보이지만 다른 일을 한다: `.get()`은 딕셔너리를 절대 수정하지 않지만, `.setdefault()`는 키가 없었다면 실제로 기본값과 함께 키를 추가한다.
- `set` membership checks (`x in my_set`) are O(1), the same reason covered in Section 2 — converting a list to a set first is a common speed-up for repeated lookups.  
`set` 멤버십 확인(`x in my_set`)은 O(1)이다, Section 2에서 다룬 것과 같은 이유다 — 반복적인 조회를 위해 먼저 리스트를 셋으로 변환하는 것은 흔한 속도 개선 방법이다.
- `a - b` and `b - a` give different results for `.difference()` — always double check which set is "the source" and which is "what to exclude."  
`.difference()`에서 `a - b`와 `b - a`는 다른 결과를 준다 — 어느 셋이 "원본"이고 어느 셋이 "제외할 것"인지 항상 다시 확인하자.
- `collections.defaultdict` is often a cleaner alternative to repeated `.setdefault()` calls when grouping many items by key — worth remembering once Section 7's standard library arrives.  
키별로 많은 항목을 그룹화할 때는 반복적인 `.setdefault()` 호출보다 `collections.defaultdict`가 더 깔끔한 대안인 경우가 많다 — Section 7의 표준 라이브러리에서 다시 만나면 기억해두자.
- `dict(zip(keys, values))` is a fast, common way to build a dict from two parallel lists — no manual loop needed.  
`dict(zip(keys, values))`는 두 개의 병렬 리스트로부터 딕셔너리를 만드는 빠르고 흔한 방법이다 — 수동 반복문이 필요 없다.

# 💼 Business Example (비즈니스 예제)
*How would a Business Analyst use this? BA는 이것을 실제로 어떻게 사용할까요?*

**Scenario (시나리오):** Two departments each sent a list of active project codes. You need to reconcile them — find overlaps, find what is missing from each side — and build a lookup table to translate project codes into readable names, all without pandas.

**시나리오:** 두 부서가 각각 활성 프로젝트 코드 리스트를 보냈다. 겹치는 것과 각 쪽에서 빠진 것을 찾아 대조하고, 프로젝트 코드를 읽기 좋은 이름으로 바꾸는 조회 테이블을 만들어야 한다 — pandas 없이.

**To do (할 일 목록):**
- [x] Use set algebra to find codes in both departments, and codes unique to each.  
셋 연산으로 두 부서 모두에 있는 코드와 각 부서에만 있는 코드를 찾는다.
- [x] Build a code-to-name lookup `dict` and use `.get()` with a fallback for unknown codes.   
코드→이름 조회 `dict`를 만들고 알 수 없는 코드에는 `.get()`의 대체값을 사용한다.
- [x] Use `.setdefault()` to group project codes by department without a nested loop. `. setdefault()`로 중첩 반복문 없이 프로젝트 코드를 부서별로 그룹화한다.

````python
marketing_codes = {"PRJ-101", "PRJ-102", "PRJ-104"}
sales_codes = {"PRJ-102", "PRJ-103", "PRJ-104"}

shared = marketing_codes & sales_codes
marketing_only = marketing_codes - sales_codes
sales_only = sales_codes - marketing_codes

print("Shared projects (공유 프로젝트):", shared)
print("Marketing only (마케팅 전용):", marketing_only)
print("Sales only (영업 전용):", sales_only)

# Lookup table — .get() falls back cleanly for any code not listed here
# 조회 테이블 — 여기 없는 코드는 .get()이 깔끔하게 대체값으로 처리
project_names = {
    "PRJ-101": "Website Relaunch",
    "PRJ-102": "Q3 Campaign",
    "PRJ-103": "Sales Enablement",
}

all_codes = marketing_codes | sales_codes
print("-" * 40)
for code in sorted(all_codes):
    name = project_names.get(code, "Unnamed Project")
    print(f"{code}: {name}")

# Group codes by department using setdefault() — no defaultdict import needed
# setdefault()로 부서별 코드 그룹화 — defaultdict import 불필요
by_department = {}
for code in marketing_codes:
    by_department.setdefault("marketing", []).append(code)
for code in sales_codes:
    by_department.setdefault("sales", []).append(code)

print("-" * 40)
for dept, codes in by_department.items():
    print(dept, "->", sorted(codes))
````

*Write today's concept in 3~5 sentences. 오늘 배운 개념을 3~5문장으로 정리해보세요.*

`dict` methods center on one critical distinction: `d[key]` and `d.pop(key)` raise `KeyError` when a key is missing, while `.get(key, default)` and `.pop(key, default)` never do — reaching for the safe versions by default prevents an entire category of crashes. `.update()` merges many keys at once, and `.setdefault()` adds a key only if it is not already there, which makes it the classic building block for grouping records without importing `collections.defaultdict`. `set` methods mirror this same safety pattern (`.discard()` vs `.remove()`) and add proper mathematical set algebra — `union`, `intersection`, `difference`, and `symmetric_difference` — available as both named methods and operators (`|`, `&`, `-`, `^`). Both `in` on a dict and membership checks on a set are O(1), making them the fast, idiomatic choice for "does this exist?" questions over collections of any real size. Every pattern here — safe lookups, set comparisons, grouping — reappears directly in pandas and in the `collections` module covered next.

`dict` 메서드는 하나의 중요한 구분을 중심으로 돌아간다: `d[key]`와 `d.pop(key)`는 키가 없으면 `KeyError`를 던지지만, `.get(key, default)`와 `.pop(key, default)`는 절대 그러지 않는다 — 기본적으로 안전한 버전을 사용하는 습관이 크래시라는 카테고리 전체를 예방한다. `.update()`는 여러 키를 한 번에 병합하고, `.setdefault()`는 아직 없을 때만 키를 추가한다. 그래서 `collections.defaultdict`를 import하지 않고도 레코드를 그룹화하는 고전적인 빌딩 블록이 된다. `set` 메서드는 같은 안전 패턴(`.discard()` vs `.remove()`)을 그대로 반영하며 제대로 된 수학적 집합 연산 — `union`, `intersection`, `difference`, `symmetric_difference` — 을 메서드와 연산자(`|`, `&`, `-`, `^`) 둘 다로 제공한다. 딕셔너리의 `in`과 셋의 멤버십 확인은 둘 다 O(1)이라, 어느 정도 규모의 컬렉션에서든 "이게 존재하나?"라는 질문에 빠르고 Pythonic하게 답하는 선택이다. 여기 나온 모든 패턴 — 안전한 조회, 셋 비교, 그룹화 — 은 pandas와 다음에 다룰 `collections` 모듈에서 그대로 다시 등장한다.

**Q1.** What is the difference between `d["missing_key"]` and `d.get("missing_key")` when the key does not exist?  
키가 존재하지 않을 때 `d["missing_key"]`와 `d.get("missing_key")`는 어떻게 다른가요?

`d["key"]` raises a KeyError if the key is missing. `d.get("key")` returns None, or the provided default value.  
`d["key"]`는 키가 없으면 KeyError를 발생시킵니다. `d.get("key")`는 None 또는 지정한 기본값을 반환합니다.

**Q2.** What is the difference between `d.get(key, default)` and `d.setdefault(key, default)`?  
`d.get(key, default)`와 `d.setdefault(key, default)`는 어떻게 다른가요?

`get()` returns the value (or default) without modifying the dictionary. `setdefault()` returns the value if the key exists; otherwise, it adds the key with the default value and returns that value.  
`get()`은 딕셔너리를 변경하지 않고 값을 반환합니다. `setdefault()`는 키가 없으면 딕셔너리에 키와 기본값을 추가한 뒤 그 값을 반환합니다.

**Q3.** Does `5 in {"a": 5}` return `True` or `False`, and why?  
`5 in {"a": 5}`는 `True`를 반환하나요, `False`를 반환하나요? 그 이유는 무엇인가요?

False. The in operator checks keys, not values. Use `d.values()` to search values.  
False입니다. in은 키를 검사합니다. 값을 확인하려면 `d.values()`를 사용해야 합니다.

**Q4.** Given `a = {1, 2, 3}` and `b = {2, 3, 4}`, what is the difference between `a - b` and `b - a`?  
`a = {1, 2, 3}`이고 `b = {2, 3, 4}`일 때, `a - b`와 `b - a`는 어떻게 다른가요?

a - b returns `{1}` because 1 exists only in a. b - a returns `{4}` because 4 exists only in b.  
a - b는 a에만 있는 `{1}`을 반환하고, b - a는 b에만 있는 `{4}`를 반환합니다.

**Q5.** When should you use `set.discard()` instead of `set.remove()`?  
언제 `set.remove()` 대신 `set.discard()`를 사용해야 하나요?

Use `discard()` when the element might not exist and you don't want a KeyError.  
제거할 요소가 없을 수도 있고, 에러를 발생시키고 싶지 않을 때 `discard()`를 사용합니다.
