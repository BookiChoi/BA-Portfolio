# 🧠 Concept (개념)

## What is it? (이게 뭔가요?)
*(Explain it in your own words. 자신의 언어로 설명해보세요.)*

The standard library is the large collection of modules that ships with every Python installation — no `pip install` required, just `import`. It covers counting and grouping (`collections`), combinatorics and function composition (`itertools`, `functools`), math (`math`), randomness (`random`), calendar math (`datetime`), filesystem paths (`os`, `pathlib`), pattern matching (`re`), and data interchange (`json`). JS/TS has no single equivalent — some of this lives in the language itself, some in Node's built-in modules, and a lot of it simply does not exist without a third-party package.

표준 라이브러리는 모든 Python 설치에 기본으로 포함되는 방대한 모듈 모음이다 — `pip install` 없이 `import`만 하면 된다. 카운팅과 그룹화(`collections`), 조합론과 함수 조합(`itertools`, `functools`), 수학(`math`), 난수(`random`), 달력 계산(`datetime`), 파일시스템 경로(`os`, `pathlib`), 패턴 매칭(`re`), 데이터 교환(`json`)을 다룬다. JS/TS에는 이와 동일한 단일 대응물이 없다 — 일부는 언어 자체에 있고, 일부는 Node의 내장 모듈에 있으며, 상당수는 서드파티 패키지 없이는 아예 존재하지 않는다.

| | JS/TS | Python |
|---|---|---|
| Count occurrences 개수 세기 | manual loop + object | `collections.Counter(items)` |
| Group by key 키로 그룹화 | manual loop + object | `collections.defaultdict(list)` |
| Double-ended queue 양방향 큐 | `Array` (slow shift/unshift) | `collections.deque` (O(1) both ends) |
| Current date/time 현재 날짜·시간 | `new Date()` | `datetime.now()` |
| Format a date 날짜 포맷 | `date.toLocaleDateString()` | `dt.strftime("%Y-%m-%d")` |
| Regex match anywhere 어디서든 정규식 매칭 | `str.match(/\d+/)` | `re.findall(r"\d+", str)` |
| JSON encode/decode | `JSON.stringify()` / `JSON.parse()` | `json.dumps()` / `json.loads()` |
| Cartesian product 데카르트 곱 | nested `for` loops | `itertools.product(a, b)` |

Given the module count, this section reads less like new syntax and more like a shopping catalog: once you know `import <module>` is how Python organizes optional tools, the question shifts from "how do I write this?" to "does the standard library already have this?" — and for counting, grouping, combinatorics, or date math, the answer is almost always yes.

모듈 개수를 생각하면, 이 섹션은 새로운 문법이라기보다 쇼핑 카탈로그에 가깝다: `import <module>`이 Python이 선택적 도구를 정리하는 방식이라는 걸 알고 나면, 질문은 "이걸 어떻게 짜지?"에서 "표준 라이브러리에 이미 있나?"로 바뀐다 — 카운팅, 그룹화, 조합론, 날짜 계산이라면 대답은 거의 항상 "있다"이다.

## Why do we use it? (왜 사용하나요?)
*(When is it useful? 언제 유용한가요?)*

Standard library functions are battle-tested and optimized in ways a quick hand-rolled version usually is not — `Counter` uses an efficient C-level dict subclass, `deque` is a proper doubly-linked structure with true O(1) operations on both ends, and `re` runs a compiled regex engine instead of manual string scanning.

표준 라이브러리 함수는 급하게 직접 짠 버전이 보통 갖추지 못하는 방식으로 검증되고 최적화되어 있다 — `Counter`는 효율적인 C 레벨 dict 서브클래스를 사용하고, `deque`는 양쪽 끝에서 진짜 O(1) 연산을 하는 제대로 된 이중 연결 구조이며, `re`는 수동 문자열 스캔 대신 컴파일된 정규식 엔진을 실행한다.

They are also a shared vocabulary — `defaultdict(list)` instantly tells another Python developer "this groups things by key," while an equivalent hand-rolled `if key not in d: d[key] = []` needs a moment longer to parse.

이것들은 공유된 어휘이기도 하다 — `defaultdict(list)`는 다른 Python 개발자에게 "이건 키로 항목을 그룹화하는 것"이라고 즉시 알려주지만, 동등한 `if key not in d: d[key] = []`를 직접 짠 코드는 파악하는 데 시간이 조금 더 걸린다.

## When is it used in Business Analytics? (Business Analytics에서는 언제 쓰이나요?)
*(Real-world use case 실제 사용 사례)*

Extracting structured values from raw text — an order ID from a log line, a date from a filename — is exactly what `re` is for, and the result usually needs `datetime.strptime()` right afterward to become a real date instead of a string. `json` is the format most APIs speak, so reading external data almost always starts with `json.loads()` before it becomes a `DataFrame`.

로그 줄에서 주문 ID를, 파일명에서 날짜를 추출하는 것처럼 원본 텍스트에서 구조화된 값을 뽑아내는 일은 정확히 `re`가 하는 일이며, 그 결과는 보통 문자열이 아니라 진짜 날짜가 되기 위해 바로 뒤에 `datetime.strptime()`이 필요하다. `json`은 대부분의 API가 사용하는 포맷이라, 외부 데이터를 읽는 작업은 `DataFrame`이 되기 전에 거의 항상 `json.loads()`로 시작한다.

`Counter` answers "what are the most common values in this column?" without pandas at all, and `itertools.product()` is the standard way to generate every combination of test scenarios or filter parameters for a report.

`Counter`는 pandas 없이도 "이 컬럼에서 가장 흔한 값은 무엇인가?"에 답하며, `itertools.product()`는 리포트를 위한 테스트 시나리오나 필터 파라미터의 모든 조합을 생성하는 표준적인 방법이다.

## Example 1 — Math & Random (수학·랜덤)

````python
import math
import random

# Rounding family — each rounds differently!
# 반올림 계열 — 각각 다르게 반올림함!
print(math.ceil(3.2), math.ceil(-3.2))          # always rounds UP (항상 올림)
print(math.floor(3.9), math.floor(-3.9))          # always rounds DOWN (항상 내림)
print(math.trunc(3.9), math.trunc(-3.9))            # always rounds toward zero (항상 0 방향으로 자름)

# Powers, roots, and logs
# 거듭제곱, 제곱근, 로그
print(math.sqrt(16), math.isqrt(16))                 # float vs exact int (float vs 정확한 int)
print(math.log(100, 10), math.log2(8), math.log10(1000))

# Trigonometry — radians by default; degrees()/radians() convert between them
# 삼각함수 — 기본은 라디안; degrees()/radians()가 서로 변환
print(math.sin(math.pi / 2))
print(math.degrees(math.pi), math.radians(180))

# Useful constants
# 유용한 상수들
print(math.pi, math.e, math.inf)

# Combinatorics — factorial, combinations (order doesn't matter), permutations (order matters)
# 조합론 — 팩토리얼, 조합(순서 무관), 순열(순서 있음)
print(math.factorial(5))
print(math.comb(10, 3), math.perm(10, 3))

# gcd / lcm / fsum — exact math that avoids float rounding traps
# gcd / lcm / fsum — float 반올림 함정을 피하는 정확한 계산
print(math.gcd(12, 8), math.lcm(4, 6))
print(math.fsum([0.1, 0.2, 0.3]))            # exact 0.6, unlike 0.1+0.2+0.3

# Validity checks for special float values, and the classic 3-4-5 triangle
# 특수 float 값 유효성 확인, 그리고 대표적인 3-4-5 삼각형
print(math.isnan(math.nan), math.isinf(math.inf), math.isfinite(42))
print(math.hypot(3, 4))

random.seed(42)     # fixes the sequence so results are reproducible here (여기서 재현 가능하도록 시드 고정)

print(random.randint(1, 6))                    # 1~6 inclusive (양 끝 포함)
print(random.choice(["a", "b", "c"]))
print(random.sample(range(10), k=3))              # 3 unique values, no repeats (중복 없는 3개)
lst = [1, 2, 3, 4, 5]
random.shuffle(lst)                                 # shuffles in place, returns None (제자리에서 섞음, None 반환)
print(lst)
````

## Example 2 — Collections I: Counter & defaultdict (컬렉션 I)

````python
from collections import Counter, defaultdict

# Counter — counts occurrences automatically
# Counter — 자동으로 개수를 세어줌
text_counts = Counter("mississippi")
print(text_counts)
print(text_counts.most_common(2))             # top 2 most frequent (가장 빈도 높은 2개)
print(text_counts["z"])                         # 0 for a missing key, no KeyError (없는 키는 0, KeyError 없음)

# Counter arithmetic — treats counts like a vector
# Counter 산술 연산 — 개수를 벡터처럼 취급
c1 = Counter(a=3, b=1)
c2 = Counter(a=1, b=2)
print(c1 + c2)              # add counts (개수를 더함)
print(c1 - c2)                # subtract, keeping only positive results (빼되 양수만 유지)

# defaultdict — supplies a default value instead of raising KeyError
# defaultdict — KeyError 대신 기본값을 제공
dd = defaultdict(int)
dd["x"] += 1                    # works even though "x" never existed (x가 없었어도 동작함)
print(dd)

# The classic defaultdict pattern: grouping items by a key
# defaultdict의 대표 패턴: 키로 항목 그룹화
groups = defaultdict(list)
for word in ["apple", "ant", "banana", "bat", "cherry"]:
    groups[word[0]].append(word)
print(dict(groups))
````

## Example 3 — Collections II: deque, namedtuple & ChainMap (컬렉션 II)

````python
from collections import deque, namedtuple, ChainMap

# deque — O(1) append/pop from BOTH ends, unlike list
# deque — list와 달리 양쪽 끝에서 O(1) 추가/제거
dq = deque([1, 2, 3])
dq.append(4)                # add to the right (오른쪽에 추가)
dq.appendleft(0)              # add to the left (왼쪽에 추가)
print(dq)
print(dq.popleft())              # O(1) — list.pop(0) would be O(n) instead
dq.rotate(1)                       # shift every element right by 1, wrapping around (모든 요소를 오른쪽으로 1칸씩, 순환)
print(dq)

# BFS pattern — the most common reason to reach for a deque
# BFS 패턴 — deque를 쓰는 가장 흔한 이유
graph = {"A": ["B", "C"], "B": ["D"], "C": ["D"], "D": []}
visited, queue = set(), deque(["A"])
order = []
while queue:
    node = queue.popleft()
    if node not in visited:
        visited.add(node)
        order.append(node)
        queue.extend(graph[node])
print(order)

# namedtuple — a lightweight, readable alternative to a plain tuple
# namedtuple — 평범한 튜플의 가볍고 읽기 좋은 대안
Point = namedtuple("Point", ["x", "y"])
p = Point(1, 2)
print(p.x, p[0])              # access by name OR by index (이름 또는 인덱스로 접근)
print(p._replace(x=10))          # a new Point; the original is still immutable (새 Point; 원본은 여전히 불변)

# ChainMap — search multiple dicts as if they were one, without merging them
# ChainMap — 병합하지 않고도 여러 딕셔너리를 하나처럼 검색
defaults = {"theme": "light", "size": 10}
user_prefs = {"theme": "dark"}
settings = ChainMap(user_prefs, defaults)
print(settings["theme"])            # found in user_prefs first (user_prefs에서 먼저 찾음)
print(settings["size"])               # falls through to defaults (defaults로 넘어감)
````

## Example 4 — Itertools & Functools (이터툴즈·펑크툴즈)

````python
import itertools as it
from functools import reduce, partial, lru_cache

# combinations / permutations — order doesn't matter vs order matters
# combinations / permutations — 순서 무관 vs 순서 있음
print(list(it.combinations([1, 2, 3], 2)))
print(list(it.permutations([1, 2, 3], 2)))

# product — the itertools replacement for nested for-loops
# product — 중첩 for문을 대체하는 itertools
print(list(it.product([0, 1], repeat=2)))

# chain — flatten several iterables into one, without building an intermediate list
# chain — 중간 리스트를 만들지 않고 여러 이터러블을 하나로 이어붙임
print(list(it.chain([1, 2], [3, 4], [5])))

# accumulate — running totals (or any running operation)
# accumulate — 누적 합계 (또는 임의의 누적 연산)
print(list(it.accumulate([1, 2, 3, 4])))                  # running sum (누적 합)
print(list(it.accumulate([1, 2, 3, 4], max)))                # running max (누적 최댓값)

# groupby — groups CONSECUTIVE matching items; sort first if the data is not already grouped
# groupby — 연속된 일치 항목을 묶음; 미리 정렬되어 있지 않다면 먼저 정렬해야 함
words = sorted(["apple", "ant", "banana", "bat"], key=lambda w: w[0])
for letter, group in it.groupby(words, key=lambda w: w[0]):
    print(letter, list(group))

# reduce — collapse a whole iterable into a single value
# reduce — 이터러블 전체를 값 하나로 축약
print(reduce(lambda acc, x: acc + x, [1, 2, 3, 4]))
print(reduce(max, [3, 1, 4, 1, 5]))

# partial — freeze some arguments of a function ahead of time
# partial — 함수의 일부 인자를 미리 고정
def power(base, exp):
    return base ** exp
square = partial(power, exp=2)
print(square(5))

# lru_cache — memoization, dramatically speeds up repeated recursive calls
# lru_cache — 메모이제이션, 반복되는 재귀 호출 속도를 크게 향상
@lru_cache(maxsize=None)
def fib(n):
    return n if n < 2 else fib(n - 1) + fib(n - 2)
print(fib(30))
print(fib.cache_info())
````

## Example 5 — Datetime & Operator (날짜시간·연산자)

````python
from datetime import datetime, timedelta
from operator import itemgetter
import operator

# Fixed dates for reproducible output (not datetime.now(), which changes every run)
# 재현 가능한 출력을 위한 고정된 날짜 (매번 바뀌는 datetime.now() 대신)
dt = datetime(2024, 1, 15, 10, 30, 0)
print(dt.year, dt.month, dt.day, dt.weekday())        # weekday(): 0=Monday (월요일)

# Formatting — strftime turns a datetime INTO a string
# 포맷팅 — strftime은 datetime을 문자열로 바꿈
print(dt.strftime("%Y-%m-%d %H:%M:%S"))

# Parsing — strptime turns a string INTO a datetime (the reverse)
# 파싱 — strptime은 문자열을 datetime으로 바꿈 (반대 방향)
parsed = datetime.strptime("2024-03-01", "%Y-%m-%d")
print(parsed)

# Arithmetic with timedelta
# timedelta를 이용한 날짜 연산
one_week_later = dt + timedelta(days=7)
print(one_week_later)
diff = datetime(2024, 12, 31) - dt
print(diff.days, "days remaining in the year")

# operator module — function versions of +, -, ==, etc., handy as a sort/reduce key
# operator 모듈 — +, -, == 등의 함수 버전, 정렬/reduce의 key로 유용
print(operator.add(3, 4), operator.mul(3, 4))

records = [{"name": "Bob", "age": 30}, {"name": "Ann", "age": 25}]
print(sorted(records, key=itemgetter("age")))          # faster than an equivalent lambda (동등한 lambda보다 빠름)

rows = [(1, "B", 3), (2, "A", 1)]
print(sorted(rows, key=itemgetter(1)))                    # sort by the 2nd element of each tuple (각 튜플의 2번째 요소 기준 정렬)
````

# ⚠️ Common Mistakes (자주 하는 실수)

### Mistake 1 — Running `itertools.groupby()` on unsorted data
`groupby()` only groups *consecutive* matching items — it does not scan the whole iterable for every match like a `dict`-based grouping would. If the input is not already sorted by the same key, semantically identical items that are not next to each other end up in separate groups.

`groupby()`는 *연속된* 일치 항목만 묶는다 — `dict` 기반 그룹화처럼 전체 이터러블을 훑어 모든 일치 항목을 찾지 않는다. 입력이 같은 키로 미리 정렬되어 있지 않으면, 의미상 같은 항목이라도 서로 붙어 있지 않으면 별개의 그룹으로 나뉜다.

✅ **Fix / 해결법:**  
Always `sorted(data, key=...)` with the SAME key before calling `groupby()`.  
`groupby()`를 호출하기 전에 항상 같은 key로 `sorted(data, key=...)`를 적용하자.
```python
import itertools as it
data = ["ant", "banana", "apple", "bat"]           # NOT pre-sorted by first letter
sorted_data = sorted(data, key=lambda w: w[0])       # ✅ sort first
for letter, group in it.groupby(sorted_data, key=lambda w: w[0]):
    print(letter, list(group))
```

### Mistake 2 — Assuming `re.match()` searches the whole string
`re.match()` only checks for a match at the very *beginning* of the string. `re.match(r"\d+", "abc123")` returns `None`, even though `"123"` clearly appears later in the string — `re.search()` is the one that looks anywhere.

`re.match()`는 문자열의 아주 *처음*에서만 매칭을 확인한다. `re.match(r"\d+", "abc123")`는 `"123"`이 뒤쪽에 분명히 있어도 `None`을 반환한다 — 어디든 찾는 쪽은 `re.search()`다.

✅ **Fix / 해결법:**  
Use `re.search()` when the pattern could appear anywhere in the string; save `re.match()` for when it truly must be at the start.  
패턴이 문자열 어디에든 나타날 수 있다면 `re.search()`를, 반드시 처음에 있어야 할 때만 `re.match()`를 사용하자.
```python
import re
print(re.match(r"\d+", "abc123"))       # None — does not start with a digit
print(re.search(r"\d+", "abc123"))        # ✅ finds "123" anywhere in the string
```

### Mistake 3 — Expecting `datetime.now()` to be a fixed snapshot
Every call to `datetime.now()` returns the current moment *at that exact call* — calling it twice in a row gives two slightly different timestamps, not the same one reused.

`datetime.now()`를 호출할 때마다 *그 호출 시점*의 현재 순간을 반환한다 — 연속으로 두 번 호출하면 재사용되는 같은 값이 아니라 미세하게 다른 두 타임스탬프를 얻는다.

✅ **Fix / 해결법:**  
Call `datetime.now()` once, store it in a variable, and reuse that variable for every calculation that should share the same "now."  
`datetime.now()`를 한 번만 호출해 변수에 저장하고, 같은 "지금"을 공유해야 하는 모든 계산에 그 변수를 재사용하자.
```python
from datetime import datetime
start = datetime.now()   # ✅ captured once
# ... later, reuse `start` instead of calling datetime.now() again
```

# 💡 Tips (팁)
Useful tips or shortcuts (유용한 팁이나 단축법)

- `collections.Counter` and `collections.defaultdict` cover the vast majority of "count things" and "group things by key" tasks — reach for them before writing a manual loop with an `if key in d` check.  
`collections.Counter`와 `collections.defaultdict`는 "개수 세기"와 "키로 그룹화"라는 작업 대부분을 처리한다 — `if key in d` 검사가 있는 수동 반복문을 쓰기 전에 먼저 떠올리자.
- `itertools.product(a, b)` replaces a nested `for` loop for generating every combination — useful for building test cases or a parameter grid.  
`itertools.product(a, b)`는 모든 조합을 생성하는 중첩 `for` 반복문을 대체한다 — 테스트 케이스나 파라미터 그리드를 만들 때 유용하다.
- `operator.itemgetter("age")` as a `sorted()` key is faster than an equivalent `lambda x: x["age"]`, since it skips the overhead of calling a Python function.  
`sorted()`의 key로 쓰는 `operator.itemgetter("age")`는 동등한 `lambda x: x["age"]`보다 빠르다, Python 함수 호출 오버헤드를 건너뛰기 때문이다.
- Prefer `pathlib.Path` over `os.path` string manipulation in new code — `Path("a") / "b" / "c.txt"` is more readable than `os.path.join("a", "b", "c.txt")` and works the same way.  
새 코드에서는 `os.path` 문자열 조작보다 `pathlib.Path`를 선호하자 — `Path("a") / "b" / "c.txt"`가 `os.path.join("a", "b", "c.txt")`보다 읽기 쉽고 동작은 같다.
- Compile a regex pattern once with `re.compile()` if it will be reused many times in a loop — it avoids re-parsing the same pattern on every call.  
정규식 패턴이 반복문에서 여러 번 재사용된다면 `re.compile()`로 한 번만 컴파일하자 — 매번 같은 패턴을 다시 파싱하지 않아도 된다.

# 💼 Business Example (비즈니스 예제)
*How would a Business Analyst use this? BA는 이것을 실제로 어떻게 사용할까요?*

**Scenario (시나리오):** Server logs came in as plain text lines. You need to extract timestamps and status codes with regex, count how often each status code appears, and save a clean summary as JSON — all standard library, no pandas.

**시나리오:** 서버 로그가 순수 텍스트 줄로 들어왔다. 정규식으로 타임스탬프와 상태 코드를 추출하고, 각 상태 코드가 얼마나 자주 나타나는지 세고, 깔끔한 요약을 JSON으로 저장해야 한다 — pandas 없이 표준 라이브러리만으로.

**To do (할 일 목록):**
- [x] Use `re.findall()` to pull the status code out of each log line.  
`re.findall()`로 각 로그 줄에서 상태 코드를 추출한다.
- [x] Use `collections.Counter` to count how often each status code appears.  
`collections.Counter`로 각 상태 코드가 얼마나 자주 나타나는지 센다.
- [x] Use `datetime.strptime()` to parse the first timestamp into a real `datetime` object.  
`datetime.strptime()`으로 첫 타임스탬프를 실제 `datetime` 객체로 파싱한다.
- [x] Use `json.dumps()` to format the final summary.  
`json.dumps()`로 최종 요약을 포맷한다.

````python
import re, json
from collections import Counter
from datetime import datetime

log_lines = [
    "2024-01-15 10:02:11 status=200 path=/home",
    "2024-01-15 10:02:14 status=404 path=/missing",
    "2024-01-15 10:02:20 status=200 path=/home",
    "2024-01-15 10:03:01 status=500 path=/api/orders",
    "2024-01-15 10:03:05 status=200 path=/home",
]

# Extract every status code with a regex
# 정규식으로 모든 상태 코드 추출
status_codes = [re.findall(r"status=(\d+)", line)[0] for line in log_lines]
print("Status codes (상태 코드):", status_codes)

# Count occurrences of each status code
# 각 상태 코드의 발생 횟수 카운트
counts = Counter(status_codes)
print("Counts (카운트):", counts)

# Parse the first log line's timestamp into a real datetime object
# 첫 로그 줄의 타임스탬프를 실제 datetime 객체로 파싱
first_timestamp_str = log_lines[0].split(" status=")[0]
first_dt = datetime.strptime(first_timestamp_str, "%Y-%m-%d %H:%M:%S")
print("First event (첫 이벤트):", first_dt.strftime("%B %d, %Y at %H:%M:%S"))

# Build a clean summary and save it as JSON text
# 깔끔한 요약을 만들어 JSON 텍스트로 저장
summary = {
    "total_requests": len(log_lines),
    "status_breakdown": dict(counts),
    "error_rate": round((counts["404"] + counts["500"]) / len(log_lines), 2),
}
print(json.dumps(summary, indent=2))
````

# ❓ Review Questions (복습 문제)

**Q1.** What does `Counter("mississippi").most_common(2)` return, and what does it do differently from a manual counting loop?  
`Counter("mississippi").most_common(2)`는 무엇을 반환하나요? 수동 카운팅 반복문과 무엇이 다른가요?

`Counter("mississippi").most_common(2)` returns `[('i', 4), ('s', 4)]`, the two most common items and their counts. It counts items automatically instead of writing a manual counting loop with a dictionary.  
`Counter("mississippi").most_common(2)`는 가장 많이 등장한 두 문자와 개수를 반환합니다. 딕셔너리로 직접 개수를 세는 반복문을 작성할 필요 없이 자동으로 카운트해 줍니다.

**Q2.** Why must data be sorted before calling `itertools.groupby()` on it?  
`itertools.groupby()`를 호출하기 전에 왜 데이터를 정렬해야 하나요?

`itertools.groupby()` only groups consecutive items, so the data must be sorted first if equal items are not already next to each other.  
`itertools.groupby()`는 연속된 같은 값만 하나의 그룹으로 묶기 때문에, 같은 값들이 붙어 있지 않다면 먼저 정렬해야 합니다.

**Q3.** What is the difference between `re.match()` and `re.search()`?  
`re.match()`와 `re.search()`는 어떻게 다른가요?

`re.match()` only checks for a match at the beginning of the string, while `re.search()` looks for the first match anywhere in the string.  
`re.match()`는 문자열의 시작에서만 패턴을 찾고, `re.search()`는 문자열 전체를 검색하여 처음 일치하는 패턴을 찾습니다.

**Q4.** Why is `collections.deque` preferred over a `list` when repeatedly adding and removing items from the front of a collection?  
컬렉션 앞쪽에서 항목을 반복적으로 추가·제거할 때 왜 `list`보다 `collections.deque`가 선호되나요?

`collections.deque` supports O(1) insertion and removal at both ends, while a `list` requires O(n) time for operations at the front.  
`collections.deque`는 앞과 뒤 모두에서 O(1)로 추가·삭제할 수 있지만, `list`는 앞쪽 작업 시 요소를 모두 이동해야 하므로 O(n)이 걸립니다.

**Q5.** What does `random.seed(42)` do, and why is it useful for tests or demos?  
`random.seed(42)`는 무엇을 하나요? 테스트나 데모에서 왜 유용한가요?

`random.seed(42)` initializes the random number generator so it produces the same sequence of random values every time. This makes tests and demos reproducible.  
`random.seed(42)`는 난수 생성기를 항상 같은 상태로 초기화하여 매번 동일한 난수 순서를 생성하게 합니다. 그래서 테스트나 데모에서 결과를 재현할 수 있습니다.
