# 🧠 Concept (개념)

## What is it? (이게 뭔가요?)

This final section is a collection of roughly twenty small, idiomatic patterns that do not fit neatly into any single earlier topic, but come up constantly in real scripts: flattening a nested list, removing duplicates while keeping order, safely dividing without a `ZeroDivisionError`, finding the index of the maximum value, and dispatching behavior through a dictionary instead of a long `if/elif` chain.

이 마지막 섹션은 앞선 어느 한 주제에도 깔끔하게 들어맞지 않지만 실제 스크립트에서 계속 등장하는, 대략 20가지의 작고 관용적인 패턴 모음이다: 중첩 리스트 평탄화, 순서를 유지하며 중복 제거, `ZeroDivisionError` 없이 안전하게 나누기, 최댓값의 인덱스 찾기, 긴 `if/elif` 체인 대신 딕셔너리로 동작 디스패치하기.

| Pattern 패턴 | One-liner 한 줄 코드 |

## Why do we use it? (왜 사용하나요?)
*(When is it useful? 언제 유용한가요?)*

Recognizing these patterns by shape saves real time: seeing `max(enumerate(lst), key=lambda x: x[1])` and immediately knowing "that's index-of-max" is faster than working it out fresh each time, the same way an experienced cook recognizes "mise en place" without re-deriving why prepping ingredients first is a good idea.

이 패턴들을 형태로 알아보는 것은 실제로 시간을 절약해준다: `max(enumerate(lst), key=lambda x: x[1])`를 보고 "아, 최댓값 인덱스 찾기구나"를 즉시 아는 것은 매번 새로 풀어내는 것보다 빠르다. 숙련된 요리사가 재료를 먼저 준비하는 이유를 매번 다시 생각하지 않고 "미장플라스"를 알아보는 것과 같다.

Several of these patterns also quietly avoid real bugs — safe division prevents a crash on a zero denominator, and the dedup-with-order pattern avoids the common mistake of reaching for `set(lst)`, which silently scrambles the original order.

이 패턴들 중 몇 개는 조용히 실제 버그도 피하게 해준다 — 안전한 나눗셈은 분모가 0일 때 크래시를 막고, 순서 유지 중복 제거 패턴은 원본 순서를 조용히 뒤섞어버리는 `set(lst)`에 손이 가는 흔한 실수를 피하게 해준다.

*(Real-world use case 실제 사용 사례)*

Nearly every one of these patterns shows up directly in a data-cleaning script before pandas ever gets involved: deduplicating a list of customer IDs while preserving the order they were first seen, safely computing a ratio that might divide by zero, or sorting a dict of category totals to find the top performers.

이 패턴들 거의 전부가 pandas가 관여하기 전 데이터 정리 스크립트에 직접 등장한다: 처음 나타난 순서를 유지하며 고객 ID 리스트의 중복을 제거하거나, 0으로 나눌 수도 있는 비율을 안전하게 계산하거나, 카테고리별 합계 딕셔너리를 정렬해 상위 실적을 찾는 일 등이다.

The dictionary-dispatch pattern in particular is how a lot of real "business rule engine" code gets written — mapping a status code, a report type, or a discount tier to the right handling function without a sprawling `if/elif` chain that grows every time a new case is added.

특히 딕셔너리 디스패치 패턴은 실제 "비즈니스 규칙 엔진" 코드가 많이 작성되는 방식이다 — 상태 코드, 리포트 타입, 할인 등급을 새 케이스가 추가될 때마다 계속 커지는 방대한 `if/elif` 체인 없이 올바른 처리 함수에 매핑하는 것이다.

# 📝 Syntax (문법)

## Basic Syntax (기본 문법)

````python
lst = [3, 1, 4, 1, 5, 9, 2, 6]

# A quick tour of five patterns you'll use constantly
# 항상 쓰게 될 다섯 가지 패턴 빠르게 훑어보기
matrix = [[1, 2], [3, 4]]
print([x for row in matrix for x in row])            # flatten

print(list(dict.fromkeys(lst)))                         # dedup, keep order

a, b = 10, 0
print(a / b if b != 0 else 0)                             # safe division

idx, val = max(enumerate(lst), key=lambda x: x[1])
print(idx, val)                                              # index of the max value

d = {"a": 3, "b": 1, "c": 2}
print(dict(sorted(d.items(), key=lambda x: x[1])))              # sort a dict by value
````

## Common Variations (자주 쓰는 변형)

````python
from itertools import chain

# Three ways to flatten a 2D list — comprehension, sum(), and itertools.chain
# 2D 리스트를 평탄화하는 세 가지 방법 — 컴프리헨션, sum(), itertools.chain
matrix = [[1, 2], [3, 4], [5]]
print([x for row in matrix for x in row])
print(sum(matrix, []))
print(list(chain.from_iterable(matrix)))

# Conditional counting — two equivalent styles
# 조건부 카운팅 — 동등한 두 가지 스타일
data = [1, -2, 3, -4, 5]
print(sum(1 for x in data if x > 0))
print(sum(x > 0 for x in data))                  # bool acts as 1/0 directly (bool이 바로 1/0로 동작)

# Nested dict safe access — chain .get() calls with defaults at every level
# 중첩 딕셔너리 안전 접근 — 각 단계마다 기본값을 가진 .get() 체이닝
config = {"a": {"b": "found it"}}
print(config.get("a", {}).get("b", "default"))
print(config.get("x", {}).get("y", "default"))       # missing at the first level, still safe (첫 단계부터 없어도 안전)
````

# 🧪 Small Examples (예제)

## Example 1 — Counting & Flattening (카운팅·평탄화)

````python
from collections import Counter

# Most common value(s) -- Counter.most_common()
# 가장 흔한 값 -- Counter.most_common()
votes = ["cat", "dog", "cat", "bird", "dog", "cat"]
c = Counter(votes)
print(c.most_common())
most_common_val = c.most_common(1)[0][0]
print(most_common_val)

top2 = [v for v, _ in c.most_common(2)]
print(top2)

# Flattening a 2D list into 1D
# 2D 리스트를 1D로 평탄화
matrix = [[1, 2, 3], [4, 5], [6]]
flat = [x for row in matrix for x in row]
print(flat)

# Deduplicate while preserving the original order
# 원본 순서를 유지하며 중복 제거
raw = [3, 1, 4, 1, 5, 9, 2, 6, 5, 3]
seen = set()
unique_manual = [x for x in raw if not (x in seen or seen.add(x))]
print(unique_manual)

# The cleaner dict.fromkeys() version -- same result, more idiomatic
# 더 깔끔한 dict.fromkeys() 버전 -- 같은 결과, 더 관용적
unique_clean = list(dict.fromkeys(raw))
print(unique_clean)
print(unique_manual == unique_clean)
````

## Example 2 — Dict & Safe-Access Patterns (딕셔너리·안전 접근 패턴)

````python
d = {"a": 1, "b": 2, "c": 3}

# Swap keys and values
# 키와 값 교환
reversed_d = {v: k for k, v in d.items()}
print(reversed_d)

# Safe division -- avoids ZeroDivisionError with a fallback
# 안전한 나눗셈 -- 대체값으로 ZeroDivisionError 방지
def safe_divide(a, b):
    return a / b if b != 0 else 0

print(safe_divide(10, 2), safe_divide(10, 0))

# Safe nested access -- chain .get() with a default at every level
# 안전한 중첩 접근 -- 각 단계마다 기본값을 가진 .get() 체이닝
settings = {"user": {"theme": "dark"}}
print(settings.get("user", {}).get("theme", "light"))
print(settings.get("admin", {}).get("theme", "light"))        # "admin" doesn't exist, still safe ("admin"이 없어도 안전)

# Sort a dict by its values, ascending or descending
# 값 기준으로 딕셔너리 정렬, 오름차순 또는 내림차순
scores = {"Bob": 70, "Ann": 95, "Kim": 60}
print(dict(sorted(scores.items(), key=lambda x: x[1])))
print(dict(sorted(scores.items(), key=lambda x: x[1], reverse=True)))

# List to index map -- reverse lookup: value -> position
# 리스트를 인덱스 맵으로 -- 역방향 조회: 값 -> 위치
items = ["apple", "banana", "cherry"]
index_map = {v: i for i, v in enumerate(items)}
print(index_map)
print(index_map["banana"])
````

## Example 3 — Aggregation & Window Patterns (집계·윈도우 패턴)

````python
from itertools import accumulate

lst = [3, 7, 1, 9, 4]

# Index of the max/min value, not just the value itself
# 최댓값/최솟값의 값이 아니라 인덱스
max_idx, max_val = max(enumerate(lst), key=lambda x: x[1])
min_idx, min_val = min(enumerate(lst), key=lambda x: x[1])
print(f"max {max_val} at index {max_idx}")
print(f"min {min_val} at index {min_idx}")

# Sliding window -- every consecutive group of k elements
# 슬라이딩 윈도우 -- 연속된 k개 요소의 모든 그룹
k = 3
windows = [lst[i:i + k] for i in range(len(lst) - k + 1)]
print(windows)

# Difference between two lists, preserving order (unlike a plain set difference)
# 순서를 유지하는 두 리스트의 차이 (단순 셋 차집합과 다름)
list1 = [1, 2, 3, 4, 5]
list2 = [2, 4]
s2 = set(list2)
diff = [x for x in list1 if x not in s2]
print(diff)

# Running totals -- cumulative sum and product
# 누적 합계 -- 누적 합과 누적 곱
print(list(accumulate([1, 2, 3, 4])))                        # cumulative sum
print(list(accumulate([1, 2, 3, 4], lambda x, y: x * y)))       # cumulative product

# The max across several separate lists at once
# 여러 개의 개별 리스트에 걸친 최댓값
lst1, lst2, lst3 = [3, 1], [9, 2], [4, 7]
global_max = max(max(sub) for sub in (lst1, lst2, lst3))
print(global_max)
````

## Example 4 — Conversion Patterns (변환 패턴)

````python
import string

# String to a list of numbers -- two common styles
# 문자열을 숫자 리스트로 -- 두 가지 흔한 스타일
raw_line = "1 2 3 4 5"
nums_from_split = list(map(int, raw_line.split()))
print(nums_from_split)

csv_line = "10,20,30"
nums_from_csv = [int(x) for x in csv_line.split(",")]
print(nums_from_csv)

# Matrix transpose -- swap rows and columns
# 행렬 전치 -- 행과 열을 바꿈
matrix = [[1, 2, 3], [4, 5, 6]]
transposed_zip = list(zip(*matrix))
print(transposed_zip)

transposed_manual = [[matrix[j][i] for j in range(len(matrix))] for i in range(len(matrix[0]))]
print(transposed_manual)

# Alphabet strings -- handy for generating labels (A, B, C...) or test data
# 알파벳 문자열 -- 라벨(A, B, C...)이나 테스트 데이터를 만들 때 유용
print(string.ascii_lowercase)
print(string.ascii_uppercase[:5])
print(string.digits)
print(list(string.ascii_lowercase[:5]))
````

# ⚠️ Common Mistakes (자주 하는 실수)

### Mistake 1 — Using `set(lst)` for deduplication when order matters
`set(lst)` removes duplicates, but sets have no guaranteed order — the result might print in a completely different order than the original list, which is a problem whenever "first occurrence order" matters (like a list of recently-viewed items).

`set(lst)`는 중복을 제거하지만 셋은 순서를 보장하지 않는다 — 결과가 원본 리스트와 완전히 다른 순서로 출력될 수 있으며, 이는 "처음 나타난 순서"가 중요할 때(최근 조회한 항목 리스트 같은) 문제가 된다.

✅ **Fix / 해결법:**  
Use `list(dict.fromkeys(lst))` to deduplicate while preserving the original order.  
원본 순서를 유지하며 중복을 제거하려면 `list(dict.fromkeys(lst))`를 사용하자.
```python
lst = [3, 1, 2, 1, 3]
print(set(lst))                        # order not guaranteed (순서 보장 안 됨)
print(list(dict.fromkeys(lst)))          # ✅ [3, 1, 2] -- original order preserved
```

### Mistake 2 — Forgetting the `if __name__ == "__main__":` guard
Without this guard, any top-level code in a `.py` file runs immediately the moment another script does `import that_file` — including test prints, sample function calls, or anything meant only for running the file directly.

이 가드가 없으면, 다른 스크립트가 `import that_file`을 하는 순간 `.py` 파일의 최상위 코드가 즉시 실행된다 — 테스트용 출력, 샘플 함수 호출, 파일을 직접 실행할 때만 의도했던 것들까지 전부 포함해서.

✅ **Fix / 해결법:**  
Wrap script-only code in `if __name__ == "__main__":` so it only runs when the file is executed directly, not when imported.  
스크립트 전용 코드를 `if __name__ == "__main__":`으로 감싸서, import될 때가 아니라 파일이 직접 실행될 때만 실행되게 하자.
```python
def main():
    print("doing real work")

if __name__ == "__main__":       # ✅ only runs when this file is executed directly
    main()
```

### Mistake 3 — Calling `max()`/`min()` directly on a dict
`max(d)` and `min(d)` iterate over a dict's KEYS by default, not its values — `max({"a": 100, "z": 1})` returns `"z"` (the alphabetically largest key), not `"a"` (the key with the largest value), which is rarely what anyone actually wants.

`max(d)`와 `min(d)`는 기본적으로 딕셔너리의 값이 아니라 **키**를 순회한다 — `max({"a": 100, "z": 1})`는 (값이 가장 큰 키인) `"a"`가 아니라 (알파벳상 가장 큰 키인) `"z"`를 반환하며, 이는 대부분의 경우 원하는 결과가 아니다.

✅ **Fix / 해결법:**  
Pass `key=d.get` to compare by value instead of by key.  
키가 아니라 값으로 비교하려면 `key=d.get`을 전달하자.
```python
d = {"a": 100, "z": 1}
print(max(d))                   # "z" -- largest KEY, probably not what you want
print(max(d, key=d.get))          # ✅ "a" -- the key with the largest VALUE
```

# 💡 Tips (팁)
Useful tips or shortcuts (유용한 팁이나 단축법)

- `list(dict.fromkeys(lst))` is worth memorizing as THE dedup-with-order pattern — it is short, fast, and avoids the common `set(lst)` ordering mistake entirely. `list(dict.fromkeys(lst))`는 순서를 유지하는 중복 제거 패턴으로 외워둘 가치가 있다 — 짧고 빠르며 흔한 `set(lst)` 순서 실수를 완전히 피한다.
- `float("inf")` and `float("-inf")` are the standard sentinel starting values for a manual min/max search — they are guaranteed to lose to any real number on the first comparison. `float("inf")`와 `float("-inf")`는 수동 최솟값/최댓값 탐색을 위한 표준 시작값이다 — 첫 비교에서 어떤 실제 숫자에도 반드시 진다.
- A dictionary dispatch table (`ops = {"+": add, "-": sub}`) scales far better than a growing `if/elif` chain — adding a new case means adding one dict entry, not editing a long conditional. 딕셔너리 디스패치 테이블(`ops = {"+": add, "-": sub}`)은 계속 커지는 `if/elif` 체인보다 훨씬 잘 확장된다 — 새 케이스를 추가하는 것은 긴 조건문을 수정하는 게 아니라 딕셔너리 항목 하나를 추가하는 것이다.
- `zip(*matrix)` for transposing reads elegantly once memorized, but the manual double-comprehension version is worth understanding too — it makes explicit what `zip(*matrix)` is doing implicitly. 전치를 위한 `zip(*matrix)`는 외워두면 우아하게 읽히지만, 수동 이중 컴프리헨션 버전도 이해해둘 가치가 있다 — `zip(*matrix)`가 암묵적으로 하는 일을 명시적으로 보여주기 때문이다.

# 💼 Business Example (비즈니스 예제)
*How would a Business Analyst use this? BA는 이것을 실제로 어떻게 사용할까요?*

**Scenario (시나리오):** End-of-quarter reporting: raw sales entries need deduplicating, ranking by revenue, and safely computing a growth rate that might divide by zero — a good final test of everything in this notebook.

**시나리오:** 분기 말 리포팅: 원본 매출 항목의 중복을 제거하고, 매출로 순위를 매기고, 0으로 나눌 수도 있는 성장률을 안전하게 계산해야 한다 — 이 노트북의 모든 것을 마지막으로 테스트하기 좋은 예제.

**To do (할 일 목록):**
- [x] Deduplicate raw sales entries while preserving their original order.  
원본 순서를 유지하며 매출 항목의 중복을 제거한다.
- [x] Find the top region by revenue using the index-of-max pattern.  
최댓값 인덱스 패턴으로 매출 상위 지역을 찾는다.
- [x] Safely compute each region's growth rate, handling a zero previous-quarter value.  
이전 분기 값이 0인 경우를 처리하며 각 지역의 성장률을 안전하게 계산한다.
- [x] Dispatch the correct formatting function per region type using a dict instead of if/elif.  
if/elif 대신 딕셔너리로 지역 타입별 올바른 포맷 함수를 디스패치한다.

````python
# Deduplicate raw region names while keeping the order they first appeared
# 원본 지역명의 중복을 처음 나타난 순서를 유지하며 제거
raw_regions = ["West", "East", "West", "North", "East", "South"]
regions = list(dict.fromkeys(raw_regions))
print("Regions, deduped in order (순서 유지 중복 제거):", regions)

this_quarter = {"West": 480, "East": 300, "North": 150, "South": 90}
last_quarter = {"West": 400, "East": 300, "North": 0, "South": 120}

# Find the top region by revenue using index-of-max on a list of (region, revenue) pairs
# (지역, 매출) 쌍 리스트에 index-of-max를 적용해 매출 상위 지역 찾기
pairs = list(this_quarter.items())
top_idx, (top_region, top_revenue) = max(enumerate(pairs), key=lambda x: x[1][1])
print(f"Top region (최고 지역): {top_region} (${top_revenue})")

# Safely compute growth rate, handling a zero last-quarter value
# 이전 분기 값이 0인 경우를 처리하며 안전하게 성장률 계산
def growth_rate(current, previous):
    return (current - previous) / previous if previous != 0 else None

print("Growth rates (성장률):")
for region in regions:
    rate = growth_rate(this_quarter[region], last_quarter[region])
    rate_str = f"{rate:.1%}" if rate is not None else "N/A (no prior data)"
    print(f"  {region}: {rate_str}")

# Dispatch a formatting function by region type instead of if/elif
# if/elif 대신 지역 타입별로 포맷 함수 디스패치
def format_domestic(region, revenue):
    return f"{region} (domestic): ${revenue:,}"

def format_international(region, revenue):
    return f"{region} (international): ${revenue:,} USD"

formatters = {"West": format_domestic, "East": format_domestic, "North": format_international, "South": format_international}

print("Formatted summary (포맷된 요약):")
for region in regions:
    print(" ", formatters[region](region, this_quarter[region]))
````
