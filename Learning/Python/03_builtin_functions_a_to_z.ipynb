# 🧠 Concept (개념)

## What is it? (이게 뭔가요?)
*(Explain it in your own words. 자신의 언어로 설명해보세요.)*

Built-in functions are the roughly 70 functions Python makes globally available without any `import` — things like `len()`, `sorted()`, `sum()`, and `zip()`. JS/TS spreads the same functionality across `Array.prototype` methods (`.map()`, `.filter()`, `.sort()`), global functions (`parseInt`), and object statics (`Object.keys()`) — Python instead puts most of it in one flat, always-available namespace, so `len(x)` works identically whether `x` is a string, list, dict, or set.

내장 함수는 Python이 `import` 없이 전역으로 제공하는 약 70개의 함수다 — `len()`, `sorted()`, `sum()`, `zip()` 같은 것들이다. JS/TS는 같은 기능을 `Array.prototype` 메서드(`.map()`, `.filter()`, `.sort()`), 전역 함수(`parseInt`), 객체 정적 메서드(`Object.keys()`)로 흩어놓는 반면, Python은 대부분을 하나의 평평하고 항상 사용 가능한 네임스페이스에 담아둔다 — 그래서 `len(x)`는 `x`가 문자열이든 리스트든 딕셔너리든 셋이든 동일하게 동작한다.

| | JS/TS | Python |
|---|---|---|
| Length 길이 | `arr.length` (property) | `len(arr)` (function) |
| Transform each item 각 항목 변환 | `arr.map(fn)` | `map(fn, arr)` or a list comprehension |
| Filter items 항목 필터링 | `arr.filter(fn)` | `filter(fn, arr)` or a list comprehension |
| Sort 정렬 | `arr.sort(fn)` (mutates!) | `sorted(arr, key=fn)` (returns new list) |
| Sum 합계 | `arr.reduce((a,b)=>a+b, 0)` | `sum(arr)` |
| Convert to string 문자열 변환 | `String(x)` or a template string | `str(x)` |
| Parse number 숫자 파싱 | `parseInt(s)` / `parseFloat(s)` | `int(s)` / `float(s)` |
| Index + value together 인덱스와 값 함께 | `arr.forEach((v,i)=>...)` | `enumerate(arr)` |
| Multiple arrays together 여러 배열을 함께 | manual index loop | `zip(arr1, arr2)` |

A subtle but important habit shift: `arr.sort()` mutates the original array in JS, while Python's `sorted()` always returns a brand-new list and leaves the original untouched (the list also has a `.sort()` method that mutates in place, mirroring JS — Python simply gives you both, and expects you to pick the right one on purpose).

미묘하지만 중요한 습관 차이가 있다: JS의 `arr.sort()`는 원본 배열을 변경하지만, Python의 `sorted()`는 항상 새로운 리스트를 반환하고 원본은 그대로 둔다 (리스트에는 제자리에서 변경하는 `.sort()` 메서드도 있어 JS와 비슷하게 동작한다 — Python은 둘 다 제공하니, 의도적으로 올바른 쪽을 골라 써야 한다).

## Why do we use it? (왜 사용하나요?)
*(When is it useful? 언제 유용한가요?)*

Built-in functions are implemented in C and heavily optimized — `sum(lst)` is faster than a hand-written `for` loop that accumulates a total, and `sorted(lst, key=...)` uses a well-tuned sort (Timsort) instead of whatever algorithm you would improvise. Reaching for them first is not just shorter code, it is usually faster code too.

내장 함수는 C로 구현되어 있고 고도로 최적화되어 있다 — `sum(lst)`는 합계를 누적하는 직접 짠 `for` 반복문보다 빠르고, `sorted(lst, key=...)`는 즉흥적으로 구현할 만한 알고리즘 대신 잘 다듬어진 정렬(Timsort)을 사용한다. 이 함수들을 먼저 떠올리는 것은 코드가 짧아질 뿐 아니라 대체로 더 빠르기도 하다.

They also read as a shared vocabulary: any Python developer instantly recognizes what `sorted(data, key=lambda x: x["age"])` does, whereas a hand-rolled sort function needs a comment to explain the same intent.

이 함수들은 공유된 어휘로도 기능한다: 어떤 Python 개발자든 `sorted(data, key=lambda x: x["age"])`를 보면 즉시 의도를 파악한다. 반면 직접 짠 정렬 함수는 같은 의도를 설명하기 위해 주석이 필요하다.

## When is it used in Business Analytics? (Business Analytics에서는 언제 쓰이나요?)
*(Real-world use case 실제 사용 사례)*

`sorted(records, key=lambda r: r["revenue"], reverse=True)[:10]` is a "top 10 accounts by revenue" report in a single line — a task that would otherwise mean writing and testing a manual sort. `zip()` and `enumerate()` show up constantly when combining parallel lists pulled from a spreadsheet, and `max`/`min` with a `key=` are the fastest way to answer "which row has the highest X" without touching pandas at all.

`sorted(records, key=lambda r: r["revenue"], reverse=True)[:10]`은 한 줄로 "매출 상위 10개 계정" 리포트가 된다 — 그렇지 않았다면 수동 정렬을 작성하고 테스트해야 했을 작업이다. 스프레드시트에서 가져온 병렬 리스트를 결합할 때 `zip()`과 `enumerate()`가 계속 등장하며, `key=`가 붙은 `max`/`min`은 pandas를 전혀 쓰지 않고도 "어떤 행이 X가 가장 큰가"에 가장 빠르게 답하는 방법이다.

`map`/`filter` and `sum` with a generator expression are also exactly what a pandas `.apply()` or boolean mask is doing under the hood — understanding them here makes pandas feel like an extension of the same ideas, not a new mental model.

`map`/`filter`와 제너레이터 표현식을 쓰는 `sum`은 pandas의 `.apply()`나 불리언 마스크가 내부적으로 하는 일과 정확히 같다 — 여기서 이해해두면 pandas가 새로운 사고방식이 아니라 같은 아이디어의 연장으로 느껴진다.

# 📝 Syntax (문법)

## Basic Syntax (기본 문법)

````python
# The handful of built-ins used in almost every script
# 거의 모든 스크립트에서 쓰이는 대표 내장 함수들
numbers = [4, -2, 7, 1, -9]

print(len(numbers))                # length
print(sum(numbers))                # total
print(max(numbers), min(numbers))  # largest, smallest
print(sorted(numbers))             # a new sorted list
print(abs(numbers[1]))             # absolute value
print(type(numbers))                # the type itself
print(isinstance(numbers, list))    # type check

# Transform and filter without writing a loop
# 반복문 없이 변환하고 필터링하기
doubled = list(map(lambda x: x * 2, numbers))
positives = list(filter(lambda x: x > 0, numbers))
print(doubled)
print(positives)
````

## Common Variations (자주 쓰는 변형)

````python
# Conversion functions — the "cast" family
# 변환 함수 — "형변환" 계열
print(int("42"), float("3.14"), str(42), bool(0))

# enumerate + zip — the two most common loop helpers
# enumerate + zip — 가장 흔한 반복 도우미 두 가지
fruits = ["apple", "banana", "cherry"]
prices = [1.2, 0.5, 3.0]
for i, fruit in enumerate(fruits, start=1):
    print(i, fruit)
for fruit, price in zip(fruits, prices):
    print(fruit, price)

# sorted() with key= — sort by any field, ascending or descending
# key=를 사용한 sorted() — 원하는 필드 기준으로 오름/내림차순 정렬
people = [{"name": "Bob", "age": 30}, {"name": "Alice", "age": 25}]
print(sorted(people, key=lambda p: p["age"]))
print(sorted(people, key=lambda p: p["age"], reverse=True))

# sum() with a generator expression — no intermediate list needed
# 제너레이터 표현식을 쓰는 sum() — 중간 리스트가 필요 없음
print(sum(p["age"] for p in people))
````

## Example 1 — abs to dict: Math, Type & Collection Creation (수학·타입·컬렉션 생성)

````python
# abs() / all() / any() — quick numeric & boolean checks
# abs()/all()/any() — 숫자·불리언 빠른 확인
print(abs(-5), abs(-3.2))
print(all([True, True, True]))            # True — every item is truthy
print(all([1, 2, 0]))                      # False — 0 is falsy (0은 falsy)
print(any([False, False, True]))            # True — at least one truthy
print(any(x > 5 for x in [1, 2, 3]))         # False

# bin() / bool() — base conversion & truthiness
# bin()/bool() — 진법 변환과 truthy 여부
print(bin(10), bin(255))
print(bool(0), bool(""), bool([]), bool(None))   # all False (모두 False)
print(bool(1), bool("a"), bool([1]))               # all True

# callable() / chr() — is it a function? / int to unicode char
# callable()/chr() — 호출 가능 여부 / 정수 → 유니코드 문자
print(callable(print), callable(42))
print(chr(65), chr(97), chr(44032))

# dict() — three equivalent ways to build a dictionary
# dict() — 딕셔너리를 만드는 세 가지 동등한 방법
print(dict(a=1, b=2))
print(dict([("a", 1), ("b", 2)]))
print(dict(zip(["a", "b"], [1, 2])))

# dir() — explore what an object can do (preview only, the full list is long)
# dir() — 객체가 할 수 있는 일 탐색 (전체가 길어서 일부만 미리보기)
print(len(dir(str)), "methods on str, first 5:", dir(str)[:5])

# divmod() — quotient and remainder in one call
# divmod() — 몫과 나머지를 한 번에
print(divmod(10, 3))          # (3, 1)
q, r = divmod(17, 5)
print(q, r)
````

## Example 2 — enumerate to map: Iteration & Transformation (순회 & 변환)

````python
# enumerate() — index + value together, no manual counter
# enumerate() — 인덱스와 값을 함께, 수동 카운터 불필요
for i, val in enumerate(["a", "b", "c"]):
    print(i, val)

# filter() — keep only what passes a condition
# filter() — 조건을 통과하는 요소만 남기기
print(list(filter(lambda x: x > 2, [1, 2, 3, 4])))
print(list(filter(None, [0, 1, "", "a", None])))   # None as fn drops every falsy value (None을 넘기면 falsy 값 전부 제거)

# float() / int() — numeric conversion
# float()/int() — 숫자 변환
print(float("3.14"), float("inf"))
print(int("42"), int(3.9), int("ff", 16))     # int() truncates, does not round (int()는 버림이지 반올림이 아님)

# format() — an alternative to f-string format specs
# format() — f-string 포맷 스펙의 대안
print(format(3.14159, ".2f"))
print(format(1000000, ","))
print(format(255, "x"), format(255, "b"))

# frozenset() — an immutable set, usable as a dict key (a regular set cannot be)
# frozenset() — 불변 셋, 딕셔너리 키로 사용 가능 (일반 set은 불가능)
fs = frozenset([1, 2, 3])
lookup = {fs: "a frozen trio"}
print(lookup[frozenset([1, 2, 3])])

# getattr() / hasattr() — dynamic, safe attribute access
# getattr()/hasattr() — 동적이고 안전한 속성 접근
class Config:
    pass
cfg = Config()
cfg.env = "production"
print(getattr(cfg, "env", "default"))
print(getattr(cfg, "missing_field", "default"))   # falls back instead of raising AttributeError (에러 대신 대체값 반환)
print(hasattr(cfg, "env"))

# hash() / id() — every object has both; only immutable types are hashable
# hash()/id() — 모든 객체가 둘 다 가짐; 해시는 불변 타입만 가능
print(isinstance(hash("hello"), int))
print(isinstance(id(cfg), int))

# hex() — hexadecimal string
# hex() — 16진수 문자열
print(hex(255), hex(16))

# input() — always returns str (shown as a comment; it would pause execution here)
# input() — 항상 str 반환 (실행이 멈추므로 여기서는 주석으로만 설명)
# name = input("Name: ")        # typing "booki" -> name == "booki" (a string, even if it looks like a number)
# age = int(input("Age: "))     # convert manually to use it as a number

# isinstance() — the preferred way to check type (over type(x) == y)
# isinstance() — 타입 확인의 권장 방법 (type(x) == y 대신)
print(isinstance(42, (int, float)))

# len() / list() / map()
# len()/list()/map()
print(len("hello"), len([1, 2, 3]), len({"a": 1}))
print(list("abc"), list(range(3)))
print(list(map(str, [1, 2, 3])))
````

## Example 3 — max to repr: Sorting, Comparison & Representation (정렬·비교·표현)

````python
# max() / min() — with and without key=
# max()/min() — key= 유무에 따라
print(max(3, 1, 4, 1, 5))
print(max([3, 1, 4]))
print(max("apple", "banana"))              # alphabetical comparison (사전순 비교)
print(max([-3, -1, -2], key=abs))            # -3 — ranked by absolute value (절댓값 기준)

people = [{"name": "Bob", "age": 30}, {"name": "Alice", "age": 25}]
print(min(people, key=lambda p: p["age"]))   # the whole dict with the smallest age (나이가 가장 작은 딕셔너리 전체)

# next() — pull values one at a time from an iterator
# next() — 이터레이터에서 값을 하나씩 꺼내기
it = iter([1, 2, 3])
print(next(it))              # 1
print(next(it))              # 2
print(next(it, "done"))      # 3
print(next(it, "done"))      # "done" — exhausted, so the default is returned instead of StopIteration (소진되어 기본값 반환)

# oct() / ord() — base conversion and character codes
# oct()/ord() — 진법 변환과 문자 코드
print(oct(8), oct(255))
print(ord("A"), ord("a"), ord("가"))

# pow() — exponent, with an optional fast modulus form
# pow() — 거듭제곱, 빠른 모듈러 연산 옵션 포함
print(pow(2, 10))                # same as 2 ** 10
print(pow(2, 10, 1000))           # (2**10) % 1000, computed efficiently (효율적으로 계산됨)

# print() — sep / end / unpacking
# print() — sep / end / 언패킹
print("hello", "world")
print("hello", "world", sep="-")
print("hello", end=" ")
print("world")
print(*[1, 2, 3])                  # unpack a list as separate arguments (리스트를 개별 인자로 언패킹)

# range() / repr()
# range()/repr()
print(list(range(0, 10, 2)))
print(repr("hello"))                # quotes included — the developer-facing form (따옴표 포함, 개발자용 표현)
print(repr([1, 2]))
````

## Example 4 — reversed to str: Reversal, Rounding & Conversion (역순·반올림·변환)

````python
# reversed() — a reverse iterator, original stays untouched
# reversed() — 역순 이터레이터, 원본은 그대로
print(list(reversed([1, 2, 3])))

# round() — watch out for banker's rounding on .5
# round() — .5에서 은행가 반올림 주의
print(round(3.14159, 2))       # 3.14
print(round(1234, -2))          # 1200 — negative ndigits rounds left of the decimal (음수 자릿수는 소수점 왼쪽으로 반올림)
print(round(2.5), round(3.5))    # 2 4 — rounds to the nearest EVEN number, not always up (짝수 쪽으로 반올림, 항상 올림은 아님)

# set() — build a set from any iterable
# set() — 어떤 이터러블로도 셋을 만들 수 있음
print(set([1, 2, 2, 3]))
print(set("hello"))

# setattr() — set an attribute dynamically, by name (a string)
# setattr() — 이름(문자열)으로 속성을 동적으로 설정
class Config:
    pass
cfg = Config()
setattr(cfg, "env", "staging")
print(cfg.env)

# slice() — a reusable slicing object
# slice() — 재사용 가능한 슬라이싱 객체
window = slice(1, 5, 2)
print([0, 1, 2, 3, 4, 5][window])

# sorted() — always returns a NEW list; supports key= and reverse=
# sorted() — 항상 새 리스트를 반환; key=와 reverse= 지원
data = [{"name": "Bob", "age": 30}, {"name": "Alice", "age": 25}]
print(sorted(data, key=lambda x: x["age"]))
print(sorted(data, key=lambda x: (-x["age"], x["name"])))   # multi-level sort: age desc, then name asc (다중 정렬: 나이 내림차순 후 이름 오름차순)

# str() — convert anything to its readable string form
# str() — 무엇이든 읽기 좋은 문자열로 변환
print(str(42), str(3.14), str(True), str(None), str([1, 2]))
````

# ⚠️ Common Mistakes (자주 하는 실수)

### Mistake 1 — Assuming `list.sort()` returns the sorted list
`lst.sort()` sorts in place and returns `None`. Writing `new_list = lst.sort()` silently sets `new_list` to `None`, not the sorted data. `sorted(lst)` is the one that returns a new list — the two are not interchangeable.

`lst.sort()`는 제자리에서 정렬하고 `None`을 반환한다. `new_list = lst.sort()`라고 쓰면 `new_list`는 조용히 `None`이 되어버린다. 새 리스트를 반환하는 쪽은 `sorted(lst)`다 — 이 둘은 서로 바꿔 쓸 수 없다.

✅ **Fix / 해결법:**  
Use `sorted(lst)` when you need the result as a value; use `lst.sort()` only when you want to mutate in place and do not need the return value.  
결과를 값으로 써야 하면 `sorted(lst)`를, 제자리에서 바꾸기만 하고 반환값이 필요 없으면 `lst.sort()`를 사용하자.
```python
lst = [3, 1, 2]
new_list = sorted(lst)   # ✅ [1, 2, 3], lst is untouched
lst.sort()                 # ✅ lst is now [1, 2, 3] in place, returns None
```

### Mistake 2 — Forgetting that `filter()` and `map()` return iterators, not lists
In Python 3, `filter()` and `map()` are lazy — they return an iterator object, not a list. `print(filter(lambda x: x > 2, [1,2,3]))` prints something like `<filter object at 0x...>`, not the filtered values.

Python 3에서 `filter()`와 `map()`은 게으르게(lazy) 동작한다 — 리스트가 아니라 이터레이터 객체를 반환한다. `print(filter(lambda x: x > 2, [1,2,3]))`는 필터링된 값이 아니라 `<filter object at 0x...>` 같은 것을 출력한다.

✅ **Fix / 해결법:**  
Wrap the result in `list()` when you need to see or reuse the values more than once.  
값을 보거나 여러 번 재사용해야 한다면 `list()`로 감싸자.
```python
result = list(filter(lambda x: x > 2, [1, 2, 3, 4]))   # ✅ [3, 4]
```

### Mistake 3 — `round(2.5)` does not always round up
Python's `round()` uses "banker's rounding" (round-half-to-even): `.5` rounds to whichever neighbor is even, not always upward. `round(2.5)` gives `2`, while `round(3.5)` gives `4`.

Python의 `round()`는 "은행가 반올림"(round-half-to-even)을 사용한다: `.5`는 항상 위로 올리는 게 아니라 짝수인 쪽으로 반올림된다. `round(2.5)`는 `2`를, `round(3.5)`는 `4`를 준다.

✅ **Fix / 해결법:**  
If you specifically need "always round half up" (like typical rounding taught in school), write that logic explicitly.  
"항상 반올림"(학교에서 배운 방식)이 필요하다면 그 로직을 명시적으로 작성하자.
```python
import math
print(round(2.5))                  # 2 — banker's rounding
print(math.floor(2.5 + 0.5))        # 3 — always-round-half-up (for positive numbers)
```

# 💡 Tips (팁)
Useful tips or shortcuts (유용한 팁이나 단축법)

- `sorted(data, key=lambda x: (-x["age"], x["name"]))` sorts by multiple fields at once — negate a numeric field to reverse just that one while keeping others ascending.  
`sorted(data, key=lambda x: (-x["age"], x["name"]))`처럼 튜플을 key로 쓰면 여러 필드로 동시에 정렬할 수 있다 — 숫자 필드 앞에 마이너스를 붙이면 그 필드만 내림차순으로 뒤집을 수 있다.
- `zip(*matrix)` is the one-liner for transposing a 2D list (swapping rows and columns) — no nested loop needed.  
`zip(*matrix)`는 2D 리스트를 전치(행과 열을 바꿈)하는 한 줄 코드다 — 중첩 반복문이 필요 없다.
- `next(iterator, default)` avoids a `StopIteration` crash when an iterator might be empty — always safer than a bare `next(iterator)` in scripts.  
`next(iterator, default)`는 이터레이터가 비어있을 수 있을 때 `StopIteration` 크래시를 피한다 — 스크립트에서는 맨 `next(iterator)`보다 항상 더 안전하다.
- `isinstance(x, (int, float))` is preferred over `type(x) == int` because it also works correctly with subclasses.  
`isinstance(x, (int, float))`는 서브클래스에서도 올바르게 동작하므로 `type(x) == int`보다 권장된다.
- When debugging, `print(f"{variable=}")` (Python 3.8+) prints both the variable's name and its value in one shot — faster than typing `print("variable:", variable)`.  
디버깅할 때 `print(f"{variable=}")`(Python 3.8+)는 변수 이름과 값을 한 번에 출력한다 — `print("variable:", variable)`을 직접 입력하는 것보다 빠르다.

# 💼 Business Example (비즈니스 예제)
*How would a Business Analyst use this? BA는 이것을 실제로 어떻게 사용할까요?*

**Scenario (시나리오):** Marketing handed you a raw list of campaign results (dicts). Before pandas, you need a quick top-N report and some cleanup — pure built-in functions, no libraries.

**시나리오:** 마케팅팀이 캠페인 결과가 담긴 원본 리스트(딕셔너리)를 넘겼다. pandas를 쓰기 전에 라이브러리 없이 순수 내장 함수만으로 빠른 상위 순위 리포트와 정리 작업을 해야 한다.

**To do (할 일 목록):**
- [x] Use `sorted()` with `key=` to rank campaigns by ROI, highest first.  
`key=`를 사용한 `sorted()`로 ROI 기준 캠페인 순위를 매긴다 (높은 순).
- [x] Use `sum()` with a generator expression to compute total spend.  
제너레이터 표현식을 쓰는 `sum()`으로 총 지출을 계산한다.
- [x] Use `max()` with `key=` to find the single best-performing campaign.  
`key=`를 사용한 `max()`로 성과가 가장 좋은 캠페인 하나를 찾는다.
- [x] Use `enumerate()` to print a numbered ranking.  
`enumerate()`로 번호가 매겨진 순위를 출력한다.

````python
campaigns = [
    {"name": "Spring Sale", "spend": 2000, "revenue": 8000},
    {"name": "Email Blast", "spend": 300,  "revenue": 2400},
    {"name": "Influencer",  "spend": 1500, "revenue": 3000},
    {"name": "Retargeting", "spend": 600,  "revenue": 3600},
]

# Compute ROI for each campaign — a plain function, no libraries needed
# 캠페인별 ROI 계산 — 라이브러리 없는 순수 함수
def roi(c):
    return (c["revenue"] - c["spend"]) / c["spend"]

ranked = sorted(campaigns, key=roi, reverse=True)

print("Campaign ranking by ROI (ROI 기준 캠페인 순위):")
for rank, c in enumerate(ranked, start=1):
    print(f"{rank}. {c['name']:<12} ROI={roi(c):.1%}")

total_spend = sum(c["spend"] for c in campaigns)
best = max(campaigns, key=roi)

print("-" * 40)
print(f"Total spend (총 지출): ${total_spend:,}")
print(f"Best campaign (최고 캠페인): {best['name']} (ROI {roi(best):.1%})")
````

# 📝 Summary (요약)
*Write today's concept in 3~5 sentences. 오늘 배운 개념을 3~5문장으로 정리해보세요.*

Python's built-in functions form a single, always-available toolbox — `len()`, `sorted()`, `sum()`, `map()`, `filter()`, `zip()`, and dozens more — that work uniformly across strings, lists, tuples, dicts, and sets, unlike JS/TS's split between array methods, global functions, and object statics. The most important habit to build is reaching for these before writing a manual loop: `sorted(data, key=...)` replaces a hand-rolled sort, `sum(x for x in data if ...)` replaces an accumulator loop, and `zip()`/`enumerate()` replace index-based iteration entirely. A few behaviors are worth memorizing early: `sorted()` returns a new list while `list.sort()` mutates and returns `None`, `filter()`/`map()` return lazy iterators that need `list()` to inspect, and `round()` uses banker's rounding rather than always rounding halves up. These same functions reappear, almost unchanged in spirit, as pandas methods later in this curriculum.

Python의 내장 함수는 하나의 항상 사용 가능한 도구 상자를 이룬다 — `len()`, `sorted()`, `sum()`, `map()`, `filter()`, `zip()`을 비롯한 수십 개가 문자열, 리스트, 튜플, 딕셔너리, 셋에 걸쳐 균일하게 동작한다. 이는 배열 메서드, 전역 함수, 객체 정적 메서드로 나뉘어 있는 JS/TS와 다르다. 가장 중요하게 길러야 할 습관은 수동 반복문을 짜기 전에 이 함수들을 먼저 떠올리는 것이다: `sorted(data, key=...)`는 직접 짠 정렬을, `sum(x for x in data if ...)`는 누적 반복문을, `zip()`/`enumerate()`는 인덱스 기반 순회를 완전히 대체한다. 일찍 외워둘 만한 동작들도 있다: `sorted()`는 새 리스트를 반환하지만 `list.sort()`는 제자리에서 변경하고 `None`을 반환하며, `filter()`/`map()`은 확인하려면 `list()`가 필요한 게으른 이터레이터를 반환하고, `round()`는 항상 반올림하는 게 아니라 은행가 반올림을 사용한다. 이 함수들은 이 커리큘럼의 뒷부분에서 거의 같은 정신으로 pandas 메서드로 다시 등장한다.
