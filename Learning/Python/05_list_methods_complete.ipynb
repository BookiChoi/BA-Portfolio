# 🧠 Concept (개념)

## What is it? (이게 뭔가요?)
*(Explain it in your own words. 자신의 언어로 설명해보세요.)*

`list` methods are the dot-notation operations attached specifically to Python's mutable, ordered sequence type — adding (`.append()`, `.insert()`, `.extend()`), removing (`.remove()`, `.pop()`, `.clear()`), searching (`.index()`, `.count()`), sorting (`.sort()`), and reordering (`.reverse()`). JS/TS's `Array.prototype` covers the same ground with a nearly one-to-one mapping, so this section is mostly about learning new names for familiar operations — plus a handful of return-value differences that trip people up.

`list` 메서드는 Python의 가변적이고 순서 있는 시퀀스 타입에 전용으로 붙은 점(dot) 표기 연산이다 — 추가(`.append()`, `.insert()`, `.extend()`), 제거(`.remove()`, `.pop()`, `.clear()`), 검색(`.index()`, `.count()`), 정렬(`.sort()`), 순서 재배열(`.reverse()`)이다. JS/TS의 `Array.prototype`도 거의 1:1로 같은 영역을 다루기 때문에, 이 섹션은 대부분 익숙한 연산에 새 이름을 배우는 일이다 — 사람들을 곧잘 헷갈리게 하는 몇 가지 반환값 차이만 빼면.

| | JS/TS | Python |
|---|---|---|
| Add to the end 끝에 추가 | `arr.push(x)` (returns new length) | `lst.append(x)` (returns `None`) |
| Add multiple items 여러 항목 추가 | `arr.push(...items)` | `lst.extend(items)` |
| Remove from the end 끝에서 제거 | `arr.pop()` (returns removed item) | `lst.pop()` (returns removed item — same!) |
| Remove from the start 시작에서 제거 | `arr.shift()` | `lst.pop(0)` |
| Add to the start 시작에 추가 | `arr.unshift(x)` | `lst.insert(0, x)` |
| Remove by value 값으로 제거 | `arr.splice(arr.indexOf(x), 1)` | `lst.remove(x)` |
| Find an index 인덱스 찾기 | `arr.indexOf(x)` (returns -1) | `lst.index(x)` (raises `ValueError`!) |
| Sort in place 제자리 정렬 | `arr.sort(fn)` | `lst.sort(key=fn)` |
| Sort into a new array 새 배열로 정렬 | `[...arr].sort(fn)` | `sorted(lst, key=fn)` |
| Reverse in place 제자리 역순 | `arr.reverse()` | `lst.reverse()` (same!) |

The pattern worth internalizing: methods that end in a verb describing a mutation (`.sort()`, `.reverse()`) always return `None` in Python — they change the list and hand back nothing, so `x = lst.sort()` is a bug waiting to happen, not a way to grab the sorted result.

내면화할 가치가 있는 패턴: 변경을 뜻하는 동사로 끝나는 메서드(`.sort()`, `.reverse()`)는 Python에서 항상 `None`을 반환한다 — 리스트를 바꾸고 아무것도 돌려주지 않으므로, `x = lst.sort()`는 정렬 결과를 얻는 방법이 아니라 곧 터질 버그다.

## Why do we use it? (왜 사용하나요?)
*(When is it useful? 언제 유용한가요?)*

Choosing the right method has real performance consequences: `lst.pop(0)` or `lst.insert(0, x)` are both `O(n)` because every remaining element has to shift over, while `lst.append(x)` and `lst.pop()` are `O(1)`. Knowing this in advance means reaching for a `collections.deque` instead of a `list` when a script needs to repeatedly add and remove from the front.

올바른 메서드를 고르는 것은 실제 성능에 영향을 준다: `lst.pop(0)`이나 `lst.insert(0, x)`는 나머지 모든 요소가 밀려야 하므로 `O(n)`이지만, `lst.append(x)`와 `lst.pop()`은 `O(1)`이다. 이를 미리 알고 있으면 스크립트가 앞쪽에서 반복적으로 추가/제거해야 할 때 `list` 대신 `collections.deque`를 쓰는 판단을 할 수 있다.

The mutate-vs-return-new-list distinction (`.sort()` vs `sorted()`, `.reverse()` vs `[::-1]`) also determines whether other variables pointing at the same list get silently affected — a shared reference to a list you did not mean to change is a classic source of hard-to-trace bugs.

제자리 변경 vs 새 리스트 반환의 구분(`.sort()` vs `sorted()`, `.reverse()` vs `[::-1]`)은 같은 리스트를 가리키는 다른 변수들이 조용히 영향을 받는지 여부도 결정한다 — 바꿀 의도가 없었던 리스트를 공유 참조로 건드리는 것은 추적하기 어려운 버그의 전형적인 원인이다.

## When is it used in Business Analytics? (Business Analytics에서는 언제 쓰이나요?)
*(Real-world use case 실제 사용 사례)*

Before a list becomes a pandas `Series`, a BA is often building it up manually: appending rows that pass a filter, popping off a processed queue item, or sorting a small list of results for a quick report without needing the overhead of a full `DataFrame`.

리스트가 pandas `Series`가 되기 전, BA는 종종 이를 수동으로 만들어간다: 필터를 통과한 행을 추가하거나, 처리된 큐 항목을 pop하거나, `DataFrame`의 오버헤드 없이 빠른 리포트를 위해 작은 결과 리스트를 정렬하는 일 등이다.

`.copy()` and `copy.deepcopy()` matter constantly when passing data between functions — accidentally mutating a shared list that another part of a script still depends on is one of the most common silent-bug patterns in real analysis code.

`.copy()`와 `copy.deepcopy()`는 함수 사이에 데이터를 전달할 때 계속 중요하다 — 스크립트의 다른 부분이 여전히 의존하고 있는 공유 리스트를 실수로 변경하는 것은 실제 분석 코드에서 가장 흔한 조용한 버그 패턴 중 하나다.

# 📝 Syntax (문법)

## Basic Syntax (기본 문법)

````python
lst = [3, 1, 4]

# The everyday four: add, remove, sort, and check membership
# 일상적으로 쓰는 네 가지: 추가, 제거, 정렬, 멤버십 확인
lst.append(5)
lst.remove(1)
lst.sort()
print(lst)
print(4 in lst)

# sorted() vs .sort() — the single most important habit in this section
# sorted() vs .sort() — 이 섹션에서 가장 중요한 습관
numbers = [3, 1, 4, 1, 5]
new_copy = sorted(numbers)     # new list, original untouched (새 리스트, 원본은 그대로)
numbers.sort()                   # original changed in place (원본이 제자리에서 변경됨)
print(new_copy, numbers)
````

## Common Variations (자주 쓰는 변형)

````python
lst = [1, 2, 3]

# extend() vs += vs append() — the three ways to grow a list
# extend() vs += vs append() — 리스트를 늘리는 세 가지 방법
lst.extend([4, 5])       # add multiple items (여러 항목 추가)
lst += [6]                 # same behavior as extend() (extend()와 동일)
lst.append(7)                 # add exactly one item (정확히 하나 추가)
print(lst)

# pop() from the end vs remove() by value
# 끝에서 pop() vs 값으로 remove()
lst.pop()             # removes 7, the last item (마지막 항목 7 제거)
lst.remove(2)            # removes the value 2, wherever it is (값 2를 위치에 상관없이 제거)
print(lst)

# Copying safely before mutating, when the original must be preserved
# 원본을 보존해야 할 때 변경 전에 안전하게 복사하기
original = [1, 2, 3]
working_copy = original.copy()
working_copy.append(4)
print(original, working_copy)

# sort(key=...) with a lambda for multi-field or custom-order sorting
# 다중 필드나 커스텀 순서 정렬을 위한 sort(key=...)와 람다
data = [{"name": "Bob", "age": 30}, {"name": "Ann", "age": 25}]
data.sort(key=lambda x: x["age"])
print(data)
````

# 🧪 Small Examples (예제)

## Example 1 — Adding Items (추가)

````python
lst = [1, 2, 3]

# append() — add ONE item to the end; note it returns None, not the list!
# append() — 끝에 항목 하나 추가; None을 반환하지 원본 리스트가 아님!
lst.append(4)
print(lst)

# insert(index, value) — insert at a specific position (O(n), shifts everything after it)
# insert(인덱스, 값) — 특정 위치에 삽입 (O(n), 이후 요소들이 밀림)
lst.insert(0, 99)
print(lst)

# extend() — add MULTIPLE items, unpacking the given iterable
# extend() — 여러 항목 추가, 주어진 이터러블을 풀어서 추가
lst.extend([7, 8])
print(lst)

# += on a list does the same thing as extend() — NOT the same as append()!
# 리스트에 += 는 extend()와 동일하게 동작 — append()와는 다름!
lst += [9, 10]
print(lst)

# a common mix-up: append(a_list) nests the list instead of extending
# 흔한 실수: append(리스트)는 확장이 아니라 리스트를 중첩시킴
demo = [1, 2]
demo.append([3, 4])
print("append with a list:", demo)      # [1, 2, [3, 4]] — nested!
demo2 = [1, 2]
demo2.extend([3, 4])
print("extend instead:     ", demo2)     # [1, 2, 3, 4] — flat, usually what you want
````

## Example 2 — Removing Items (제거)

````python
lst = [3, 1, 4, 1, 5, 9]

# remove(value) — removes the FIRST matching value (raises ValueError if missing)
# remove(값) — 첫 번째로 일치하는 값을 제거 (없으면 ValueError)
lst.remove(1)
print(lst)                       # only the first 1 is gone (첫 번째 1만 사라짐)

# pop() — removes and returns the LAST item, O(1)
# pop() — 마지막 항목을 제거하고 반환, O(1)
last = lst.pop()
print(last, lst)

# pop(index) — removes and returns an item at a specific position
# pop(인덱스) — 특정 위치의 항목을 제거하고 반환
first = lst.pop(0)
print(first, lst)                 # ⚠️ O(n) — everything after index 0 shifts left (collections.deque is faster for this)

# clear() — removes everything, leaving an empty list
# clear() — 전체를 제거해 빈 리스트로 만듦
lst.clear()
print(lst)

# remove() raises ValueError if the value is not present — validate first
# remove()는 값이 없으면 ValueError를 던짐 — 먼저 확인하자
data = [1, 2, 3]
target = 99
if target in data:
    data.remove(target)
else:
    print(f"{target} not found, nothing removed")
````

## Example 3 — Searching (검색)

````python
lst = [3, 1, 4, 1, 5, 9, 1]

# index(value) — the position of the FIRST match (raises ValueError if missing)
# index(값) — 첫 번째로 일치하는 위치 (없으면 ValueError)
print(lst.index(4))

# index(value, start) — search starting from a given position
# index(값, 시작위치) — 주어진 위치부터 검색
print(lst.index(1, 2))       # skip the first 1 at index 1, find the next one (인덱스 1의 첫 1을 건너뛰고 다음을 찾음)

# count(value) — how many times a value appears
# count(값) — 값이 몇 번 나타나는지
print(lst.count(1))

# combine index() with a safety check to avoid ValueError
# ValueError를 피하기 위해 index()를 안전 검사와 함께 사용
target = 100
if target in lst:
    print(lst.index(target))
else:
    print(f"{target} is not in the list")

# a common pattern: find the positions of every match, not just the first
# 흔한 패턴: 첫 번째뿐 아니라 모든 일치 위치 찾기
positions = [i for i, val in enumerate(lst) if val == 1]
print(positions)
````

## Example 4 — Sorting (정렬)

````python
lst = [3, 1, 4, 1, 5, 9]

# sort() — sorts IN PLACE, ascending, returns None
# sort() — 제자리에서 오름차순 정렬, None 반환
lst.sort()
print(lst)

# sort(reverse=True) — descending
# sort(reverse=True) — 내림차순
lst.sort(reverse=True)
print(lst)

# sort(key=...) — sort by a computed value instead of the value itself
# sort(key=...) — 값 자체가 아니라 계산된 값을 기준으로 정렬
words = ["banana", "kiwi", "fig", "apple"]
words.sort(key=len)
print(words)

signed = [-3, 1, -4, 1, -5]
signed.sort(key=abs)
print(signed)                 # sorted by absolute value, original signs kept (절댓값 기준 정렬, 부호는 유지)

# sorted() — the non-mutating twin of .sort(); returns a NEW list
# sorted() — .sort()의 비파괴적 쌍둥이; 새 리스트를 반환
original = [3, 1, 4, 1, 5]
new_list = sorted(original)
print("original untouched:", original)
print("new sorted list:   ", new_list)

# sorted() also accepts key= and reverse=, same as .sort()
# sorted()도 .sort()와 마찬가지로 key=와 reverse=를 받음
records = [{"name": "Bob", "score": 70}, {"name": "Ann", "score": 95}]
print(sorted(records, key=lambda r: r["score"], reverse=True))
````

# ⚠️ Common Mistakes (자주 하는 실수)

### Mistake 1 — Capturing the return value of `.append()`
`.append()` mutates the list in place and returns `None`. Writing `lst = lst.append(x)` silently destroys the list, replacing it with `None`.

`.append()`는 리스트를 제자리에서 변경하고 `None`을 반환한다. `lst = lst.append(x)`라고 쓰면 리스트가 조용히 파괴되어 `None`으로 바뀐다.

✅ **Fix / 해결법:**  
Call `.append()` on its own line; never assign its result.  
`.append()`는 단독으로 호출하자; 결과를 대입하지 말자.
```python
lst = [1, 2, 3]
lst.append(4)          # ✅ correct — no assignment needed
# lst = lst.append(4)  # ❌ lst is now None
```

### Mistake 2 — Confusing `.remove(value)` with `.pop(index)`
`.remove()` takes a *value* and deletes its first occurrence. `.pop()` takes an *index* and deletes whatever sits at that position. Calling `lst.remove(0)` does not remove the item at index 0 — it removes the first element whose *value* is `0`, which may not exist and raises `ValueError`.

`.remove()`는 *값*을 받아 첫 번째로 나타나는 위치를 삭제한다. `.pop()`은 *인덱스*를 받아 그 위치의 항목을 삭제한다. `lst.remove(0)`은 인덱스 0의 항목을 제거하는 게 아니라, *값*이 `0`인 첫 번째 요소를 제거한다 — 그런 값이 없으면 `ValueError`가 발생한다.

✅ **Fix / 해결법:**  
Use `.pop(i)` for "delete by position" and `.remove(v)` for "delete by value."  
"위치로 삭제"에는 `.pop(i)`를, "값으로 삭제"에는 `.remove(v)`를 사용하자.
```python
lst = [10, 20, 30]
lst.pop(0)         # ✅ removes 10 — the item AT index 0
lst.remove(30)       # ✅ removes the VALUE 30, wherever it is
```

### Mistake 3 — Removing items while looping over the same list
Modifying a list's length while a `for` loop is iterating over it causes elements to be silently skipped, because the loop's internal index keeps advancing even as the list shrinks underneath it.

`for` 반복문이 순회하는 도중 리스트의 길이를 바꾸면, 리스트가 줄어드는 동안에도 반복문의 내부 인덱스는 계속 앞으로 나아가기 때문에 요소가 조용히 건너뛰어진다.

✅ **Fix / 해결법:**  
Build a new filtered list with a comprehension instead of removing items mid-loop.  
반복 중간에 항목을 제거하는 대신 컴프리헨션으로 새 필터링된 리스트를 만들자.
```python
lst = [1, 2, 3, 4, 5, 6]
lst = [x for x in lst if x % 2 != 0]   # ✅ removes all even numbers safely
```

# 💡 Tips (팁)
Useful tips or shortcuts (유용한 팁이나 단축법)

- `x in lst` before `.remove(x)` avoids a `ValueError` when the value might not be present — the same defensive pattern used for `.index()`.  
`.remove(x)` 전에 `x in lst`를 확인하면 값이 없을 수도 있을 때 `ValueError`를 피할 수 있다 — `.index()`에도 쓰이는 같은 방어 패턴이다.
- `lst.pop()` (no argument) and `lst.append(x)` are both O(1) — prefer them over `lst.pop(0)` and `lst.insert(0, x)`, which are O(n) because everything after them has to shift.  
`lst.pop()`(인자 없음)과 `lst.append(x)`는 둘 다 O(1)이다 — 이후 모든 요소가 밀려야 하는 O(n)인 `lst.pop(0)`, `lst.insert(0, x)`보다 선호하자.
- `list.copy()` and `lst[:]` only copy one level deep — use `copy.deepcopy()` whenever the list contains other lists or dicts that also need to be independent.  
`list.copy()`와 `lst[:]`는 한 단계까지만 복사한다 — 리스트 안에 독립적으로 만들어야 할 다른 리스트나 딕셔너리가 있다면 `copy.deepcopy()`를 사용하자.
- `sorted(lst, key=..., reverse=True)` in one call is usually clearer than sorting ascending and then reversing separately.  
`sorted(lst, key=..., reverse=True)`를 한 번에 호출하는 것이 오름차순 정렬 후 따로 뒤집는 것보다 대체로 더 명확하다.
- A list comprehension like `[x for x in lst if condition]` is almost always the safer, more Pythonic replacement for looping with `.remove()` or `.pop()` mid-iteration.  
`[x for x in lst if condition]` 같은 리스트 컴프리헨션은 반복 중간에 `.remove()`나 `.pop()`을 쓰는 것보다 거의 항상 더 안전하고 Pythonic한 대안이다.

# 💼 Business Example (비즈니스 예제)
*How would a Business Analyst use this? BA는 이것을 실제로 어떻게 사용할까요?*

**Scenario (시나리오):** A support queue is coming in as a plain Python list of ticket dicts. You need to process it: pull off the oldest ticket, sort the rest by priority, and remove any that get cancelled — using nothing but list methods.

**시나리오:** 지원 티켓 큐가 순수 Python 딕셔너리 리스트로 들어온다. 가장 오래된 티켓을 꺼내고, 나머지를 우선순위로 정렬하고, 취소된 티켓을 제거해야 한다 — 오직 리스트 메서드만으로.

**To do (할 일 목록):**
- [x] Use `.pop(0)` to pull the oldest ticket off the front of the queue.  
`.pop(0)`으로 큐 맨 앞의 가장 오래된 티켓을 꺼낸다.
- [x] Use `.sort()` with `key=` to order the remaining tickets by priority.  
`key=`를 사용한 `.sort()`로 남은 티켓을 우선순위로 정렬한다.
- [x] Use a list comprehension to drop any cancelled tickets safely.  
리스트 컴프리헨션으로 취소된 티켓을 안전하게 제거한다.
- [x] Use `.copy()` before experimenting, so the original queue is never lost.  
원본 큐를 절대 잃지 않도록 실험 전에 `.copy()`를 사용한다.

````python
ticket_queue = [
    {"id": "T1", "priority": 2, "status": "open"},
    {"id": "T2", "priority": 1, "status": "cancelled"},
    {"id": "T3", "priority": 3, "status": "open"},
    {"id": "T4", "priority": 1, "status": "open"},
]

# Work on a copy so the original queue is never accidentally lost
# 원본 큐를 절대 잃지 않도록 복사본에서 작업
queue = ticket_queue.copy()

# Pull the oldest ticket off the front (pop(0) is fine at this small scale)
# 맨 앞의 가장 오래된 티켓을 꺼냄 (이 작은 규모에서는 pop(0)으로 충분)
oldest = queue.pop(0)
print("Handling first (가장 먼저 처리):", oldest)

# Drop cancelled tickets safely with a comprehension, not by mutating mid-loop
# 반복 중간에 변경하지 않고 컴프리헨션으로 안전하게 취소된 티켓 제거
queue = [t for t in queue if t["status"] != "cancelled"]

# Sort the rest by priority, lowest number first
# 나머지를 우선순위로 정렬, 낮은 숫자가 먼저
queue.sort(key=lambda t: t["priority"])

print("Remaining queue, sorted by priority (우선순위로 정렬된 남은 큐):")
for t in queue:
    print(f"  {t['id']} priority={t['priority']}")

print("-" * 40)
print("Original queue untouched (원본 큐는 그대로):", [t["id"] for t in ticket_queue])
````

# 📝 Summary (요약)
*Write today's concept in 3~5 sentences. 오늘 배운 개념을 3~5문장으로 정리해보세요.*

List methods split cleanly into two families: ones that mutate in place and return `None` (`.append()`, `.insert()`, `.extend()`, `.remove()`, `.sort()`, `.reverse()`, `.clear()`) and ones that return a value without necessarily mutating (`.pop()` removes AND returns, `.index()` and `.count()` only search). The single habit worth building above all others is never assigning the result of a mutating method — `x = lst.append(y)` and `x = lst.sort()` are both bugs, since both silently set `x` to `None`. Performance also depends on which end of the list is touched: `.append()`/`.pop()` are O(1), while `.insert(0, x)`/`.pop(0)` are O(n) because every other element has to shift. `.copy()` and slicing (`lst[:]`) only copy one level deep, so a list of lists still needs `copy.deepcopy()` to be fully independent. These same distinctions — mutate vs. return-new, shallow vs. deep — resurface directly as pandas' `inplace=` parameter and its own copy-vs-view behavior.

리스트 메서드는 크게 두 부류로 나뉜다: 제자리에서 변경하고 `None`을 반환하는 것들(`.append()`, `.insert()`, `.extend()`, `.remove()`, `.sort()`, `.reverse()`, `.clear()`)과, 반드시 변경하지는 않으면서 값을 반환하는 것들(`.pop()`은 제거하면서 반환, `.index()`와 `.count()`는 검색만 한다)이다. 무엇보다 길러야 할 습관 하나는 변경형 메서드의 결과를 절대 대입하지 않는 것이다 — `x = lst.append(y)`와 `x = lst.sort()`는 둘 다 버그다, `x`가 조용히 `None`이 되기 때문이다. 성능은 리스트의 어느 쪽 끝을 건드리는지에도 달려 있다: `.append()`/`.pop()`은 O(1)이지만, `.insert(0, x)`/`.pop(0)`은 나머지 모든 요소가 밀려야 하므로 O(n)이다. `.copy()`와 슬라이싱(`lst[:]`)은 한 단계까지만 복사하므로, 리스트의 리스트가 완전히 독립적이려면 `copy.deepcopy()`가 필요하다. 이런 구분들 — 변경 vs 새로 반환, 얕은 vs 깊은 — 은 pandas의 `inplace=` 파라미터와 자체적인 복사-vs-뷰 동작으로 그대로 다시 등장한다.
