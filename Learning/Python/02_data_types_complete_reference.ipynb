# 🧠 Concept (개념)

## What is it? (이게 뭔가요?)
*(Explain it in your own words. 자신의 언어로 설명해보세요.)*

Python organizes data into a small number of built-in types: numbers (`int`, `float`), an immutable sequence of characters (`str`), and four container types — `list`, `tuple`, `dict`, and `set` — each with a distinct mutability and ordering contract. JS/TS collapses most of this into `Array`, `Object`, and (less commonly) `Map`/`Set`, so the biggest adjustment is not learning new concepts, it is learning which of these four containers is the *right default* for a given job.

Python은 데이터를 소수의 내장 타입으로 정리한다: 숫자형(`int`, `float`), 불변 문자 시퀀스인 `str`, 그리고 각기 다른 가변성과 순서 규약을 가진 네 가지 컨테이너 — `list`, `tuple`, `dict`, `set`이다. JS/TS는 이 대부분을 `Array`, `Object`, (상대적으로 덜 쓰이는) `Map`/`Set`으로 뭉뚱그리기 때문에, 가장 크게 적응해야 할 부분은 새로운 개념이 아니라 상황에 맞는 "기본 컨테이너"를 고르는 감각이다.

| | JS/TS | Python |
|---|---|---|
| Ordered, mutable list 순서 있는 가변 리스트 | `Array` | `list` |
| Ordered, immutable list 순서 있는 불변 리스트 | (no direct equivalent — `Object.freeze(arr)`) | `tuple` |
| Key-value store 키-값 저장소 | `Object` / `Map` | `dict` |
| Unique-value collection 중복 없는 값 모음 | `Set` | `set` |
| String 문자열 | `String` (immutable) | `str` (immutable) |
| Template literal 템플릿 리터럴 | `${name} is ${age}` (JS/TS backtick string) | `f"{name} is {age}"` |
| Negative indexing 음수 인덱싱 | `arr.at(-1)` | `arr[-1]` (built in everywhere) |
| Numbers 숫자 | one `Number` type (always float) | `int` (arbitrary precision) + `float` separately |

The type that has no JS/TS equivalent at all is `tuple` — an ordered, immutable sequence. It exists because sometimes you *want* a collection that cannot be accidentally mutated, such as coordinates, a function's multiple return values, or a dictionary key (which must be hashable, so a `list` cannot be used but a `tuple` can).

JS/TS에 전혀 대응되는 개념이 없는 타입은 `tuple`이다 — 순서가 있고 불변인 시퀀스다. 좌표값, 함수의 다중 반환값, 딕셔너리 키(해시 가능해야 하므로 `list`는 안 되지만 `tuple`은 가능)처럼 실수로 변경되면 안 되는 컬렉션이 필요할 때 존재 의미가 있다.

## Why do we use it? (왜 사용하나요?)
*(When is it useful? 언제 유용한가요?)*

Choosing the right type up front prevents entire categories of bugs later: using a `tuple` for a fixed-shape record (like a coordinate) makes accidental mutation a `TypeError` instead of a silent data corruption; using a `set` instead of a `list` for membership checks turns an `O(n)` scan into an `O(1)` lookup; using a `dict` instead of parallel lists keeps related data physically attached to its key.

타입을 처음부터 올바르게 고르면 나중에 발생할 여러 종류의 버그를 예방할 수 있다: 좌표처럼 크기가 고정된 레코드에 `tuple`을 쓰면 실수로 값을 바꾸려 할 때 조용한 데이터 손상 대신 `TypeError`가 발생한다. 멤버십 검사에 `list` 대신 `set`을 쓰면 `O(n)` 탐색이 `O(1)` 조회로 바뀐다. 병렬 리스트 대신 `dict`를 쓰면 관련 데이터가 물리적으로 키에 묶여 있게 된다.

Strings and numbers being the way they are also matters for correctness: Python's `int` never silently overflows the way JS numbers can near `Number.MAX_SAFE_INTEGER`, and `float` arithmetic carries the same binary rounding surprises (`0.1 + 0.2 != 0.3`) that JS has — knowing this in advance avoids chasing a "why is my total off by a cent" bug in a spreadsheet-adjacent script.

문자열과 숫자의 동작 방식도 정확성에 영향을 준다: Python의 `int`는 JS의 `Number.MAX_SAFE_INTEGER` 근처에서처럼 조용히 오버플로되지 않는다. 반면 `float` 연산은 JS와 똑같은 이진 반올림 문제(`0.1 + 0.2 != 0.3`)를 갖고 있다 — 이를 미리 알고 있으면 스프레드시트 관련 스크립트에서 "왜 합계가 1센트 어긋나지" 같은 버그를 쫓아다니지 않아도 된다.

## When is it used in Business Analytics? (Business Analytics에서는 언제 쓰이나요?)
*(Real-world use case 실제 사용 사례)*

Every pandas `DataFrame` is, under the hood, built from these same primitives: a column of numbers is a sequence of `int`/`float`, row labels are often `str`, and `df.to_dict()` produces exactly the `dict` structure covered here. Long before a BA opens pandas, they are usually holding raw JSON or CSV data as nested `list`/`dict` structures and need to reshape it by hand.

모든 pandas `DataFrame`은 결국 이와 같은 기본 타입들로 만들어진다: 숫자 컬럼은 `int`/`float`의 시퀀스이고, 행 라벨은 종종 `str`이며, `df.to_dict()`는 정확히 여기서 다루는 `dict` 구조를 만든다. pandas를 열기 훨씬 전부터, BA는 보통 JSON이나 CSV 원본 데이터를 중첩된 `list`/`dict` 구조로 들고 있으며 이를 손으로 재구성해야 한다.

`set` shows up constantly for quick data-quality checks — "which customer IDs are in file A but not file B?" is one line (`set(a) - set(b)`) instead of a nested loop — and dict/set comprehensions replace what would otherwise be several lines of manual filtering logic.

`set`은 빠른 데이터 품질 검사에서 항상 등장한다 — "파일 A에는 있지만 파일 B에는 없는 고객 ID는?"이라는 질문이 중첩 반복문 대신 한 줄(`set(a) - set(b)`)로 해결된다. 딕셔너리/셋 컴프리헨션은 그렇지 않았다면 여러 줄이 필요했을 수동 필터링 로직을 대체한다.

# 📝 Syntax (문법)

## Basic Syntax (기본 문법)

````python
# One example of each core type
# 핵심 타입 하나씩
n = 42                   # int
flt = 3.14                 # float
s = "hello"                  # str
lst = [1, 2, 3]                  # list — mutable, ordered
tpl = (1, 2, 3)                      # tuple — immutable, ordered
d = {"a": 1, "b": 2}                     # dict — key/value, ordered since 3.7
st = {1, 2, 3}                              # set — unique, unordered

for value in (n, flt, s, lst, tpl, d, st):
    print(type(value).__name__, value)

# f-string — the standard way to build strings with values
# f-string — 값을 넣어 문자열을 만드는 표준 방법
name, age = "Booki", 28
print(f"{name} is {age} years old")

# Indexing & slicing work the same way across str / list / tuple
# 인덱싱과 슬라이싱은 str / list / tuple에서 동일하게 동작
print(s[0], s[-1], s[1:3])
print(lst[0], lst[-1], lst[1:3])
````

## Common Variations (자주 쓰는 변형)

````python
# Underscore for readable large numbers
# 가독성을 위한 언더바
population = 51_780_000
print(population)

# Number base literals — all equal 10
# 진법 리터럴 — 전부 10과 같음
print(0b1010, 0o12, 0x0A)

# Type conversion — int() truncates, does not round
# 타입 변환 — int()는 버림이지 반올림이 아님
print(int(3.9))       # 3, not 4
print(int("ff", 16))   # 255 — hex string to int

# Multi-line string
# 여러 줄 문자열
note = """Line one
Line two"""
print(note)

# Dict merge (3.9+) and dict comprehension
# 딕셔너리 병합 (3.9+)과 딕셔너리 컴프리헨션
base_settings = {"theme": "light", "lang": "en"}
overrides = {"theme": "dark"}
print(base_settings | overrides)
squares = {x: x**2 for x in range(5)}
print(squares)

# Safe 2D list initialization
# 안전한 2D 리스트 초기화
grid = [[0] * 3 for _ in range(3)]
print(grid)
````

# 🧪 Small Examples (예제)

## Example 1 — Numeric Types (숫자형)

````python
# int has unlimited precision — no overflow like JS numbers
# int는 무한 정밀도 — JS 숫자처럼 오버플로되지 않음
big = 10 ** 100
print(len(str(big)), "digits, computed exactly")

# float precision — the classic gotcha shared with JS
# float 정밀도 — JS와 공유하는 대표적인 함정
print(0.1 + 0.2)               # 0.30000000000000004
print(round(0.1 + 0.2, 2))     # 0.3

# Type conversion
# 타입 변환
print(int("42"))         # 42
print(int(3.7))            # 3  — truncates, not round()
print(int(True))            # 1
print(float("inf"))          # inf
print(bool(0), bool(1))       # False True

# isinstance across numeric types — True is technically an int too
# 숫자 타입에 대한 isinstance — True도 사실은 int
for value in (5, 5.0, True):
    print(value, "-> int?", isinstance(value, int))
````

## Example 2 — Strings (문자열)

````python
s = "Hello, World!"

# Strings are immutable + iterable
# 문자열은 불변이며 순회 가능
print(s[0], s[-1], s[7:12], s[::-1])
for ch in s[:5]:
    print(ch, end=" ")
print()

# f-string formatting - the format spec after the colon does the heavy lifting
# f-string 포맷 - 콜론 뒤의 포맷 스펙이 핵심
price = 1234.5
print(f"{price:,.2f}")        # 1,234.50
print(f"{price:>12,.2f}")      # right-aligned in a 12-char field

# Raw strings — no escape processing, useful for paths & regex
# raw string — 이스케이프 처리를 하지 않음, 경로/정규식에 유용
path = r"C:\Users\booki"
print(path)
print(len(path))    # backslashes count as real characters

# concatenation & repetition
# 결합과 반복
print("ab" + "cd")
print("ab" * 3)
````

## Example 3 — Lists (리스트)

````python
lst = [3, 1, 4, 1, 5, 9]

# Slicing [start:end:step] — end is exclusive
# 슬라이싱 [시작:끝:증감폭] — 끝은 미포함
print(lst[1:3])      # [1, 4]
print(lst[::2])       # [3, 4, 5]
print(lst[::-1])       # reversed copy

# Membership & concatenation
# 멤버십 검사와 결합
print(9 in lst)
print(lst + [100])

# Lists are mutable — this is the key difference from tuples
# 리스트는 mutable — tuple과의 핵심 차이
lst.append(2)
print(lst)

# Safe vs unsafe 2D initialization
# 안전한 vs 위험한 2D 초기화
safe = [[0] * 3 for _ in range(3)]
unsafe = [[0] * 3] * 3            # all 3 rows are the SAME list object!
safe[0][0] = 9
unsafe[0][0] = 9
print("safe:  ", safe)        # only row 0 changed
print("unsafe:", unsafe)       # every row changed — the trap
````

## Example 4 — Dictionaries (딕셔너리)

````python
d = {"name": "Booki", "role": "BA-in-training"}

# Access, add, delete
# 접근, 추가, 삭제
print(d["name"])
d["city"] = "Vancouver"
del d["role"]
print(d)
print("city" in d)

# Merging (Python 3.9+) — right side wins on key conflicts
# 병합 (Python 3.9+) — 충돌 시 오른쪽이 우선
base_settings = {"theme": "light", "notifications": True}
user_prefs = {"theme": "dark"}
merged = base_settings | user_prefs
print(merged)

# Dict comprehension — build a lookup table in one line
# 딕셔너리 컴프리헨션 — 한 줄로 조회 테이블 만들기
prices = {"coffee": 4.5, "tea": 3.5, "juice": 5.5}
expensive = {item: price for item, price in prices.items() if price > 4}
print(expensive)

# Since Python 3.7, dicts preserve insertion order — reliable for reporting
# Python 3.7부터 딕셔너리는 삽입 순서를 유지 — 리포팅에 신뢰 가능
for item, price in prices.items():
    print(item, price)
````

## Example 5 — Sets (셋)

````python
raw_ids = [101, 102, 101, 103, 102, 104]

# Sets automatically drop duplicates
# 셋은 중복을 자동으로 제거
unique_ids = set(raw_ids)
print(unique_ids)

# O(1) membership check - much faster than a list scan for large data
# O(1) 멤버십 검사 - 큰 데이터에서 리스트 탐색보다 훨씬 빠름
print(103 in unique_ids)

# Set algebra — a one-line answer to common data-quality questions
# 셋 연산 — 흔한 데이터 품질 질문에 대한 한 줄 답
file_a = {"cust_1", "cust_2", "cust_3"}
file_b = {"cust_2", "cust_3", "cust_4"}
print("in both:       ", file_a & file_b)      # intersection
print("only in A:     ", file_a - file_b)      # difference
print("in either:     ", file_a | file_b)      # union
print("in exactly one:", file_a ^ file_b)      # symmetric difference
````

# ⚠️ Common Mistakes (자주 하는 실수)

### Mistake 1 — `[[0]*3]*3` creates 3 references to the SAME row
`[0]*3` makes one list `[0, 0, 0]`. Multiplying THAT by 3 does not create three independent copies — it creates a list containing the same inner list object three times. Changing one "row" silently changes all of them.

`[0]*3`은 리스트 하나 `[0, 0, 0]`을 만든다. 그걸 다시 3을 곱하면 독립적인 복사본 3개가 아니라, 같은 내부 리스트 객체를 세 번 담은 리스트가 된다. 한 "행"을 바꾸면 조용히 전부 바뀐다.

✅ **Fix / 해결법:**  
Use a list comprehension so each row is created fresh.  
각 행이 새로 생성되도록 리스트 컴프리헨션을 사용하자.
```python
matrix = [[0] * 3 for _ in range(3)]   # ✅ 3 independent rows
```

### Mistake 2 — `{}` creates an empty dict, not an empty set
It is easy to assume `{}` is the empty version of `{1, 2, 3}`, but Python already claimed `{}` for the empty dictionary. Writing `s = {}` and then calling `s.add(1)` fails with an `AttributeError` because `s` is actually a `dict`.

`{}`가 `{1, 2, 3}`의 빈 버전이라고 생각하기 쉽지만, Python은 이미 `{}`를 빈 딕셔너리로 정해두었다. `s = {}`라고 쓰고 `s.add(1)`을 호출하면 `s`가 사실 `dict`이기 때문에 `AttributeError`가 발생한다.

✅ **Fix / 해결법:**  
Use `set()` to create an empty set.  
빈 셋을 만들 때는 `set()`을 사용하자.
```python
s = set()   # ✅ empty set, not {}
```

### Mistake 3 — Comparing floats directly with `==`
Because of binary floating-point rounding, `0.1 + 0.2 == 0.3` evaluates to `False`. This surfaces constantly in anything that sums money or percentages.

이진 부동소수점 반올림 때문에 `0.1 + 0.2 == 0.3`은 `False`로 평가된다. 금액이나 퍼센트를 합산하는 모든 코드에서 계속 나타나는 문제다.

✅ **Fix / 해결법:**  
Round both sides before comparing, or compare within a small tolerance.  
비교 전에 양쪽을 반올림하거나, 작은 오차 범위 내에서 비교하자.
```python
print(round(0.1 + 0.2, 2) == round(0.3, 2))   # ✅ True
```

# 💡 Tips (팁)
Useful tips or shortcuts (유용한 팁이나 단축법)

- `list`, `dict`, and `set` are mutable — pass them into a function and the function can change the original. `tuple` and `str` are immutable, so they are always safe to hand off without copying.  
`list`, `dict`, `set`은 mutable이라 함수에 넘기면 함수가 원본을 바꿀 수 있다. `tuple`과 `str`은 immutable이라 복사 없이 넘겨도 항상 안전하다.
- f-string format specs (`:.2f`, `:,`, `:>10`, `:.1%`) cover almost every formatting need — reach for them before writing manual string padding or rounding logic.  
f-string 포맷 스펙(`:.2f`, `:,`, `:>10`, `:.1%`)은 거의 모든 포맷 요구를 처리한다 — 수동으로 패딩이나 반올림 로직을 짜기 전에 먼저 떠올리자.
- Need a fast "does this exist?" check on a large collection? Convert it to a `set` first — membership checks on a `set` are O(1), while on a `list` they are O(n).  
큰 컬렉션에서 빠른 "이게 존재하나?" 검사가 필요하면 먼저 `set`으로 변환하자 — `set`의 멤버십 검사는 O(1)이지만 `list`는 O(n)이다.
- A `tuple` is the natural choice for a dictionary key made of multiple values (e.g. `(year, month)`), since `list` cannot be hashed but `tuple` can.  
여러 값으로 이루어진 딕셔너리 키(예: `(year, month)`)에는 `tuple`이 자연스러운 선택이다. `list`는 해시가 불가능하지만 `tuple`은 가능하기 때문이다.
- `d.get(key, default)` is almost always safer than `d[key]` when the key might be missing — it returns a fallback instead of raising `KeyError`.  
키가 없을 수도 있다면 `d[key]`보다 `d.get(key, default)`가 거의 항상 안전하다 — `KeyError`를 던지는 대신 대체값을 반환한다.

# 💼 Business Example (비즈니스 예제)
*How would a Business Analyst use this? BA는 이것을 실제로 어떻게 사용할까요?*

**Scenario (시나리오):** You have pulled a raw customer list from two different systems — the CRM export and the billing export — as plain Python `list`/`dict` data. Before this ever becomes a pandas merge, you need to spot data-quality issues using basic types.

**시나리오:** 서로 다른 두 시스템 — CRM 내보내기와 청구 내보내기 — 에서 고객 리스트를 순수 Python `list`/`dict` 데이터로 받았다. pandas 병합으로 넘어가기 전에, 기본 타입만으로 데이터 품질 문제를 먼저 확인해야 한다.

**To do (할 일 목록):**
- [x] Use `set` operations to find customer IDs present in one file but not the other. `set` 연산으로 한 파일에만 있는 고객 ID를 찾는다.
- [x] Build a `dict` lookup from customer ID to name for fast access. 빠른 조회를 위해 고객 ID → 이름 딕셔너리를 만든다.
- [x] Use an f-string to format a clean summary line per mismatch. f-string으로 불일치 건별 요약 줄을 깔끔하게 포맷한다.
- [x] Store each mismatch as an immutable `tuple` record, since it should not be edited after the fact. 이후 수정되면 안 되므로 각 불일치 기록을 불변 `tuple`로 저장한다.

````python
crm_customers = {
    "C001": "Kim Song",
    "C002": "Lee Jun",
    "C003": "Park Min",
    "C004": "Choi Ye",
}
billing_customer_ids = {"C002", "C003", "C005"}   # C005 is not in the CRM at all

crm_ids = set(crm_customers.keys())

only_in_crm = crm_ids - billing_customer_ids
only_in_billing = billing_customer_ids - crm_ids
in_both = crm_ids & billing_customer_ids

print(f"Customers in CRM only:     {sorted(only_in_crm)}")
print(f"Customers in billing only: {sorted(only_in_billing)}")
print(f"Customers in both:         {sorted(in_both)}")

# Immutable mismatch records - a (customer_id, issue) tuple should not be
# edited after it is logged, so a tuple is a better fit than a list here.
# 불변 불일치 기록 - (고객ID, 이슈) 튜플은 로그된 후 수정되면 안 되므로
# list보다 tuple이 더 알맞다.
mismatches = []
for cid in sorted(only_in_crm):
    mismatches.append((cid, "missing_from_billing"))
for cid in sorted(only_in_billing):
    mismatches.append((cid, "missing_from_crm"))

print("-" * 45)
for cid, issue in mismatches:
    name = crm_customers.get(cid, "Unknown")
    print(f"{cid} | {name:<10} | {issue}")
````

# ❓ Review Questions (복습 문제)

**Q1.** Why does `[[0]*3]*3` produce three rows that all change together, and how does `[[0]*3 for _ in range(3)]` fix it?  
`[[0]*3]*3`이 왜 함께 바뀌는 세 개의 행을 만들어내며, `[[0]*3 for _ in range(3)]`은 이를 어떻게 해결하나요?

`[[0]*3]*3` copies the same inner list, so all rows share the same object. A list comprehension creates three separate lists.  
`[[0]*3]*3`은 같은 내부 리스트를 복사하므로 같은 메모리를 공유합니다. 리스트 컴프리헨션은 서로 다른 리스트 3개를 생성합니다.

**Q2.** What type does `{}` create, and how do you create an empty set instead?  
`{}`는 어떤 타입을 만드나요, 그리고 빈 셋은 어떻게 만드나요?

`{}` creates an empty dictionary. Use `set()` to create an empty set.  
`{}`는 빈 딕셔너리를 만듭니다. 빈 셋은 `set()`으로 생성합니다.

**Q3.** Given `a = {1, 2, 3}` and `b = {2, 3, 4}`, what do `a & b`, `a - b`, and `a ^ b` each return?  
`a = {1, 2, 3}`이고 `b = {2, 3, 4}`일 때, `a & b`, `a - b`, `a ^ b`는 각각 무엇을 반환하나요?

a & b -> {2, 3} (intersection, 교집합)  
a - b -> {1} (difference, 차집합)  
a ^ b -> {1, 4} (symmetric difference, 대칭차집합)

**Q4.** Why can a `tuple` be used as a dictionary key while a `list` cannot?  
왜 `tuple`은 딕셔너리 키로 쓸 수 있지만 `list`는 쓸 수 없나요?

A tuple is immutable and hashable, so it can be used as a dictionary key. A list is mutable, so it cannot.  
Tuple은 변경 불가능(immutable) 하고 hashable이라 딕셔너리 키로 사용할 수 있습니다. List는 변경 가능하므로 사용할 수 없습니다.

**Q5.** Why does `0.1 + 0.2 == 0.3` evaluate to `False` in Python, and what is the safer way to compare two floats?  
Python에서 `0.1 + 0.2 == 0.3`이 왜 `False`로 평가되나요? 그리고 두 float을 더 안전하게 비교하는 방법은 무엇인가요?  


Floating-point numbers cannot always be represented exactly. Use `math.isclose()` instead of `==`.  
부동소수점은 정확하게 표현되지 않을 수 있으므로 `==` 대신 `math.isclose()`를 사용합니다.
