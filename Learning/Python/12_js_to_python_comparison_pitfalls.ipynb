# 📝 Syntax (문법)

## Basic Syntax (기본 문법)

````python
# The five translations you'll use constantly
# 항상 쓰게 될 다섯 가지 번역
arr = [1, 2, 3]
print(len(arr))                          # JS: arr.length
print(arr[-1])                             # JS: arr.at(-1)
print([x * 2 for x in arr])                  # JS: arr.map(x => x*2)
status = "yes" if True else "no"               # JS: true ? "yes" : "no"
print(status)
d = {"a": 1}
print(d.get("b", "default"))                     # JS: d.b ?? "default"
````

## Common Variations (자주 쓰는 변형)

````python
import json

# A few more common translations
# 몇 가지 더 흔한 번역
print(3 in [1, 2, 3])                     # JS: [1,2,3].includes(3)

people = [{"age": 30}, {"age": 25}]
print(sorted(people, key=lambda p: p["age"]))     # JS: people.sort((a,b) => a.age - b.age)

# Falsy values differ slightly in which values count — but the concept is the same
# falsy 값은 어떤 값이 포함되는지 살짝 다르지만 개념은 같음
print(bool(0), bool(""), bool([]), bool(None))       # all False, same spirit as JS falsy values

print(json.dumps({"ok": True}))                         # JS: JSON.stringify({ok: true})
````

# 🧪 Small Examples (예제)

## Example 1 — Syntax Basics Comparison (기본 문법 비교표)

````python
# Comments — # instead of //
# 주석 — //대신 #
# This is a comment, just like // in JS/TS

# Variables — no const/let/var keyword needed
# 변수 — const/let/var 키워드 불필요
x = 10                    # JS: const x = 10;

# Booleans and None — capitalized, and None instead of null/undefined
# 불리언과 None — 대문자, null/undefined 대신 None
is_active = True            # JS: true
is_deleted = False            # JS: false
missing = None                   # JS: null or undefined

# Logical operators — words instead of symbols
# 논리 연산자 — 기호 대신 단어
print(True and False)             # JS: &&
print(True or False)                # JS: ||
print(not True)                       # JS: !

# Ternary — condition sits in the MIDDLE, not first
# 삼항 연산자 — 조건이 처음이 아니라 중간에 옴
age = 20
status = "adult" if age >= 18 else "minor"     # JS: age >= 18 ? "adult" : "minor"
print(status)

# for-loop — range() replaces the classic three-part for(;;)
# for 반복문 — range()가 고전적인 for(;;)를 대체
for i in range(3):                              # JS: for (let i=0; i<3; i++)
    print(i)

# match-case — Python's switch-case (3.10+)
# match-case — Python의 switch-case (3.10 이상)
code = 200
match code:
    case 200:
        print("OK")
    case 404:
        print("Not Found")
    case _:
        print("Unknown")
````

## Example 2 — Collections & Methods Comparison (컬렉션·메서드 비교표)

````python
from functools import reduce

arr = [3, 1, 4, 1, 5]

# Length, add, remove
# 길이, 추가, 제거
print(len(arr))                       # JS: arr.length
arr.append(9)                           # JS: arr.push(9)
arr.pop()                                 # JS: arr.pop()

# Front operations — O(n) on a list in both languages, but named differently
# 앞쪽 연산 — 두 언어 모두 리스트에서 O(n), 이름만 다름
arr.insert(0, 99)                          # JS: arr.unshift(99)
arr.pop(0)                                   # JS: arr.shift()

# Search
# 검색
print(3 in arr)                                # JS: arr.includes(3)
print(arr.index(4))                              # JS: arr.indexOf(4) -- raises ValueError if missing, JS returns -1!

# Transform / filter / reduce — comprehensions and functools replace array methods
# 변환/필터/축약 — 컴프리헨션과 functools가 배열 메서드를 대체
print([x * 2 for x in arr])                        # JS: arr.map(x => x*2)
print([x for x in arr if x > 3])                      # JS: arr.filter(x => x>3)
print(reduce(lambda a, b: a + b, arr))                   # JS: arr.reduce((a,b)=>a+b)

# some / every
# some / every
print(any(x > 10 for x in arr))                            # JS: arr.some(x => x>10)
print(all(x > 0 for x in arr))                               # JS: arr.every(x => x>0)

# join — called on the SEPARATOR in Python, the reverse of JS
# join — Python에서는 구분자에서 호출, JS와 반대
print(",".join(str(x) for x in arr))                          # JS: arr.join(",")

# Spread / merge
# 펼치기 / 병합
a, b = [1, 2], [3, 4]
print([*a, *b])                                                  # JS: [...a, ...b]
print(a + b)                                                       # same result, simpler syntax (같은 결과, 더 단순한 문법)
````

## Example 3 — Modules & Classes Comparison (모듈·클래스 비교표)

````python
import math                                         # JS: import * as math from 'math'
from math import sqrt                                 # JS: import { sqrt } from 'math'

d = {"name": "Booki", "role": "BA"}

# Object-equivalent access
# Object에 대응하는 접근
print(list(d.keys()))                   # JS: Object.keys(d)
print(list(d.values()))                   # JS: Object.values(d)
print(list(d.items()))                      # JS: Object.entries(d)

# Console output, type checking, parsing
# 콘솔 출력, 타입 확인, 파싱
print("hello")                                # JS: console.log("hello")
print(type(42))                                 # JS: typeof x
print(int("42"), float("3.14"))                   # JS: parseInt/parseFloat
print(str(42))                                      # JS: String(x)

# Map/Set equivalents
# Map/Set 대응물
m = dict()                                             # JS: new Map()
s = set()                                                # JS: new Set()

# Classes and inheritance
# 클래스와 상속
class Animal:
    def __init__(self, name):                 # JS: constructor(name)
        self.name = name

class Dog(Animal):                               # JS: class Dog extends Animal
    pass

print(Dog("Rex").name)

# Import styles
# import 스타일
print(math.pi, sqrt(16))

# No destructuring from a dict -- must pull keys out manually
# 딕셔너리에서 구조분해 불가 -- 수동으로 키를 꺼내야 함
# JS: const {name, role} = d;
name, role = d["name"], d["role"]                        # the Pythonic workaround (Pythonic한 대안)
print(name, role)
````

## Example 4 — Anti-patterns I (함정 정리 I)

````python
# -- Anti-pattern 1: append()/sort() return None -- see Section 3 & 5 --
lst = [3, 1, 2]
correct = sorted(lst)              # correct: new list, returns a value (정답: 새 리스트를 반환)
# wrong = lst.sort()               # wrong: would silently be None (오답: 조용히 None이 됨)
print("1) sorted() returns a value:", correct)

# -- Anti-pattern 2: mutable default arguments -- see Section 9 --
def add_item(item, bucket=None):        # correct: None default, not [] (정답: [] 대신 None 기본값)
    if bucket is None:
        bucket = []
    bucket.append(item)
    return bucket
print("2) safe default:", add_item("a"), add_item("b"))

# -- Anti-pattern 3: [[0]*3]*3 shares one row across all three -- see Section 1 & 2 --
safe_grid = [[0] * 3 for _ in range(3)]        # correct: independent rows (정답: 독립적인 행)
print("3) independent rows:", safe_grid)

# -- Anti-pattern 4: mutating a list while looping over it -- see Section 5 --
nums = [1, 2, 3, 4, 5, 6]
odds_only = [x for x in nums if x % 2 != 0]      # correct: comprehension instead of removing mid-loop (정답: 반복 중간에 제거하는 대신 컴프리헨션)
print("4) filtered safely:", odds_only)

# -- Anti-pattern 5: string += in a loop is O(n^2) -- join() is O(n) --
words = ["This", "is", "much", "faster"]
# slow = ""
# for w in words: slow += w + " "      # wrong: O(n^2), each += copies the whole string so far (오답: O(n^2))
fast = " ".join(words)                    # correct: O(n), join() builds the result once (정답: O(n))
print("5) joined efficiently:", fast)
````

# ⚠️ Common Mistakes (자주 하는 실수)

### Mistake 1 — Calling `.map()`/`.filter()` directly on a Python list
JS arrays have built-in `.map()`/`.filter()` methods, but Python lists do not — `[1,2,3].map(...)` raises `AttributeError: 'list' object has no attribute 'map'`. `map()` and `filter()` are Python's global FUNCTIONS, not list methods, and comprehensions are usually preferred anyway.

JS 배열은 내장 `.map()`/`.filter()` 메서드를 갖지만, Python 리스트는 그렇지 않다 — `[1,2,3].map(...)`는 `AttributeError: 'list' object has no attribute 'map'`을 던진다. `map()`과 `filter()`는 Python의 전역 함수이지 리스트 메서드가 아니며, 어차피 보통은 컴프리헨션이 선호된다.

✅ **Fix / 해결법:**  
Use `map(fn, arr)` as a function call, or better, a list comprehension.  
`map(fn, arr)`를 함수 호출로 사용하거나, 더 나은 방법으로 리스트 컴프리헨션을 사용하자.
```python
arr = [1, 2, 3]
result = [x * 2 for x in arr]   # ✅ the idiomatic choice
```

### Mistake 2 — Assuming Python's `==` coerces types like JS's `==` does
JS's `==` performs type coercion (`1 == "1"` is `true` in JS), but Python's `==` never does — `1 == "1"` is `False` in Python. Python's `==` behaves more like JS's strict `===`.

JS의 `==`는 타입 강제 변환을 한다(JS에서 `1 == "1"`은 `true`), 하지만 Python의 `==`는 절대 그러지 않는다 — Python에서 `1 == "1"`은 `False`다. Python의 `==`는 JS의 엄격한 `===`에 더 가깝게 동작한다.

✅ **Fix / 해결법:**  
Convert types explicitly before comparing, since Python will never silently coerce them for you.  
비교하기 전에 명시적으로 타입을 변환하자, Python은 절대 조용히 대신 변환해주지 않는다.
```python
print(1 == "1")             # False in Python -- no coercion
print(1 == int("1"))          # ✅ True -- explicit conversion first
```

### Mistake 3 — Trying to use JS-style destructuring on a dict
JS's object destructuring has no Python equivalent — writing `{name, age} = d` raises a `SyntaxError`. Python's closest tool, tuple unpacking, only works positionally on ordered sequences, not by key name on a dict.

JS의 객체 구조분해는 Python 대응물이 없다 — `{name, age} = d`라고 쓰면 `SyntaxError`가 발생한다. Python에서 가장 가까운 도구인 튜플 언패킹은 딕셔너리에서 키 이름이 아니라 순서가 있는 시퀀스에서 위치 기반으로만 동작한다.

✅ **Fix / 해결법:**  
Pull keys out individually, or use `.get()` for each one.  
키를 개별적으로 꺼내거나, 각각에 `.get()`을 사용하자.
```python
d = {"name": "Booki", "age": 28}
name, age = d["name"], d["age"]    # ✅ the Pythonic workaround
```

# 💼 Business Example (비즈니스 예제)
*How would a Business Analyst use this? BA는 이것을 실제로 어떻게 사용할까요?*

**Scenario (시나리오):** A teammate who is more comfortable in JS/TS sent you a rough pseudocode sketch for a small report script. Before running with it, you need to translate it into idiomatic, trap-free Python.

**시나리오:** JS/TS에 더 익숙한 동료가 작은 리포트 스크립트의 대략적인 의사코드 스케치를 보냈다. 그대로 쓰기 전에, 이를 관용적이고 함정 없는 Python으로 번역해야 한다.

**To do (할 일 목록):**
- [x] Replace JS-style array methods (`.filter()`, `.map()`) with comprehensions.  
JS 스타일 배열 메서드(`.filter()`, `.map()`)를 컴프리헨션으로 바꾼다.
- [x] Fix the `sort()`-return-value trap and the mutable default argument trap.  
`sort()` 반환값 함정과 가변 기본 인자 함정을 고친다.
- [x] Replace `==` against `None` with `is`, and build the summary dict without destructuring syntax.  
`None`과의 `==` 비교를 `is`로 바꾸고, 구조분해 문법 없이 요약 딕셔너리를 만든다.

````python
# The original JS-flavored (buggy) approach, commented out for reference:
# 원본 JS 스타일(버그가 있는) 접근 방식, 참조용으로 주석 처리:
#
# function summarize(orders, tag = []) {
#     const completed = orders.filter(o => o.status === "completed");
#     const amounts = completed.map(o => o.amount);
#     tag.push("processed");                          // mutable default trap in Python
#     return { count: completed.length, total: amounts.reduce((a,b)=>a+b, 0) };
# }

orders = [
    {"status": "completed", "amount": 120, "note": None},
    {"status": "pending",   "amount": 80,  "note": None},
    {"status": "completed", "amount": 300, "note": "rush order"},
]

def summarize(orders, tag=None):                        # fixed: None default, not a mutable list (수정: 가변 리스트 대신 None 기본값)
    if tag is None:                                        # fixed: `is`, not `==` (수정: `==`가 아니라 `is`)
        tag = []
    completed = [o for o in orders if o["status"] == "completed"]     # fixed: comprehension, not .filter() (수정: .filter() 대신 컴프리헨션)
    amounts = [o["amount"] for o in completed]                          # fixed: comprehension, not .map() (수정: .map() 대신 컴프리헨션)
    tag.append("processed")
    return {"count": len(completed), "total": sum(amounts), "tags": tag}    # built manually, no destructuring needed (수동으로 구성, 구조분해 불필요)

result = summarize(orders)
print(result)

# Confirm the mutable-default fix actually works: a second call starts fresh
# 가변 기본값 수정이 실제로 동작하는지 확인: 두 번째 호출은 새로 시작함
result2 = summarize(orders)
print(result2)                    # tags should be ['processed'] again, NOT ['processed', 'processed']
````
