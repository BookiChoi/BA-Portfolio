# 🧠 Concept (개념)

## What is it? (이게 뭔가요?)
*(Explain it in your own words. 자신의 언어로 설명해보세요.)*

This section covers everything about functions that goes beyond "define it, call it, return a value": accepting a flexible number of arguments (`*args`, `**kwargs`), forcing certain arguments to be passed a specific way (keyword-only, positional-only), documenting expected types (type hints), combining assignment with an expression (the walrus operator `:=`), functions that remember state between calls (closures), and functions that wrap other functions to add behavior (decorators).

이 섹션은 "정의하고, 호출하고, 값을 반환한다"를 넘어서는 함수에 관한 모든 것을 다룬다: 유연한 개수의 인자 받기(`*args`, `**kwargs`), 특정 인자를 정해진 방식으로만 전달하도록 강제하기(키워드 전용, 위치 전용), 예상되는 타입 문서화하기(타입 힌트), 할당과 표현식을 결합하기(왈러스 연산자 `:=`), 호출 사이에 상태를 기억하는 함수(클로저), 다른 함수를 감싸 동작을 추가하는 함수(데코레이터)다.

| | JS/TS | Python |
|---|---|---|
| Rest parameters 나머지 매개변수 | `function f(...args)` | `def f(*args)` |
| Spread into keyword args 키워드 인자로 펼치기 | (no direct equivalent) | `def f(**kwargs)` |
| Type annotations 타입 주석 | TypeScript: `function f(x: number): string` | `def f(x: int) -> str:` (optional, not enforced) |
| Assignment inside an expression 표현식 안의 할당 | `if (n = arr.length) { }` (rare, discouraged in JS) | `if (n := len(arr)):` (idiomatic in Python 3.8+) |
| Closures 클로저 | `function makeCounter() { let count = 0; return () => ++count; }` | `def make_counter(): count = 0; def inner(): nonlocal count` |
| Higher-order wrapping function 고차 래핑 함수 | manual HOC pattern, no special syntax | `@decorator` syntax sugar |

The biggest syntax difference: Python has dedicated `@decorator` syntax for the "wrap a function to add behavior" pattern, while JS/TS developers write the equivalent by hand (a higher-order function) without special language support for it.

가장 큰 문법 차이: Python은 "함수를 감싸서 동작을 추가하는" 패턴을 위한 전용 `@decorator` 문법을 갖고 있지만, JS/TS 개발자는 이를 위한 특별한 언어 지원 없이 동등한 것(고차 함수)을 직접 손으로 작성한다.

## Why do we use it? (왜 사용하나요?)
*(When is it useful? 언제 유용한가요?)*

`*args`/`**kwargs` let a single function accept whatever a caller wants to throw at it — essential for wrapper functions and decorators that need to forward arguments to another function without knowing its signature in advance. Keyword-only and positional-only parameters make a function's call signature self-documenting and prevent a specific class of bugs: accidentally passing arguments in the wrong order, or relying on a parameter name that was never meant to be part of the public API.

`*args`/`**kwargs`는 함수 하나가 호출자가 무엇을 던지든 받을 수 있게 해준다 — 다른 함수의 시그니처를 미리 알지 못한 채 인자를 전달해야 하는 래퍼 함수와 데코레이터에 필수적이다. 키워드 전용·위치 전용 인자는 함수의 호출 시그니처를 자기 설명적으로 만들고, 특정 종류의 버그를 예방한다: 인자를 잘못된 순서로 실수로 전달하거나, 공개 API의 일부가 될 의도가 없었던 파라미터 이름에 의존하는 것 등이다.

Decorators exist because "add logging/timing/caching to this function" is such a common need that Python gives it first-class syntax instead of forcing every developer to write the same wrapping boilerplate by hand each time.

데코레이터가 존재하는 이유는 "이 함수에 로깅/시간측정/캐싱을 추가하기"가 너무나 흔한 요구라서, Python이 모든 개발자가 매번 같은 감싸기 보일러플레이트를 직접 작성하게 강요하는 대신 일급 문법을 제공하기 때문이다.

## When is it used in Business Analytics? (Business Analytics에서는 언제 쓰이나요?)
*(Real-world use case 실제 사용 사례)*

A `@timer` or `@retry` decorator around a slow data-pulling function is one of the most common real uses — measuring how long a query takes, or automatically retrying a flaky API call, without cluttering the actual data logic with that concern.

느린 데이터 조회 함수에 `@timer`나 `@retry` 데코레이터를 붙이는 것은 가장 흔한 실제 사용 사례 중 하나다 — 실제 데이터 로직을 그런 관심사로 어지럽히지 않으면서 쿼리에 걸리는 시간을 측정하거나 불안정한 API 호출을 자동으로 재시도하는 일이다.

The walrus operator shows up constantly in cleaning loops — `while (chunk := file.read(1024)):` reads a file in pieces without a separate "read, then check" step, and `[y for x in data if (y := transform(x)) is not None]` computes a value once and reuses it, instead of calling `transform(x)` twice.

왈러스 연산자는 정리용 반복문에서 계속 등장한다 — `while (chunk := file.read(1024)):`는 별도의 "읽고, 그다음 확인"하는 단계 없이 파일을 조각조각 읽으며, `[y for x in data if (y := transform(x)) is not None]`은 `transform(x)`를 두 번 호출하는 대신 값을 한 번 계산해서 재사용한다.

# 📝 Syntax (문법)

## Basic Syntax (기본 문법)

````python
# *args collects extra positional arguments into a tuple
# *args는 추가 위치 인자를 튜플로 모음
# **kwargs collects extra keyword arguments into a dict
# **kwargs는 추가 키워드 인자를 딕셔너리로 모음
def describe(*args, **kwargs):
    print("args:", args)
    print("kwargs:", kwargs)

describe(1, 2, 3, name="booki", role="BA")

# A closure — an inner function that remembers a variable from its enclosing scope
# 클로저 — 감싸는 스코프의 변수를 기억하는 내부 함수
def make_multiplier(factor):
    def multiply(x):
        return x * factor
    return multiply

double = make_multiplier(2)
print(double(5))

# A decorator — a function that wraps another function to add behavior
# 데코레이터 — 다른 함수를 감싸 동작을 추가하는 함수
def announce(func):
    def wrapper(*args, **kwargs):
        print(f"calling {func.__name__}...")
        return func(*args, **kwargs)
    return wrapper

@announce
def add(a, b):
    return a + b

print(add(2, 3))
````

## Common Variations (자주 쓰는 변형)

````python
from typing import Optional

# The walrus operator := assigns AND returns a value in one expression
# 왈러스 연산자 :=는 하나의 표현식에서 할당과 반환을 동시에 함
data = [1, 2, 3, 4, 5, 6, 7, 8]
if (n := len(data)) > 5:
    print(f"data has {n} items, which is more than 5")

# ...reused inside a comprehension to avoid computing twice
# ...컴프리헨션 안에서 재사용해 두 번 계산하지 않음
results = [y for x in range(10) if (y := x ** 2) > 20]
print(results)

# Type hints on a function — documentation, not enforcement
# 함수의 타입 힌트 — 강제가 아니라 문서화
def find_first_negative(numbers: list) -> Optional[int]:
    for n in numbers:
        if n < 0:
            return n
    return None

print(find_first_negative([3, 1, -5, 2]))

# Keyword-only parameter — must be passed by name after *
# 키워드 전용 매개변수 — * 이후에는 반드시 이름으로 전달
def connect(host, *, timeout=30):
    print(host, timeout)

connect("db.example.com", timeout=5)
````

# 🧪 Small Examples (예제)

## Example 1 — *args & **kwargs (가변 인자)

````python
# *args — extra positional arguments arrive as a tuple
# *args — 추가 위치 인자는 튜플로 들어옴
def add_all(*args):
    return sum(args)

print(add_all(1, 2, 3, 4))
print(add_all())                    # works with zero arguments too (인자가 0개여도 동작)

# **kwargs — extra keyword arguments arrive as a dict
# **kwargs — 추가 키워드 인자는 딕셔너리로 들어옴
def build_profile(**kwargs):
    return kwargs

print(build_profile(name="booki", role="BA", city="Vancouver"))

# Combining regular params with *args and **kwargs
# 일반 매개변수와 *args, **kwargs를 함께 사용
def summarize(title, *details, **metadata):
    print("title:", title)
    print("details:", details)
    print("metadata:", metadata)

summarize("Report", "Q1", "Q2", author="booki", pages=12)

# Unpacking a list/dict INTO a function call, using the same * and ** syntax
# 같은 * 와 ** 문법으로 리스트/딕셔너리를 함수 호출에 언패킹
values = [1, 2, 3]
print(add_all(*values))               # unpacks the list as separate arguments (리스트를 개별 인자로 언패킹)

config = {"name": "booki", "role": "BA"}
print(build_profile(**config))          # unpacks the dict as keyword arguments (딕셔너리를 키워드 인자로 언패킹)
````

## Example 2 — Keyword-only & Positional-only Parameters (키워드·위치 전용 인자)

````python
# Keyword-only parameters — everything after a bare * must be passed by name
# 키워드 전용 매개변수 — 단독 * 이후의 모든 것은 이름으로 전달해야 함
def create_user(name, *, admin=False):
    return {"name": name, "admin": admin}

print(create_user("booki", admin=True))          # works (정상 동작)
try:
    create_user("booki", True)                       # fails — admin must be a keyword (실패 — admin은 반드시 키워드)
except TypeError as e:
    print("caught:", e)

# Positional-only parameters — everything before / must be passed by position (3.8+)
# 위치 전용 매개변수 — / 이전의 모든 것은 위치로 전달해야 함 (3.8+)
def power(base, exp, /):
    return base ** exp

print(power(2, 10))                       # works (정상 동작)
try:
    power(base=2, exp=10)                     # fails — base/exp cannot be passed by keyword (실패 — 키워드 전달 불가)
except TypeError as e:
    print("caught:", e)

# Combining all three regions: positional-only / normal / keyword-only
# 세 영역을 모두 결합: 위치 전용 / 일반 / 키워드 전용
def request(url, /, method="GET", *, timeout=30):
    return f"{method} {url} (timeout={timeout})"

print(request("api.example.com"))
print(request("api.example.com", "POST", timeout=5))
````

## Example 3 — Type Hints (타입 힌트)

````python
from typing import Optional, Union

# Basic parameter and return type hints — documentation, checked by editors/linters, not Python itself
# 기본 매개변수·반환 타입 힌트 — 문서화 목적, 에디터/린터가 확인, Python 자체는 강제 안 함
def add(a: int, b: int) -> int:
    return a + b

print(add(2, 3))
print(add(2.5, 3.5))               # runs fine even though the hint says int — hints are NOT enforced (힌트가 int여도 실행됨 — 강제되지 않음)

# Optional[X] means "X or None" — very common for functions that might not find a result
# Optional[X]는 "X 또는 None"을 의미 — 결과를 못 찾을 수도 있는 함수에 흔히 사용
def find_user(user_id: int, users: dict) -> Optional[str]:
    return users.get(user_id)

print(find_user(1, {1: "booki"}))
print(find_user(99, {1: "booki"}))

# Union[X, Y] means "X or Y" — a value that could be more than one type
# Union[X, Y]는 "X 또는 Y"를 의미 — 한 가지 이상의 타입일 수 있는 값
def to_display(value: Union[int, str]) -> str:
    return f"Value: {value}"

print(to_display(42))
print(to_display("hello"))

# Modern syntax (3.10+): X | Y instead of Union[X, Y], and built-in generics like list[int]
# 최신 문법 (3.10+): Union[X, Y] 대신 X | Y, list[int]처럼 내장 제네릭 사용
def process(value: int | str) -> str:
    return str(value)

def total(numbers: list[int]) -> int:
    return sum(numbers)

print(process(5), total([1, 2, 3]))
````

## Example 4 — Walrus Operator (왈러스 연산자)

````python
# Assign AND use a value in the same expression — no separate line needed
# 같은 표현식에서 값을 할당하고 사용 — 별도의 줄 필요 없음
data = [3, 1, 4, 1, 5, 9, 2, 6]

if (count := len(data)) > 5:
    print(f"{count} items — more than 5")

# Without walrus, this would take two lines:
# 왈러스 없이는 두 줄이 필요했을 것:
count2 = len(data)
if count2 > 5:
    print(f"{count2} items — more than 5 (the old way / 예전 방식)")

# Reusing a computed value inside a comprehension, without calling the function twice
# 함수를 두 번 호출하지 않고 컴프리헨션 안에서 계산된 값을 재사용
def slow_square(x):
    return x ** 2

big_squares = [y for x in data if (y := slow_square(x)) > 10]
print(big_squares)

# The classic while-loop pattern: pull a value, then check it, in one line
# 대표적인 while 반복문 패턴: 값을 꺼내고 확인하는 것을 한 줄로
numbers_iter = iter([5, 3, 8, 0, 2])
while (item := next(numbers_iter, None)) is not None:
    print("processing:", item)          # note: 0 is still processed correctly, since we check "is not None" (0도 올바르게 처리됨, "is not None"으로 확인하기 때문)
````

## Example 5 — Closures & nonlocal (클로저)

````python
# A closure — make_counter "remembers" `count` even after it returns
# 클로저 — make_counter는 반환된 후에도 count를 "기억"함
def make_counter(start=0):
    count = start
    def counter():
        nonlocal count               # without this, count += 1 would raise an error (이게 없으면 count += 1은 에러)
        count += 1
        return count
    return counter

c1 = make_counter()
print(c1())              # 1
print(c1())                # 2
print(c1())                  # 3

# Each call to make_counter() creates an INDEPENDENT closure with its own count
# make_counter()를 호출할 때마다 독립적인 count를 가진 별개의 클로저가 생성됨
c2 = make_counter(start=100)
print(c2())                # 101 — independent from c1 (c1과 독립적)
print(c1())                  # 4  — c1 keeps its own separate count (c1은 자신만의 count를 유지)

# A closure-based accumulator — useful for running totals without a class
# 클로저 기반 누적기 — 클래스 없이 누적 합계를 만들 때 유용
def make_accumulator():
    total = 0
    def add(amount):
        nonlocal total
        total += amount
        return total
    return add

running_total = make_accumulator()
print(running_total(10))
print(running_total(25))
print(running_total(5))
````

````python
from functools import wraps

# ============================================================
# Practice — Decorators (연습 — 데코레이터)
# ============================================================

# 1) A basic decorator — wraps a function to change its result
#    기본 데코레이터 — 함수를 감싸 결과를 바꿈
def shout(func):
    def wrapper(*args, **kwargs):
        result = func(*args, **kwargs)
        return result.upper()               # TODO: which string method makes this UPPERCASE?
    return wrapper

@shout
def greet(name):
    return f"hello, {name}"

print("1)", greet("booki"))                        # Expected: HELLO, BOOKI

# 2) A decorator that counts how many times a function was called
#    함수가 몇 번 호출되었는지 세는 데코레이터
def count_calls(func):
    def wrapper(*args, **kwargs):
        wrapper.calls += 1                 # TODO: increment by how much each call?
        return func(*args, **kwargs)
    wrapper.calls = 0
    return wrapper

@count_calls
def ping():
    return "pong"

ping()
ping()
print("2)", ping.calls)                              # Expected: 2

# 3) Forwarding arguments through a decorator with *args and **kwargs
#    *args와 **kwargs로 데코레이터를 통해 인자 전달하기
def logger(func):
    def wrapper(*args, **kwargs):
        print(f"calling {func.__name__} with {args}")
        return func(*args, **kwargs)              # TODO: replace each blank to forward ALL original arguments
    return wrapper

@logger
def add(a, b):
    return a + b

print("3)", add(2, 3))                                 # Expected: a log line, then 5

# 4) functools.wraps — preserves the original function's name and docstring
#    functools.wraps — 원래 함수의 이름과 docstring을 보존
def identity_decorator(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        return func(*args, **kwargs)
    return wrapper

@identity_decorator
def multiply(a, b):
    """Multiplies two numbers."""
    return a * b

print("4)", multiply.__name__)                           # Expected: multiply, NOT wrapper

# 5) Stacking two decorators — they apply bottom-up (closest to the function first)
#    두 데코레이터 쌓기 — 아래에서 위로 적용됨 (함수와 가장 가까운 것부터)
def add_exclaim(func):
    def wrapper(*args, **kwargs):
        return func(*args, **kwargs) + "!"
    return wrapper

@add_exclaim
@shout
def excited_greet(name):
    return f"hi, {name}"

print("5)", excited_greet("booki"))                          # TODO: reuse `shout` from exercise 1 -> Expected: HI, BOOKI!
````

# ⚠️ Common Mistakes (자주 하는 실수)

### Mistake 1 — Forgetting `@wraps(func)` inside a decorator
Without `@functools.wraps(func)`, the wrapped function silently loses its original `__name__` and docstring — `my_function.__name__` becomes `"wrapper"` instead of `"my_function"`, which breaks debugging, logging, and introspection tools that rely on that metadata.

`@functools.wraps(func)`가 없으면, 감싸진 함수는 원래의 `__name__`과 docstring을 조용히 잃는다 — `my_function.__name__`이 `"my_function"` 대신 `"wrapper"`가 되어, 그 메타데이터에 의존하는 디버깅, 로깅, 인트로스펙션 도구가 망가진다.

✅ **Fix / 해결법:**  
Always decorate the inner `wrapper` function with `@wraps(func)`.  
항상 내부 `wrapper` 함수를 `@wraps(func)`로 데코레이트하자.
```python
from functools import wraps
def my_decorator(func):
    @wraps(func)          # ✅ preserves func's real name and docstring
    def wrapper(*args, **kwargs):
        return func(*args, **kwargs)
    return wrapper
```

### Mistake 2 — Forgetting `nonlocal` when modifying a closure's variable
Assigning to a variable inside a nested function, without `nonlocal`, creates a brand-new LOCAL variable instead of modifying the enclosing one — `count += 1` without `nonlocal count` raises `UnboundLocalError`, because Python sees the assignment and treats `count` as local for the entire function, even before the `+=` line runs.

`nonlocal` 없이 중첩 함수 안에서 변수에 대입하면, 감싸는 변수를 수정하는 대신 완전히 새로운 지역 변수가 생성된다 — `nonlocal count` 없이 `count += 1`을 쓰면 `UnboundLocalError`가 발생한다, Python이 대입문을 보고 `+=` 줄이 실행되기도 전에 함수 전체에 걸쳐 `count`를 지역 변수로 취급하기 때문이다.

✅ **Fix / 해결법:**  
Add `nonlocal variable_name` before modifying an enclosing-scope variable inside a nested function.  
중첩 함수 안에서 감싸는 스코프의 변수를 수정하기 전에 `nonlocal 변수명`을 추가하자.
```python
def make_counter():
    count = 0
    def counter():
        nonlocal count    # ✅ without this line, count += 1 raises UnboundLocalError
        count += 1
        return count
    return counter
```

### Mistake 3 — Using a mutable default argument
`def add_item(item, history=[]):` looks like it creates a fresh empty list every call, but the default list is actually created ONCE, when the function is defined, and reused across every call that does not supply its own — items silently pile up across unrelated calls.

`def add_item(item, history=[]):`는 매 호출마다 새로운 빈 리스트를 만드는 것처럼 보이지만, 실제로는 함수가 정의될 때 딱 한 번 기본 리스트가 만들어지고 자신만의 값을 넘기지 않는 모든 호출에서 재사용된다 — 서로 관련 없는 호출들 사이에서 항목이 조용히 쌓인다.

✅ **Fix / 해결법:**  
Use `None` as the default, and create a fresh list inside the function body when it is `None`.  
기본값으로 `None`을 사용하고, `None`일 때 함수 본문 안에서 새 리스트를 만들자.
```python
def add_item(item, history=None):
    if history is None:
        history = []       # ✅ a fresh list every call that needs one
    history.append(item)
    return history
```

# 💡 Tips (팁)
Useful tips or shortcuts (유용한 팁이나 단축법)

- Use `*args, **kwargs` in a wrapper function whenever you do not know (or do not want to restrict) what arguments the wrapped function accepts — it is the standard way to make a decorator work with ANY function.  
감싸는 함수가 어떤 인자를 받는지 모르거나 제한하고 싶지 않을 때는 래퍼 함수에 `*args, **kwargs`를 사용하자 — 데코레이터가 어떤 함수와도 동작하게 만드는 표준적인 방법이다.
- A bare `*` in a parameter list (`def f(a, *, b):`) forces everything after it to be a keyword argument, without needing to collect the rest into `**kwargs` — useful purely for readability and preventing positional mistakes.  
매개변수 목록의 단독 `*`(`def f(a, *, b):`)는 `**kwargs`로 나머지를 모을 필요 없이 그 뒤의 모든 것을 키워드 인자로 강제한다 — 순전히 가독성과 위치 실수 방지를 위해 유용하다.
- Type hints do not slow down or change how Python runs your code at all — they are purely for humans and tools (editors, `mypy`), so there is little downside to adding them.  
타입 힌트는 Python이 코드를 실행하는 방식을 전혀 느리게 하거나 바꾸지 않는다 — 순전히 사람과 도구(에디터, `mypy`)를 위한 것이라, 추가해서 잃을 게 거의 없다.
- The walrus operator (`:=`) is best used sparingly, for genuinely reusing a value — overusing it just to save a line can make code harder to read, not easier.  
왈러스 연산자(`:=`)는 진짜로 값을 재사용할 때만 아껴 쓰는 것이 좋다 — 단순히 한 줄을 줄이려고 남용하면 코드가 더 읽기 쉬워지는 게 아니라 더 어려워진다.
- A decorator can itself take arguments by adding one more layer of nested functions — `@retry(times=3)` is a function that returns a decorator, which then returns the wrapper.  
데코레이터 자체도 중첩 함수를 한 겹 더 추가해서 인자를 받을 수 있다 — `@retry(times=3)`는 데코레이터를 반환하는 함수이고, 그 데코레이터가 다시 래퍼를 반환한다.

# 💼 Business Example (비즈니스 예제)
*How would a Business Analyst use this? BA는 이것을 실제로 어떻게 사용할까요?*

**Scenario (시나리오):** You are building a small internal reporting tool. You need a decorator that times how long each report-generation function takes, and a flexible function that can accept any number of named filters without knowing them in advance.

**시나리오:** 작은 내부 리포팅 도구를 만들고 있다. 각 리포트 생성 함수가 얼마나 걸리는지 시간을 재는 데코레이터와, 미리 알지 못하는 임의 개수의 이름 붙은 필터를 받을 수 있는 유연한 함수가 필요하다.

**To do (할 일 목록):**
- [x] Write a `@timer` decorator using `*args`/`**kwargs` so it works on any function.  
`@timer` 데코레이터를 `*args`/`**kwargs`로 작성해 어떤 함수에도 동작하게 한다.
- [x] Use `functools.wraps` so the decorated function keeps its real name.  
`functools.wraps`를 사용해 데코레이트된 함수가 원래 이름을 유지하게 한다.
- [x] Write a report function that accepts any number of filter keyword arguments with `**kwargs`.  `**kwargs`로 임의 개수의 필터 키워드 인자를 받는 리포트 함수를 작성한다.
- [x] Add a type hint to the report function's return type for documentation.  
문서화를 위해 리포트 함수의 반환 타입에 타입 힌트를 추가한다.

````python
import time

def timer(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        start = time.perf_counter()
        result = func(*args, **kwargs)
        elapsed = time.perf_counter() - start
        print(f"{func.__name__} took {elapsed:.4f}s")
        return result
    return wrapper

sales_data = [
    {"region": "West", "amount": 500, "status": "completed"},
    {"region": "East", "amount": 300, "status": "completed"},
    {"region": "West", "amount": 150, "status": "pending"},
    {"region": "East", "amount": 700, "status": "completed"},
]

@timer
def generate_report(data: list, **filters) -> list:
    """Filter records by any number of keyword criteria."""
    result = data
    for key, value in filters.items():
        result = [row for row in result if row.get(key) == value]
    return result

# Any combination of filters can be passed, without changing the function
# 함수를 바꾸지 않고도 어떤 필터 조합이든 전달 가능
west_completed = generate_report(sales_data, region="West", status="completed")
print(west_completed)

print(f"Function name preserved by @wraps: {generate_report.__name__}")
````

# 📝 Summary (요약)
*Write today's concept in 3~5 sentences. 오늘 배운 개념을 3~5문장으로 정리해보세요.*

This section rounds out everything Python functions can do beyond the basics from Section 1: `*args`/`**kwargs` accept a flexible number of positional/keyword arguments and are essential for writing wrapper functions that forward arguments to anything; keyword-only (`*`) and positional-only (`/`) markers make a function's call signature self-documenting; type hints (`Optional`, `Union`, `list[int]`) document expected types without Python enforcing them at runtime; the walrus operator (`:=`) assigns and uses a value in the same expression, avoiding a repeated computation or a separate line; closures let an inner function remember and modify (`nonlocal`) a variable from its enclosing scope across multiple calls; and decorators use `@` syntax to wrap a function with extra behavior — timing, logging, caching — without touching its internal code. Two habits are worth building immediately: always use `@functools.wraps(func)` inside a decorator's wrapper, and never use a mutable object as a default argument value.

이 섹션은 Section 1의 기초를 넘어서는 Python 함수의 모든 것을 마무리한다: `*args`/`**kwargs`는 유연한 개수의 위치·키워드 인자를 받으며, 무엇이든 인자를 전달해야 하는 래퍼 함수를 작성하는 데 필수적이다. 키워드 전용(`*`)과 위치 전용(`/`) 마커는 함수의 호출 시그니처를 자기 설명적으로 만든다. 타입 힌트(`Optional`, `Union`, `list[int]`)는 Python이 런타임에 강제하지 않으면서 예상되는 타입을 문서화한다. 왈러스 연산자(`:=`)는 같은 표현식에서 값을 할당하고 사용해, 반복된 계산이나 별도의 줄을 피한다. 클로저는 내부 함수가 여러 호출에 걸쳐 감싸는 스코프의 변수를 기억하고 수정(`nonlocal`)하게 해준다. 데코레이터는 `@` 문법을 사용해 내부 코드를 건드리지 않고 함수를 추가 동작(시간 측정, 로깅, 캐싱)으로 감싼다. 즉시 길러야 할 습관 두 가지: 데코레이터의 wrapper 안에서는 항상 `@functools.wraps(func)`를 사용하고, 기본 인자 값으로 가변 객체를 절대 사용하지 말자.
