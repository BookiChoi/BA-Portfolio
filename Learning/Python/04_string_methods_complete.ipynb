# 🧠 Concept (개념)

## What is it? (이게 뭔가요?)
*(Explain it in your own words. 자신의 언어로 설명해보세요.)*

Every Python string carries a large set of built-in methods — `.upper()`, `.split()`, `.strip()`, `.replace()`, and dozens more — that clean, search, and reshape text without ever needing `import re`. JS/TS has an equivalent set on `String.prototype`, so the *concepts* transfer directly; what changes is naming (`.trim()` becomes `.strip()`) and a couple of genuinely different design choices, like `.join()` being called on the separator rather than the array.

모든 Python 문자열은 `.upper()`, `.split()`, `.strip()`, `.replace()`를 비롯한 수십 개의 내장 메서드를 가지고 있어 `import re` 없이도 텍스트를 정리하고 검색하고 재구성할 수 있다. JS/TS도 `String.prototype`에 동등한 메서드 집합을 가지고 있어서 *개념* 자체는 그대로 옮겨온다. 달라지는 것은 이름(`.trim()`이 `.strip()`이 됨)과 `.join()`이 배열이 아니라 구분자에서 호출된다는 것처럼 진짜로 다른 몇 가지 설계 선택이다.

| | JS/TS | Python |
|---|---|---|
| Trim whitespace 공백 제거 | `s.trim()` | `s.strip()` |
| Change case 대소문자 변경 | `s.toUpperCase()` / `s.toLowerCase()` | `s.upper()` / `s.lower()` |
| Split into array 배열로 분리 | `s.split(",")` | `s.split(",")` (same!) |
| Join an array 배열 결합 | `arr.join(",")` | `",".join(arr)` (called on the separator) |
| Search for a substring 부분 문자열 검색 | `s.includes("x")` | `"x" in s` |
| Find an index 인덱스 찾기 | `s.indexOf("x")` | `s.find("x")` (returns -1 if missing) |
| Replace text 텍스트 대체 | `s.replace("a","b")` (first match only) | `s.replace("a","b")` (ALL matches by default) |
| Pad a string 문자열 패딩 | `s.padStart(10)` / `s.padEnd(10)` | `s.rjust(10)` / `s.ljust(10)` |
| Check content type 내용 타입 확인 | regex (`/^\d+$/.test(s)`) | `s.isdigit()`, `s.isalpha()`, etc. |

The single biggest gotcha for a JS/TS developer: JS's `.replace()` only touches the *first* match unless you use a global regex, while Python's `.replace()` replaces *every* match by default — the opposite default behavior for the same-looking method name.

JS/TS 개발자에게 가장 큰 함정 하나: JS의 `.replace()`는 전역 정규식을 쓰지 않는 한 *첫 번째* 매칭만 바꾸지만, Python의 `.replace()`는 기본적으로 *모든* 매칭을 바꾼다 — 이름은 같아 보이는데 기본 동작이 정반대다.

## Why do we use it? (왜 사용하나요?)
*(When is it useful? 언제 유용한가요?)*

Almost every piece of real-world data starts life as messy text: a CSV cell with stray whitespace, a name typed in the wrong case, a comma-separated field that needs splitting. String methods handle all of this without needing to import anything, and they read as a clear, linear pipeline — `raw.strip().lower().split(",")` describes exactly what happens, in the order it happens.

실제 데이터는 거의 항상 지저분한 텍스트로 시작한다: 앞뒤에 공백이 붙은 CSV 셀, 대소문자가 잘못 입력된 이름, 분리해야 하는 콤마 구분 필드. 문자열 메서드는 아무것도 import하지 않고 이 모든 걸 처리하며, `raw.strip().lower().split(",")`처럼 일어나는 순서 그대로를 명확하고 선형적인 파이프라인으로 읽을 수 있다.

Knowing `.find()` vs `.index()` (silent `-1` vs a raised `ValueError`) and `.isdigit()`/`.isalpha()` also matters for defensive code — validating a field before converting it avoids a crash three lines later with a confusing error message.

`.find()`와 `.index()`의 차이(조용한 `-1` vs `ValueError` 발생), `.isdigit()`/`.isalpha()`를 아는 것도 방어적인 코드를 짜는 데 중요하다 — 필드를 변환하기 전에 검증하면 세 줄 뒤에서 알 수 없는 에러 메시지와 함께 크래시가 나는 걸 피할 수 있다.

## When is it used in Business Analytics? (Business Analytics에서는 언제 쓰이나요?)
*(Real-world use case 실제 사용 사례)*

Cleaning inconsistent text is one of the most common first steps in any BA workflow: standardizing customer names with `.strip().title()`, splitting a "City, Country" field with `.split(",")`, or validating that an ID column is all digits with `.isdigit()` before converting it to `int`.

일관되지 않은 텍스트를 정리하는 것은 거의 모든 BA 워크플로우의 가장 흔한 첫 단계다: `.strip().title()`로 고객 이름을 표준화하거나, `.split(",")`로 "City, Country" 필드를 분리하거나, ID 컬럼을 `int`로 변환하기 전에 `.isdigit()`로 전부 숫자인지 검증하는 일 등이다.

Every one of these methods is also available directly on a pandas column via the `.str` accessor (`df["name"].str.strip()`), so mastering the plain-Python version here means Section 13's pandas string cleaning will look completely familiar.

이 메서드들은 pandas 컬럼에서도 `.str` accessor를 통해 그대로 사용할 수 있다(`df["name"].str.strip()`). 그래서 여기서 순수 Python 버전을 익혀두면 Section 13의 pandas 문자열 정리가 완전히 익숙하게 느껴질 것이다.

# 📝 Syntax (문법)

## Basic Syntax (기본 문법)

````python
s = "  Hello, World!  "

# The most common cleanup chain: strip, then change case
# 가장 흔한 정리 체인: strip 후 대소문자 변경
clean = s.strip().lower()
print(repr(clean))

# Split into parts, then join them back differently
# 나눈 뒤, 다르게 다시 결합
parts = clean.split(", ")
print(parts)
print(" | ".join(parts))

# Search and replace
# 검색과 대체
print("world" in clean)
print(clean.replace("world", "python"))

# Validate before converting
# 변환하기 전에 검증
code_str = "00123"
if code_str.isdigit():
    print(int(code_str))
````

## Common Variations (자주 쓰는 변형)

````python
# strip() variants — both sides, left only, right only, or custom characters
# strip() 변형 — 양쪽, 왼쪽만, 오른쪽만, 또는 커스텀 문자
print(repr("  hi  ".strip()))
print(repr("  hi  ".lstrip()))
print(repr("  hi  ".rstrip()))
print("xxhixx".strip("x"))

# split() with a limit, and its right-to-left twin rsplit()
# 제한 개수를 갖는 split()과 오른쪽부터 나누는 rsplit()
print("a,b,c".split(",", 1))       # stop after the first split (첫 분리 후 멈춤)
print("a,b,c".rsplit(",", 1))       # stop after the first split, counting from the right (오른쪽부터 세어 첫 분리 후 멈춤)

# find() returns -1 quietly; index() raises ValueError loudly
# find()는 조용히 -1을 반환; index()는 크게 ValueError를 던짐
print("hello".find("z"))
try:
    "hello".index("z")
except ValueError as e:
    print("caught:", e)

# startswith() / endswith() accept a tuple of options
# startswith() / endswith()는 튜플로 여러 옵션을 받을 수 있음
print("report.csv".endswith((".csv", ".xlsx")))
````

# 🧪 Small Examples (예제)

## Example 1 — Case & Whitespace (대소문자·공백)

````python
s = "Hello, World!"

# Case conversion
# 대소문자 변환
print(s.upper())              # "HELLO, WORLD!"
print(s.lower())                # "hello, world!"
print(s.title())                  # "Hello, World!" — capitalizes each word (각 단어 첫 글자 대문자)
print(s.capitalize())               # "Hello, world!" — capitalizes only the first letter (첫 글자만 대문자)
print(s.swapcase())                   # "hELLO, wORLD!" — flips every letter's case (모든 글자 대소문자 반전)

# Whitespace removal
# 공백 제거
padded = "   hello   "
print(repr(padded.strip()))        # both sides removed (양쪽 제거)
print(repr(padded.lstrip()))        # left side only (왼쪽만)
print(repr(padded.rstrip()))         # right side only (오른쪽만)

# strip() with a custom character set — removes ANY of the given characters
# 커스텀 문자 세트를 갖는 strip() — 주어진 문자들 중 어느 것이든 제거
print("xxhelloxx".strip("x"))
print("**title**".strip("*"))
````

## Example 2 — Split & Join (분리·결합)

````python
# split() — default splits on any whitespace, collapsing repeats
# split() — 기본값은 공백 기준으로 분리, 연속 공백도 처리
print("a b   c".split())

# split() with a separator
# 구분자를 지정한 split()
print("a,b,c".split(","))

# split() with a maxsplit limit
# maxsplit 제한을 갖는 split()
print("a,b,c".split(",", 1))          # only splits once (한 번만 분리)

# splitlines() — split on line breaks, without keeping the newline character
# splitlines() — 줄바꿈 기준으로 분리, 줄바꿈 문자는 남기지 않음
print("line1\nline2\nline3".splitlines())

# rsplit() — same as split(), but counts from the right
# rsplit() — split()과 같지만 오른쪽부터 셈
print("a,b,c".rsplit(",", 1))

# join() — called on the SEPARATOR, not the list (the reverse of JS)
# join() — 리스트가 아니라 구분자에서 호출 (JS와 반대)
print(",".join(["a", "b", "c"]))
print("\n".join(["line1", "line2"]))

# join() requires every element to already be a str
# join()은 모든 요소가 이미 str이어야 함
numbers = [1, 2, 3]
print(",".join(str(x) for x in numbers))    # ✅ convert first / 먼저 변환해야 함
````

## Example 3 — Search & Count (검색·카운트)

````python
s = "hello world"

# find() / rfind() — return an index, or -1 if not found
# find()/rfind() — 인덱스를 반환, 없으면 -1
print(s.find("o"))          # 4 — first match, searching left to right (왼쪽부터 첫 매칭)
print(s.rfind("o"))          # 7 — first match, searching right to left (오른쪽부터 첫 매칭)
print(s.find("z"))            # -1 — not found, no exception (없으면 예외 없이 -1)

# index() / rindex() — same idea, but raise ValueError instead of returning -1
# index()/rindex() — 같은 개념이지만 -1 대신 ValueError를 던짐
print(s.index("o"))
try:
    s.index("z")
except ValueError as e:
    print("caught:", e)

# count() — how many times a substring appears
# count() — 부분 문자열이 몇 번 나타나는지
print(s.count("o"))
print("mississippi".count("ss"))

# startswith() / endswith() — accept a single string OR a tuple of options
# startswith() / endswith() — 문자열 하나 또는 튜플로 여러 옵션 가능
print(s.startswith("hello"))
print(s.endswith("world"))
print("report.xlsx".endswith((".csv", ".xlsx", ".json")))
````

## Example 4 — Replace & Validate (대체·확인)

````python
# replace() — replaces EVERY match by default (unlike JS's .replace())
# replace() — 기본적으로 모든 매칭을 대체 (JS의 .replace()와 다름)
print("hello world world".replace("world", "Python"))

# replace() with a count limit — only the first N matches
# 개수를 제한한 replace() — 처음 N개 매칭만
print("aabbcc".replace("b", "X", 1))

# Content-type checks — useful for validating input before converting it
# 내용 타입 확인 — 변환 전 입력을 검증할 때 유용
print("hello".isalpha())              # True — every character is a letter (모든 문자가 알파벳)
print("hello123".isalnum())            # True — letters and/or digits only (알파벳과 숫자만)
print("hello123".isalpha())             # False — contains digits too (숫자도 포함)
print("123".isdigit())                   # True — every character is a digit (모든 문자가 숫자)
print("   ".isspace())                    # True — every character is whitespace (모든 문자가 공백)
print("HELLO".isupper())                   # True
print("hello".islower())                    # True

# A practical pattern: validate before you convert
# 실전 패턴: 변환하기 전에 검증
user_input = "042"
if user_input.isdigit():
    print(int(user_input))
else:
    print("invalid number")
````

# ⚠️ Common Mistakes (자주 하는 실수)

### Mistake 1 — Using `.find()` when you need `.index()` (or vice versa)
`.find()` returns `-1` when the substring is missing, which is easy to overlook and can silently become a wrong index used later in the code. `.index()` raises `ValueError` instead, which is safer to leave unhandled during development because it fails loudly right where the problem is.

`.find()`는 부분 문자열이 없을 때 `-1`을 반환하는데, 이걸 놓치기 쉽고 나중에 잘못된 인덱스로 조용히 쓰일 수 있다. `.index()`는 대신 `ValueError`를 던지므로, 개발 중에는 처리하지 않고 두는 편이 더 안전하다 — 문제가 발생한 바로 그 지점에서 크게 실패하기 때문이다.

✅ **Fix / 해결법:**  
Use `.find()` only when you plan to check for `-1` explicitly; use `.index()` (often inside a `try/except`) when a missing substring should be treated as an error.  
`-1`을 명시적으로 확인할 계획일 때만 `.find()`를 쓰고, 없는 경우를 에러로 취급해야 한다면 (보통 `try/except`와 함께) `.index()`를 사용하자.
```python
pos = "hello".find("z")
if pos == -1:               # ✅ explicit check
    print("not found")
```

### Mistake 2 — Passing non-string elements to `.join()`
`.join()` expects every element of the iterable to already be a `str`. Passing a list that contains an `int` raises a `TypeError` saying it expected a str instance.

`.join()`은 이터러블의 모든 요소가 이미 `str`이어야 한다. `int`가 포함된 리스트를 넘기면 str 인스턴스를 기대했다는 `TypeError`가 발생한다.

✅ **Fix / 해결법:**  
Convert every element with `str()` first, typically inside a generator expression.  
먼저 모든 요소를 `str()`로 변환하자, 보통 제너레이터 표현식 안에서.
```python
numbers = [1, 2, 3]
print(",".join(str(x) for x in numbers))   # ✅ "1,2,3"
```

### Mistake 3 — Trying to modify a string in place
Strings are immutable, just like `tuple`. `s[0] = "H"` raises `TypeError: 'str' object does not support item assignment` — there is no way to change a character inside an existing string object.

문자열은 `tuple`과 마찬가지로 불변이다. `s[0] = "H"`는 `TypeError: 'str' object does not support item assignment`를 던진다 — 기존 문자열 객체 내부의 문자 하나를 바꿀 방법은 없다.

✅ **Fix / 해결법:**  
Build a new string instead — slicing and concatenation, or a method that already returns a new string.  
대신 새 문자열을 만들자 — 슬라이싱과 결합, 또는 이미 새 문자열을 반환하는 메서드를 사용하자.
```python
s = "hello"
s = "H" + s[1:]   # ✅ "Hello" — a brand new string, not a mutation
```

# 💡 Tips (팁)
Useful tips or shortcuts (유용한 팁이나 단축법)

- Chain string methods for a clean one-line pipeline: `raw.strip().lower().split(",")` reads left-to-right in the exact order it executes.  
문자열 메서드를 체이닝하면 깔끔한 한 줄 파이프라인이 된다: `raw.strip().lower().split(",")`는 실행되는 순서 그대로 왼쪽에서 오른쪽으로 읽힌다.
- `"x" in s` is almost always clearer and faster to write than `s.find("x") != -1` for a simple yes/no membership check.  
단순한 예/아니오 멤버십 확인에는 `s.find("x") != -1`보다 `"x" in s`가 거의 항상 더 명확하고 쓰기도 빠르다.
- Validate with `.isdigit()` before calling `int()` on user-provided or scraped text — it turns a possible crash into a clean "invalid input" branch.  
사용자 입력이나 스크래핑한 텍스트에 `int()`를 호출하기 전에 `.isdigit()`으로 검증하면, 발생할 수 있는 크래시가 깔끔한 "잘못된 입력" 분기가 된다.
- `startswith()` and `endswith()` both accept a tuple of options in one call — no need to chain multiple `or` conditions.  
`startswith()`와 `endswith()`는 둘 다 한 번의 호출로 튜플 옵션을 받을 수 있다 — 여러 `or` 조건을 이어 쓸 필요가 없다.
- `f"{value:>10}"` and `.rjust(10)` do the same padding job — the f-string version is usually preferred when you are already building a formatted string, while `.rjust()` reads more clearly as a standalone step.  
`f"{value:>10}"`와 `.rjust(10)`은 같은 패딩 작업을 한다 — 이미 포맷된 문자열을 만드는 중이라면 f-string 쪽이 보통 더 선호되고, `.rjust()`는 독립된 단계로서 더 명확하게 읽힌다.

# 💼 Business Example (비즈니스 예제)
*How would a Business Analyst use this? BA는 이것을 실제로 어떻게 사용할까요?*

**Scenario (시나리오):** You have received a messy CSV export of customer emails and names — inconsistent capitalization, stray whitespace, and a mix of separators. You need to clean it up with pure string methods before it goes anywhere near pandas.

**시나리오:** 대소문자가 뒤섞이고 앞뒤에 공백이 붙어 있으며 구분자도 제각각인 지저분한 고객 이메일·이름 CSV를 받았다. pandas 근처에 가기 전에 순수 문자열 메서드만으로 정리해야 한다.

**To do (할 일 목록):**
- [x] Strip whitespace and standardize the case of each name. 각 이름의 공백을 제거하고 대소문자를 표준화한다.
- [x] Validate each email contains "@" before treating it as valid. 각 이메일이 유효하다고 처리하기 전에 "@"가 포함되어 있는지 검증한다.
- [x] Split a "Last, First" field into separate first/last name parts. "성, 이름" 필드를 성과 이름으로 나눈다.
- [x] Build a clean, padded summary table using `.ljust()`. `.ljust()`로 정렬된 깔끔한 요약 테이블을 만든다.

````python
raw_records = [
    "  KIM, song   ", "kim.song@email.com  ",
    "lee, jun",        "leejunATemail.com",     # missing @ — invalid email (@ 누락 — 잘못된 이메일)
    "  Park, Min  ",   "park.min@email.com",
]

# Pair up (name, email) — every 2 items form one record
# (이름, 이메일) 짝짓기 — 2개씩 묶어 한 레코드
raw_pairs = list(zip(raw_records[0::2], raw_records[1::2]))

print(f"{'Last':<10}{'First':<10}{'Email':<25}{'Valid?'}")
for raw_name, raw_email in raw_pairs:
    last, first = raw_name.strip().split(",")
    last = last.strip().title()
    first = first.strip().title()
    email = raw_email.strip().lower()
    is_valid = "@" in email
    print(f"{last:<10}{first:<10}{email:<25}{is_valid}")
````

Every method covered here reappears almost unchanged on pandas columns via the `.str` accessor, making this section direct preparation for real data cleaning work.

**Q1.** What is the difference between `.find()` and `.index()` when the target substring is not present?  
찾는 부분 문자열이 없을 때 `.find()`와 `.index()`는 어떻게 다른가요?

`.find()` returns -1 if the substring is not found. `.index()` raises a ValueError.  
`.find()`는 찾지 못하면 -1을 반환하고, `.index()`는 ValueError를 발생시킵니다.

**Q2.** Why is `",".join(["a", "b", "c"])` called on the string `","` rather than on the list, and how is this different from JS/TS's `arr.join(",")`?  
왜 `",".join(["a", "b", "c"])`는 리스트가 아니라 문자열 `","`에서 호출되나요? JS/TS의 `arr.join(",")`와는 어떻게 다른가요?

In Python, `join()` belongs to the separator string, which joins all elements of an iterable. In JS/TS, `join()` is a method of the array.  
Python에서는 `join()`이 구분자 문자열의 메서드입니다. 문자열이 iterable의 요소들을 연결합니다. JS/TS에서는 배열의 메서드입니다.

**Q3.** What error occurs if you run `",".join([1, 2, 3])` directly, and how do you fix it?  
`",".join([1, 2, 3])`을 그대로 실행하면 어떤 에러가 발생하나요? 어떻게 고칠 수 있나요?

It raises a TypeError because `join()` only accepts strings. Convert the elements using `map(str, ...)` or a list comprehension.  
`join()`은 문자열만 연결할 수 있으므로 TypeError가 발생합니다. `map(str, ...)`이나 리스트 컴프리헨션으로 문자열로 변환해야 합니다.

**Q4.** How many matches does Python's `"aaa".replace("a", "b")` replace by default, and how is that different from JS/TS's `.replace()`?  
Python의 `"aaa".replace("a", "b")`는 기본적으로 몇 개의 매칭을 대체하나요? JS/TS의 `.replace()`와는 어떻게 다른가요?

Python replaces all matches by default. JS/TS replaces only the first match unless you use a global regular expression (or `replaceAll()`).  
Python은 기본적으로 모든 매칭을 바꿉니다. JS/TS는 기본적으로 첫 번째만 바꾸며, 모두 바꾸려면 `replaceAll()`이나 전역 정규식을 사용해야 합니다.

**Q5.** Why does `s[0] = "H"` raise a `TypeError` on a Python string, and what should you write instead to change the first character?  
왜 Python 문자열에서 `s[0] = "H"`는 `TypeError`를 던지나요? 첫 글자를 바꾸려면 대신 무엇을 써야 하나요?

Strings are immutable, so you cannot change a character in place. Create a new string instead.  
문자열은 변경 불가능(immutable)하므로 문자를 직접 수정할 수 없습니다. 대신 새로운 문자열을 만들어야 합니다.
