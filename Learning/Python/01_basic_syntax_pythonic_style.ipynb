# 🧠 Concept (개념)

## What is it? (이게 뭔가요?)
*(Explain it in your own words. 자신의 언어로 설명해보세요.)*

Basic syntax is the small set of rules Python uses to declare variables, choose between types, branch with conditionals, repeat with loops, and package logic into functions. Coming from JS/TS, most of these ideas are already familiar — what changes is the notation: indentation instead of braces, `snake_case` instead of `camelCase`, and a handful of keywords (`and`/`or`/`not` instead of `&&`/`||`/`!`) that read closer to plain English.

기초 문법은 Python에서 변수를 선언하고, 타입을 고르고, 조건문으로 분기하고, 반복문으로 되풀이하고, 함수로 로직을 묶는 데 쓰이는 작은 규칙들의 집합이다. JS/TS를 알고 있다면 개념 자체는 이미 익숙하며, 달라지는 것은 표기법이다: 중괄호 대신 들여쓰기, `camelCase` 대신 `snake_case`, 그리고 `&&`/`||`/`!` 대신 평범한 영어에 가까운 `and`/`or`/`not` 키워드를 쓴다.

| | JS/TS | Python |
|---|---|---|
| Declare a variable 변수 선언 | `const x = 10;` | `x = 10` |
| Block delimiter 블록 구분 | `{ }` braces | Indentation (들여쓰기) |
| Statement end 문장 종료 | `;` semicolon | Newline (줄바꿈) |
| Boolean literals 불리언 리터럴 | `true` / `false` | `True` / `False` |
| Null value 널 값 | `null` / `undefined` | `None` |
| Logical operators 논리 연산자 | `&&` \|\| `!` | `and` `or` `not` |
| Strict equality 엄격한 동등 비교 | `===` | `==` (no `===` in Python) |
| Ternary 삼항 연산자 | `cond ? a : b` | `a if cond else b` |

Note the ternary order flips: JS/TS puts the condition first, Python puts the "true" result first.  
삼항 연산자의 순서가 뒤바뀐다는 점에 주의하자: JS/TS는 조건이 먼저 오지만 Python은 "참일 때" 결과가 먼저 온다.

## Why do we use it? (왜 사용하나요?)
*(When is it useful? 언제 유용한가요?)*

Every Python program — from a five-line script to a full pandas pipeline — is built out of these same primitives. Getting comfortable with them now means every later section (data types, functions, OOP, pandas) reads like syntax you already know, instead of a new language to decode on top of a new library.

다섯 줄짜리 스크립트부터 완전한 pandas 파이프라인까지, 모든 Python 프로그램은 결국 이 기본 요소들로 만들어진다. 지금 이 부분을 편하게 익혀두면, 이후에 나오는 모든 섹션(자료형, 함수, OOP, pandas)이 새로 배우는 언어가 아니라 이미 아는 문법으로 읽히게 된다.

Pythonic style also matters for readability: idioms like dictionary-based dispatch or index mapping exist specifically to replace long `if/elif` chains, which keeps business logic short and easy to review in code that a teammate — or future you — has to read months later.

Pythonic 스타일은 가독성 면에서도 중요하다. 딕셔너리 기반 분기, 인덱스 매핑 같은 관용구는 긴 `if/elif` 체인을 대체하기 위해 존재하며, 몇 달 뒤 동료나 미래의 자신이 코드를 다시 읽을 때도 비즈니스 로직을 짧고 검토하기 쉽게 유지해준다.

## When is it used in Business Analytics? (Business Analytics에서는 언제 쓰이나요?)
*(Real-world use case 실제 사용 사례)*

Before a single row of data touches pandas, a BA is usually writing exactly this kind of basic logic: reading a config value into a variable, looping through a small list of file names, branching on a status code, or wrapping a repeated calculation into a function so it isn't copy-pasted five times across a notebook.

pandas에 데이터가 한 행이라도 들어가기 전에, BA는 보통 이런 기초 로직부터 작성한다: 설정 값을 변수에 담고, 파일 이름이 담긴 작은 리스트를 반복문으로 순회하고, 상태 코드에 따라 분기하고, 반복되는 계산을 함수로 감싸서 노트북 여기저기에 복붙하지 않도록 만드는 일이다.

Even inside pandas code, these fundamentals resurface constantly: a `lambda` inside `.apply()` is just a function, a list comprehension is just a loop, and a boolean mask (`df["age"] > 25`) is built on the same truthy logic covered in this section.

pandas 코드 안에서도 이 기본기는 계속 다시 등장한다: `.apply()` 안의 `lambda`는 결국 함수이고, 리스트 컴프리헨션은 결국 반복문이며, 불리언 마스크(`df["age"] > 25`)도 이 섹션에서 다루는 truthy 로직 위에 만들어진다.

# 📝 Syntax (문법)

## Basic Syntax (기본 문법)

````python
# Variables & basic types — no keyword, no semicolon
# 변수 & 기본 타입 — 키워드도, 세미콜론도 없음
name = "Booki"             # str
age = 28                  # int
gpa = 3.75                 # float
is_employed = True        # bool (capital T!)
manager = None             # None instead of null/undefined

print(name, age, gpa, is_employed, manager)

# Conditional — colon + indentation define the block
# 조건문 — 콜론과 들여쓰기가 블록을 정의함
if age >= 30:
    print("Senior")
elif age >= 20:
    print("Junior")
else:
    print("Intern")

# Loop — range() generates the sequence, no i++ needed
# 반복문 — range()가 시퀀스를 생성하며 i++가 필요 없음
for i in range(3):
    print(f"iteration {i}")

# Function — def keyword, optional type hints
# 함수 — def 키워드, 타입 힌트는 선택사항
def greet(person: str) -> str:
    return f"Hello, {person}!"

print(greet(name))
````

## Common Variations (자주 쓰는 변형)

````python
# Multiple assignment — declare several variables in one line
# 다중 할당 — 한 줄로 여러 변수를 선언
a, b, c = 1, 2, 3
x = y = z = 0   # all three point at the same starting value

# Chained comparison — reads like real math, JS/TS has no equivalent
# 다중 비교 — 실제 수학처럼 읽힘, JS/TS에는 없는 문법
score = 82
print(0 <= score <= 100)   # True

# Augmented assignment — same idea as JS/TS
# 복합 할당 연산자 — JS/TS와 같은 개념
count = 0
count += 1
count *= 10

# Type checking — isinstance() is preferred over type()
# 타입 확인 — type()보다 isinstance()가 권장됨
print(isinstance(count, int))
gpa = 3.75
print(isinstance(gpa, (int, float)))   # multiple types at once

# Ternary — Python flips the JS/TS order: value_if_true first
# 삼항 연산자 — Python은 JS/TS와 순서가 반대: 참일 때 값이 먼저
status = "pass" if score >= 60 else "fail"
print(status)

# Dict-based branching — a Pythonic alternative to long if/elif chains
# 딕셔너리 기반 분기 — 긴 if/elif 체인을 대체하는 Pythonic한 방법
http_messages = {200: "OK", 404: "Not Found", 500: "Server Error"}
print(http_messages.get(404, "Unknown"))
````

# 🧪 Small Examples (예제)

## Example 1 — Variables & Basic Types (변수 & 기본 타입)

````python
# Every basic type used in one place
# 모든 기본 타입을 한곳에 모아보기
x = 10                  # int — unlimited precision, no overflow like JS numbers
y = 3.14                 # float
s = "hello"                # str
b = True                    # bool — capitalized!
n = None                     # None — Python's null/undefined
lst = [1, 2, 3]                 # list — mutable, ordered
tpl = (1, 2, 3)                    # tuple — immutable, ordered
d = {"a": 1}                          # dict — key/value pairs
st = {1, 2, 3}                           # set — unique, unordered

for value in (x, y, s, b, n, lst, tpl, d, st):
    print(type(value).__name__, "->", value)

# Multiple assignment & type checking
# 다중 할당 & 타입 확인
p, q, r = 1, 2, 3
print(type(x))                       # <class 'int'>
print(isinstance(x, int))            # True
print(isinstance(x, (int, float)))   # True — multiple types at once
````

## Example 2 — Operators & Truthy/Falsy (연산자 & Truthy/Falsy)

````python
# Arithmetic — // and % behave differently for negatives than JS/TS
# 산술 연산 — 음수에 대해 //와 %가 JS/TS와 다르게 동작함
print(10 // 3)     # 3  — floor division
print(-7 // 2)      # -4 — rounds toward negative infinity (JS gives -3)
print(2 ** 10)        # 1024 — exponent operator, no Math.pow needed

# Bool is a subtype of int — True acts like 1, False like 0
# Bool은 int의 서브타입 — True는 1처럼, False는 0처럼 동작
print(True + True)                      # 2
print(sum([True, False, True, True]))     # 3

# Short-circuit evaluation for default values
# 단축 평가로 기본값 지정하기
name = ""
display_name = name or "Anonymous"
print(display_name)   # "Anonymous" because "" is falsy

# Falsy values: 0, 0.0, "", [], {}, (), set(), None, False
# Truthy values: everything else
# Falsy 값: 0, 0.0, "", [], {}, (), set(), None, False
# Truthy 값: 그 외 전부

# Identity (is) vs equality (==)
# 동일성(is) vs 동등성(==)
a = [1, 2]
b = [1, 2]
print(a == b)   # True  — same values
print(a is b)   # False — different objects in memory
print(a is None)        # False
print(a is not None)    # True — the Pythonic way to check "not None"
````

## Example 3 — Conditionals (조건문)

````python
order_total = 120

# Standard if / elif / else
# 표준 if / elif / else
if order_total >= 100:
    discount = 0.10
elif order_total >= 50:
    discount = 0.05
else:
    discount = 0.0
print(f"discount: {discount:.0%}")

# match-case — Python 3.10+, similar to switch-case
# match-case — Python 3.10 이상, switch-case와 유사
status_code = 404
match status_code:
    case 200:
        message = "OK"
    case 404:
        message = "Not Found"
    case 400 | 401 | 403:   # multiple patterns at once / 여러 패턴을 한번에
        message = "Client Error"
    case _:                     # default case / 기본값
        message = "Unknown"
print(message)

# Pythonic alternative: index mapping instead of if/elif
# Pythonic한 대안: if/elif 대신 인덱스 매핑
labels = ["even", "odd"]
n = 7
print(labels[n % 2])   # index mapping replaces an if/else
````

## Example 4 — Loops (반복문)

````python
# range() — no i++, no separate counter variable required
# range() — i++도, 별도의 카운터 변수도 필요 없음
for i in range(3):
    print("range:", i)

# enumerate() — index + value together, replaces a manual counter
# enumerate() — 인덱스와 값을 함께, 수동 카운터를 대체
fruits = ["apple", "banana", "cherry"]
for idx, fruit in enumerate(fruits, start=1):
    print(idx, fruit)

# zip() — walk two lists together, like an index-based loop in JS/TS
# zip() — 두 리스트를 함께 순회, JS/TS의 인덱스 기반 반복을 대체
names = ["Alice", "Bob"]
scores = [90, 85]
for name_, score in zip(names, scores):
    print(name_, score)

# Dictionary iteration with .items() — the most common pattern
# .items()를 이용한 딕셔너리 순회 — 가장 흔한 패턴
prices = {"coffee": 4.5, "tea": 3.5}
for item, price in prices.items():
    print(f"{item}: ${price}")

# while loop
# while 반복문
count = 0
while count < 3:
    count += 1
print("final count:", count)
````

## Example 5 — Functions Basics (함수 기본)

````python
# Default parameters
# 기본값 파라미터
def welcome(name, msg="Hello"):
    return f"{msg}, {name}!"
print(welcome("Booki"))
print(welcome("Booki", "Welcome back"))

# Multiple return values — Python returns a tuple
# 여러 값 반환 — Python은 튜플을 반환
def min_max(values: list[int]) -> tuple[int, int]:
    return min(values), max(values)
lo, hi = min_max([3, 1, 4, 1, 5])
print(lo, hi)

# Lambda — anonymous, single-expression function
# 람다 — 익명의 단일 표현식 함수
square = lambda x: x ** 2
print(square(5))

# sorted() with a lambda key — very common in BA scripts
# sorted()와 람다 key — BA 스크립트에서 매우 흔하게 사용
records = [{"name": "Bob", "age": 30}, {"name": "Alice", "age": 25}]
# by_age = sorted(records, key=lambda r: r["age"], reverse=True)
by_age = sorted(records, key=lambda r: r["age"])
print(by_age)

# Type hints — optional but recommended for readability
# 타입 힌트 — 선택사항이지만 가독성을 위해 권장됨
def add(a: int, b: int) -> int:
    return a + b
print(add(2, 3))
````

# ⚠️ Common Mistakes (자주 하는 실수)

### Mistake 1 — Comparing to `None` with `==` instead of `is`
`x == None` usually works, but it is not the Pythonic way, and it can silently give the wrong answer for custom objects that override `__eq__`. Python's style guide recommends `is` for singleton comparisons like `None`, `True`, and `False`.

`x == None`도 대부분 동작하지만 Pythonic한 방식이 아니며, `__eq__`를 재정의한 커스텀 객체에서는 조용히 잘못된 결과를 줄 수 있다. Python 스타일 가이드는 `None`, `True`, `False`처럼 유일한(singleton) 값과 비교할 때는 `is`를 권장한다.

✅ **Fix / 해결법:**  
Use `x is None` / `x is not None` instead of `==`.  
`==` 대신 `x is None` / `x is not None`을 사용하자.
```python
x = None
if x is None:       # ✅ correct / 올바른 방법
    print("empty")
```

### Mistake 2 — Flipping the ternary operator order
JS/TS ternary reads `condition ? valueIfTrue : valueIfFalse`. Python's reads `valueIfTrue if condition else valueIfFalse` — the condition sits in the *middle*, not first. Typing it in JS/TS order is a very common early bug.

JS/TS의 삼항 연산자는 `condition ? valueIfTrue : valueIfFalse` 순서로 읽는다. Python은 `valueIfTrue if condition else valueIfFalse`로, 조건이 처음이 아니라 *중간*에 온다. JS/TS 순서로 입력하는 것은 초반에 매우 흔한 실수다.

✅ **Fix / 해결법:**  
Read it right-to-left as "[this], if [condition] is true, otherwise [that]".  
"[조건]이 참이면 [이 값], 아니면 [저 값]"으로 오른쪽에서 왼쪽으로 읽는 습관을 들이자.
```python
score = 75
result = "pass" if score >= 60 else "fail"   # ✅ condition in the middle
```

### Mistake 3 — Assuming `//` always rounds like JS/TS truncation
Python's `//` is *floor* division — it always rounds toward negative infinity, not toward zero. For positive numbers this matches JS/TS intuition, but for negatives it does not: `-7 // 2` is `-4` in Python, while the equivalent truncating division in JS/TS gives `-3`.

Python의 `//`는 *내림(floor)* 나눗셈으로, 항상 음의 무한대 방향으로 내림한다. 0 방향으로 자르지 않는다. 양수에서는 JS/TS 직관과 일치하지만 음수에서는 다르다: Python에서 `-7 // 2`는 `-4`이지만, JS/TS의 동등한 절삭 나눗셈은 `-3`을 준다.

✅ **Fix / 해결법:**  
If you need truncation toward zero, use `math.trunc(a / b)` instead of `//`.  
0 방향으로 자르고 싶다면 `//` 대신 `math.trunc(a / b)`를 사용하자.
```python
import math
print(-7 // 2)              # -4  (floor)
print(math.trunc(-7 / 2))   # -3  (truncate toward zero)
```

# 💡 Tips (팁)
Useful tips or shortcuts (유용한 팁이나 단축법)

- Use an index-mapping list (`labels[x % 2]`) or a dict (`handler.get(key, default)`) instead of a short `if/elif` chain — it is shorter and easier to extend.  
짧은 `if/elif` 체인 대신 인덱스 매핑 리스트(`labels[x % 2]`)나 딕셔너리(`handler.get(key, default)`)를 사용하면 더 짧고 확장하기도 쉽다.
- `x or default` is Python's quick way to fill in a fallback value, but remember it treats `0`, `""`, and `[]` as falsy too — not just `None`.  
`x or default`는 대체값을 채우는 Python의 빠른 방법이지만, `None`뿐 아니라 `0`, `""`, `[]`도 falsy로 취급한다는 점을 기억하자.
- `enumerate(items, start=1)` is almost always nicer than manually tracking a counter variable inside a loop — reach for it whenever you need both an index and a value.  
반복문 안에서 카운터 변수를 수동으로 관리하는 것보다 `enumerate(items, start=1)`을 쓰는 편이 거의 항상 낫다 — 인덱스와 값이 둘 다 필요할 때마다 이걸 떠올리자.
- `isinstance(x, (int, float))` checks several types at once with a tuple — no need for `type(x) == int or type(x) == float`.  
`isinstance(x, (int, float))`는 튜플로 여러 타입을 한 번에 검사한다 — `type(x) == int or type(x) == float`처럼 쓸 필요가 없다.
- Type hints (`def add(a: int, b: int) -> int:`) do not change runtime behavior, but they make functions self-documenting and let editors catch mistakes early.  
타입 힌트(`def add(a: int, b: int) -> int:`)는 런타임 동작을 바꾸지는 않지만 함수를 자기 설명적으로 만들고 에디터가 실수를 미리 잡아주게 한다.

*How would a Business Analyst use this? BA는 이것을 실제로 어떻게 사용할까요?*

**Scenario (시나리오):** You are a junior BA at an online store. Before the sales data ever reaches pandas, you have been handed a small, plain-Python list of today's orders and asked for a same-day flag report — no libraries, just the basics from this notebook.

**시나리오:** 당신은 온라인 쇼핑몰의 주니어 BA다. 매출 데이터가 pandas로 넘어가기 전, 오늘의 주문이 담긴 순수 Python 리스트를 전달받아 라이브러리 없이 이 노트북에서 배운 기초만으로 당일 플래그 리포트를 만들어야 한다.

**To do (할 일 목록):**
- [x] Loop through the order records and compute the total revenue.  
주문 레코드를 순회하며 총 매출을 계산한다.
- [x] Flag any order over $100 as `"high_value"` using a conditional.  
조건문을 사용해 $100 이상 주문을 `"high_value"`로 표시한다.
- [x] Use a dict lookup to translate each order's status code into a readable label.  
딕셔너리 조회로 각 주문의 상태 코드를 읽기 쉬운 라벨로 변환한다.
- [x] Write a small function that summarizes one order as a formatted string.  
주문 하나를 포맷된 문자열로 요약하는 작은 함수를 작성한다.

````python
orders = [
    {"id": 1, "customer": "Kim",  "amount": 120.0, "status": 1},
    {"id": 2, "customer": "Lee",  "amount": 45.5,  "status": 2},
    {"id": 3, "customer": "Park", "amount": 300.0, "status": 1},
    {"id": 4, "customer": "Choi", "amount": 60.0,  "status": 3},
]

status_labels = {1: "Paid", 2: "Pending", 3: "Cancelled"}

def summarize_order(order):
    """Return a one-line, human-readable summary of a single order.
    주문 하나를 사람이 읽기 좋은 한 줄 요약으로 반환한다."""
    flag = "high_value" if order["amount"] > 100 else "standard"
    label = status_labels.get(order["status"], "Unknown")
    return f"#{order['id']} {order['customer']:<6} ${order['amount']:>7.2f} [{flag:<10}] {label}"

total_revenue = 0
for order in orders:
    print(summarize_order(order))
    total_revenue += order["amount"]

print("-" * 50)
print(f"Total revenue (총 매출): ${total_revenue:,.2f}")
````

*Write today's concept in 3~5 sentences. 오늘 배운 개념을 3~5문장으로 정리해보세요.*

Python expresses the same core ideas as JS/TS — variables, types, conditionals, loops, and functions — through a smaller, more uniform syntax: indentation instead of braces, `snake_case` instead of `camelCase`, and English-like keywords (`and`, `or`, `not`, `is`) instead of symbols. A handful of details genuinely differ and cause real bugs if ignored: floor division rounds toward negative infinity, the ternary operator's condition sits in the middle, and identity checks (`is`) are preferred over equality (`==`) for `None`. Pythonic style also introduces alternatives — dict-based dispatch, index mapping, `enumerate`, `zip` — that replace patterns a JS/TS developer would normally reach for a manual loop or counter to solve. Every later section in this curriculum, from data types to pandas, is written using exactly this vocabulary, so fluency here pays off immediately.

Python은 JS/TS와 동일한 핵심 개념 — 변수, 타입, 조건문, 반복문, 함수 — 을 더 작고 일관된 문법으로 표현한다: 중괄호 대신 들여쓰기, `camelCase` 대신 `snake_case`, 기호 대신 영어에 가까운 키워드(`and`, `or`, `not`, `is`)를 사용한다. 몇 가지는 실제로 다르며 무시하면 진짜 버그로 이어진다: 나눗셈 `//`는 음의 무한대 방향으로 내림하고, 삼항 연산자의 조건은 중간에 오며, `None`과 비교할 때는 동등성(`==`)보다 동일성(`is`)이 권장된다. Pythonic 스타일은 딕셔너리 기반 분기, 인덱스 매핑, `enumerate`, `zip`처럼 JS/TS 개발자가 보통 수동 반복문이나 카운터로 해결하던 패턴을 대체하는 대안도 제공한다. 이 커리큘럼의 이후 모든 섹션, 자료형부터 pandas까지, 바로 이 어휘로 쓰여지므로 여기서의 숙련도는 곧바로 도움이 된다.
