# 🧠 Concept (개념)

## What is it? (이게 뭔가요?)
*(Explain it in your own words. 자신의 언어로 설명해보세요.)*

A `class` bundles data (attributes) and behavior (methods) into a single reusable blueprint. Every instance method takes `self` as its first parameter — Python's explicit stand-in for the JS/TS `this` — and magic methods (`__init__`, `__str__`, `__eq__`, and dozens more) are how a class plugs into Python's built-in operators and functions instead of needing special-cased support.

`class`는 데이터(속성)와 동작(메서드)을 하나의 재사용 가능한 청사진으로 묶는다. 모든 인스턴스 메서드는 첫 번째 매개변수로 `self`를 받는다 — JS/TS의 `this`에 해당하는 Python의 명시적인 자리 표시자다. 매직 메서드(`__init__`, `__str__`, `__eq__` 등 수십 개)는 클래스가 특별 취급 없이 Python의 내장 연산자·함수에 연결되는 방법이다.

| | JS/TS | Python |
|---|---|---|
| Instance reference 인스턴스 참조 | `this` (implicit) | `self` (explicit first parameter) |
| Constructor 생성자 | `constructor(name) { this.name = name; }` | `def __init__(self, name): self.name = name` |
| Custom string conversion 커스텀 문자열 변환 | `toString()` | `__str__(self)` |
| Custom equality 커스텀 동등성 | `equals()` method (no operator overload) | `__eq__(self, other)` — plugs directly into `==` |
| Static method 정적 메서드 | `static method() {}` | `@staticmethod` |
| Getter/setter 게터·세터 | `get x() {}` / `set x(v) {}` | `@property` / `@x.setter` |
| Inheritance 상속 | `class Dog extends Animal` | `class Dog(Animal):` |
| Call parent constructor 부모 생성자 호출 | `super(name)` | `super().__init__(name)` |
| Lightweight data class 가벼운 데이터 클래스 | plain object literal `{x, y}` | `@dataclass class Point: x: float; y: float` |

The concept with no real JS/TS equivalent: Python lets a class define exactly how it behaves with `==`, `<`, `len()`, `in`, and even hashing — a custom class can be sorted, compared, or used as a dict key just as naturally as a built-in type, once the right magic methods are implemented.

JS/TS에 딱히 대응물이 없는 개념: Python은 클래스가 `==`, `<`, `len()`, `in`, 심지어 해싱과 어떻게 동작할지 정확히 정의하게 해준다 — 올바른 매직 메서드만 구현하면 커스텀 클래스도 내장 타입만큼 자연스럽게 정렬되거나 비교되거나 딕셔너리 키로 쓰일 수 있다.

## Why do we use it? (왜 사용하나요?)
*(When is it useful? 언제 유용한가요?)*

Bundling related data and the functions that operate on it into one class keeps a codebase organized as it grows — instead of passing the same dict around to a dozen separate functions, a class groups the data and its behavior together, with a name that documents what it represents.

관련된 데이터와 그것을 다루는 함수를 하나의 클래스로 묶으면 코드베이스가 커져도 정리된 상태를 유지할 수 있다 — 같은 딕셔너리를 십여 개의 개별 함수에 계속 전달하는 대신, 클래스는 데이터와 그 동작을 함께 묶고, 그것이 무엇을 나타내는지 문서화하는 이름을 갖는다.

Magic methods matter because they let a custom type integrate seamlessly with the rest of the language: implement `__lt__` once, and `sorted()`, `min()`, and `max()` all immediately work correctly on a list of your objects, without writing a single `key=` lambda.

매직 메서드가 중요한 이유는 커스텀 타입이 언어의 나머지 부분과 매끄럽게 통합되게 해주기 때문이다: `__lt__`를 한 번 구현하면 `sorted()`, `min()`, `max()`가 객체 리스트에서 `key=` 람다를 하나도 쓰지 않고 즉시 올바르게 동작한다.

## When is it used in Business Analytics? (Business Analytics에서는 언제 쓰이나요?)
*(Real-world use case 실제 사용 사례)*

A small class to represent a "Customer" or "Order" — with validation in `__init__`, a readable `__str__` for logging, and comparison methods for sorting — is often cleaner than passing the same handful of dict keys through every function in a script, especially once the logic grows past a few functions.

`__init__`에 검증 로직을, 로깅을 위한 읽기 좋은 `__str__`을, 정렬을 위한 비교 메서드를 갖춘 작은 "Customer"나 "Order" 클래스는, 특히 로직이 몇 개의 함수를 넘어서 커지고 나면, 같은 몇 개의 딕셔너리 키를 스크립트의 모든 함수에 계속 전달하는 것보다 대체로 더 깔끔하다.

`@dataclass` in particular shows up constantly for configuration objects and lightweight records — it gives a class `__init__`, `__repr__`, and `__eq__` for free from a short field list, which is exactly the shape of most business records (an order, a report config, a filter set).

`@dataclass`는 특히 설정 객체와 가벼운 레코드에서 계속 등장한다 — 짧은 필드 목록만으로 클래스에 `__init__`, `__repr__`, `__eq__`를 공짜로 제공하며, 이는 대부분의 비즈니스 레코드(주문, 리포트 설정, 필터 세트)의 형태와 정확히 일치한다.

# 📝 Syntax (문법)

## Basic Syntax (기본 문법)

````python
class Employee:
    # Class variable — shared by every instance
    # 클래스 변수 — 모든 인스턴스가 공유
    company = "Acme Analytics"

    def __init__(self, name, role):
        # Instance variables — unique to each object
        # 인스턴스 변수 — 각 객체마다 고유함
        self.name = name
        self.role = role

    def __str__(self):
        return f"{self.name} ({self.role})"

    def promote(self, new_role):
        self.role = new_role

emp = Employee("booki", "BA-in-training")
print(emp)
emp.promote("Junior BA")
print(emp)
print(Employee.company, emp.company)     # accessible from both the class and any instance (클래스와 인스턴스 양쪽에서 접근 가능)
````

## Common Variations (자주 쓰는 변형)

````python
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def __repr__(self):
        return f"Product({self.name!r}, {self.price})"

    def __eq__(self, other):
        return self.price == other.price

    def __lt__(self, other):
        return self.price < other.price

    @classmethod
    def free_sample(cls, name):
        return cls(name, 0.0)          # a factory that builds a Product a different way (다른 방식으로 Product를 만드는 팩토리)

    @staticmethod
    def apply_tax(price, rate=0.1):
        return round(price * (1 + rate), 2)

products = [Product("Tea", 3.5), Product("Coffee", 4.5), Product("Water", 1.0)]
print(sorted(products))                # works because __lt__ is defined (__lt__가 정의되어 있어 동작함)

free = Product.free_sample("Trial Pack")
print(free)
print(Product.apply_tax(4.5))
````

# 🧪 Small Examples (예제)

## Example 1 — Classes & Instances (클래스·인스턴스)

````python
class BankAccount:
    bank_name = "Python Bank"            # class variable — same for every account (클래스 변수 — 모든 계좌가 동일)

    def __init__(self, owner, balance=0):
        self.owner = owner                 # instance variable — unique per account (인스턴스 변수 — 계좌마다 고유)
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount

    def withdraw(self, amount):
        if amount > self.balance:
            print("Insufficient funds")
        else:
            self.balance -= amount

# Creating instances — each one gets its own independent state
# 인스턴스 생성 — 각각 독립적인 상태를 가짐
acc1 = BankAccount("booki", 100)
acc2 = BankAccount("Kim", 50)

acc1.deposit(20)
acc2.withdraw(10)
print(acc1.owner, acc1.balance)
print(acc2.owner, acc2.balance)

# Class variables are shared — but assigning through an INSTANCE creates a new instance variable instead
# 클래스 변수는 공유되지만 — 인스턴스를 통해 대입하면 새로운 인스턴스 변수가 생성됨
print(acc1.bank_name, acc2.bank_name)
acc1.bank_name = "booki's Custom Bank"          # this only affects acc1, not the class or acc2 (acc1에만 영향, 클래스나 acc2는 그대로)
print(acc1.bank_name, acc2.bank_name)
````

## Example 2 — Magic Methods (매직 메서드)

````python
class Money:
    def __init__(self, amount):
        self.amount = amount

    def __str__(self):                       # used by print() and str() (print()와 str()이 사용)
        return f"${self.amount:.2f}"

    def __repr__(self):                        # used by the REPL and debugging (REPL과 디버깅에서 사용)
        return f"Money({self.amount})"

    def __eq__(self, other):                     # powers == (==를 동작시킴)
        return self.amount == other.amount

    def __lt__(self, other):                       # powers < and sorted()/min()/max() (<와 sorted()/min()/max()를 동작시킴)
        return self.amount < other.amount

    def __add__(self, other):                        # powers + (+를 동작시킴)
        return Money(self.amount + other.amount)

    def __len__(self):                                  # powers len() — here, "number of whole dollars" (len()을 동작시킴 — 여기선 "달러 정수부")
        return int(self.amount)

wallet = Money(25.50)
print(wallet)                        # uses __str__
print(repr(wallet))                    # uses __repr__
print(Money(10) == Money(10))            # uses __eq__
print(Money(5) < Money(10))                # uses __lt__
print(Money(10) + Money(5))                  # uses __add__
print(len(Money(42.99)))                       # uses __len__

prices = [Money(30), Money(10), Money(20)]
print(sorted(prices))                # sorted() works automatically thanks to __lt__ (sorted()가 __lt__ 덕분에 자동으로 동작)
````

## Example 3 — Class Methods & Static Methods (클래스·정적 메서드)

````python
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

    def __repr__(self):
        return f"Employee({self.name!r}, {self.salary})"

    # classmethod — receives the CLASS itself (cls), commonly used as an alternative constructor
    # classmethod — 클래스 자체(cls)를 받음, 주로 대안 생성자로 사용
    @classmethod
    def from_csv_row(cls, row):
        name, salary = row.split(",")
        return cls(name, float(salary))

    # staticmethod — receives NEITHER self NOR cls; a regular function living in the class namespace
    # staticmethod — self도 cls도 받지 않음; 클래스 네임스페이스에 있는 평범한 함수
    @staticmethod
    def is_valid_salary(salary):
        return salary > 0

# Calling the classmethod like a second constructor
# 두 번째 생성자처럼 classmethod 호출
emp1 = Employee.from_csv_row("booki,65000")
print(emp1)

# Calling the staticmethod — no instance needed at all
# staticmethod 호출 — 인스턴스가 전혀 필요 없음
print(Employee.is_valid_salary(65000))
print(Employee.is_valid_salary(-500))

# Both are also accessible from an instance, though that's less common
# 둘 다 인스턴스에서도 접근 가능하지만 흔히 쓰이지는 않음
print(emp1.is_valid_salary(emp1.salary))
````

## Example 4 — Properties (프로퍼티)

````python
class Circle:
    def __init__(self, radius):
        self._radius = radius              # underscore prefix signals "internal" by convention (밑줄 접두사는 관례상 "내부용"을 의미)

    @property
    def radius(self):
        return self._radius

    @radius.setter
    def radius(self, value):
        if value <= 0:
            raise ValueError("radius must be positive")
        self._radius = value

    @property
    def area(self):                        # a READ-ONLY computed property — no setter defined (읽기 전용 계산 프로퍼티 — 세터 없음)
        return round(3.14159 * self._radius ** 2, 2)

c = Circle(5)
print(c.radius)              # looks like attribute access, but actually calls the getter (속성 접근처럼 보이지만 실제로는 게터 호출)
print(c.area)                   # computed on the fly every time it's accessed (접근할 때마다 즉석에서 계산됨)

c.radius = 10                     # looks like attribute assignment, but actually calls the setter (속성 대입처럼 보이지만 실제로는 세터 호출)
print(c.radius, c.area)

try:
    c.radius = -5                     # the setter's validation kicks in (세터의 검증이 작동함)
except ValueError as e:
    print("caught:", e)
````

## Example 5 — Inheritance (상속)

````python
class Animal:
    def __init__(self, name, sound="..."):
        self.name = name
        self.sound = sound

    def speak(self):
        return f"{self.name} says {self.sound}"

    def __str__(self):
        return f"Animal({self.name})"

class Dog(Animal):
    def __init__(self, name, breed):
        super().__init__(name, sound="Woof")        # reuse the parent's __init__ instead of repeating it (반복하지 않고 부모의 __init__ 재사용)
        self.breed = breed

    def __str__(self):                                  # overriding — Dog gets its own version (오버라이딩 — Dog는 자신만의 버전을 가짐)
        return f"Dog({self.name}, {self.breed})"

class Cat(Animal):
    def __init__(self, name):
        super().__init__(name, sound="Meow")
    # no __str__ override — Cat falls back to Animal's version (오버라이딩 없음 — Cat은 Animal의 버전을 그대로 씀)

dog = Dog("Rex", "Labrador")
cat = Cat("Whiskers")

print(dog.speak())                # inherited method, works unchanged (상속된 메서드, 변경 없이 동작)
print(cat.speak())
print(dog)                          # uses Dog's own __str__
print(cat)                            # falls back to Animal's __str__

print(isinstance(dog, Animal))          # True — Dog IS an Animal (Dog는 Animal이다)
print(isinstance(dog, Dog))
print(isinstance(cat, Dog))               # False — Cat is not a Dog
````

# ⚠️ Common Mistakes (자주 하는 실수)

### Mistake 1 — Forgetting `self` as a method's first parameter
Every instance method needs `self` as its first parameter, even though Python passes it automatically when called as `obj.method()`. Forgetting it causes a `TypeError` about the wrong number of arguments the moment the method is called.

모든 인스턴스 메서드는 `obj.method()`로 호출될 때 Python이 자동으로 전달함에도 불구하고 첫 번째 매개변수로 `self`가 필요하다. 이를 빠뜨리면 메서드가 호출되는 순간 인자 개수가 맞지 않는다는 `TypeError`가 발생한다.

✅ **Fix / 해결법:**  
Every method defined inside a class needs `self` as its first parameter, without exception.  
클래스 안에 정의된 모든 메서드는 예외 없이 첫 번째 매개변수로 `self`가 필요하다.
```python
class Greeter:
    def greet(self):        # ✅ self is required, even if unused inside
        return "hello"
```

### Mistake 2 — Defining `__eq__` without `__hash__`
The moment a class defines `__eq__`, Python automatically sets `__hash__` to `None` for that class, making instances unhashable — they can no longer be put in a `set` or used as a `dict` key, even if that worked before `__eq__` was added.

클래스가 `__eq__`를 정의하는 순간, Python은 그 클래스의 `__hash__`를 자동으로 `None`으로 설정해 인스턴스를 해시 불가능하게 만든다 — `__eq__`를 추가하기 전에는 가능했더라도, 더 이상 `set`에 넣거나 `dict` 키로 쓸 수 없다.

✅ **Fix / 해결법:**  
If instances still need to be hashable, define `__hash__` explicitly alongside `__eq__`.  
인스턴스가 여전히 해시 가능해야 한다면 `__eq__`와 함께 `__hash__`도 명시적으로 정의하자.
```python
class Money:
    def __init__(self, amount):
        self.amount = amount
    def __eq__(self, other):
        return self.amount == other.amount
    def __hash__(self):                # ✅ without this, {Money(5), Money(10)} raises TypeError
        return hash(self.amount)
```

### Mistake 3 — Using a mutable class variable expecting per-instance behavior
A class variable defined as a mutable object (like `items = []`) is shared by EVERY instance, just like a mutable default argument. Calling `self.items.append(x)` on one instance silently affects every other instance too, since they are all pointing at the exact same list.

가변 객체(`items = []` 같은)로 정의된 클래스 변수는 가변 기본 인자와 마찬가지로 모든 인스턴스가 공유한다. 한 인스턴스에서 `self.items.append(x)`를 호출하면 모두가 정확히 같은 리스트를 가리키고 있기 때문에 다른 모든 인스턴스에도 조용히 영향을 준다.

✅ **Fix / 해결법:**  
Create mutable attributes inside `__init__` so each instance gets its own independent object.  
각 인스턴스가 독립적인 객체를 갖도록 가변 속성은 `__init__` 안에서 생성하자.
```python
class Cart:
    def __init__(self):
        self.items = []      # ✅ a fresh list for every instance, not shared
```

# 💡 Tips (팁)
Useful tips or shortcuts (유용한 팁이나 단축법)

- `@dataclass` automatically generates `__init__`, `__repr__`, and `__eq__` from just the field list — for simple data-holding classes, it eliminates almost all of the boilerplate this section covers.  
`@dataclass`는 필드 목록만으로 `__init__`, `__repr__`, `__eq__`를 자동으로 생성한다 — 단순한 데이터 보관용 클래스라면 이 섹션에서 다룬 보일러플레이트를 거의 전부 없애준다.
- Implementing `__repr__` is almost always worth it, even if `__str__` already exists — `__repr__` is what shows up in a list, in the debugger, and at the REPL, and its convention is to look like valid Python code that could recreate the object.  
`__str__`이 이미 있어도 `__repr__`을 구현하는 것은 거의 항상 가치가 있다 — `__repr__`은 리스트, 디버거, REPL에서 보이는 것이며, 관례상 객체를 다시 만들 수 있는 유효한 Python 코드처럼 보이게 작성한다.
- Prefix an attribute with a single underscore (`self._radius`) to signal "this is internal, please use the property instead" — Python has no true private attributes, this is a convention, not enforcement.  
속성 앞에 밑줄 하나(`self._radius`)를 붙이면 "이건 내부용이니 프로퍼티를 대신 사용하라"는 신호가 된다 — Python에는 진짜 비공개 속성이 없으며, 이는 강제가 아니라 관례다.
- `isinstance(obj, ParentClass)` returns `True` for both the parent class and any subclass — it is the preferred way to check "is this an Animal (of any kind)?" instead of comparing `type(obj) == Animal` directly.  
`isinstance(obj, ParentClass)`는 부모 클래스와 모든 서브클래스 모두에 대해 `True`를 반환한다 — `type(obj) == Animal`을 직접 비교하는 대신 "이게 (어떤 종류든) Animal인가?"를 확인하는 권장 방법이다.
- Reach for `@classmethod` whenever a class needs more than one way to be constructed — `Employee.from_csv_row(...)` reads far more clearly at the call site than a constructor with a confusing extra flag.  
클래스에 하나 이상의 생성 방법이 필요할 때는 `@classmethod`를 떠올리자 — `Employee.from_csv_row(...)`는 헷갈리는 추가 플래그가 있는 생성자보다 호출부에서 훨씬 명확하게 읽힌다.

# 💼 Business Example (비즈니스 예제)
*How would a Business Analyst use this? BA는 이것을 실제로 어떻게 사용할까요?*

**Scenario (시나리오):** You are building a small internal library to represent customer accounts consistently across several scripts, instead of passing around raw dicts everywhere.

**시나리오:** 여러 스크립트에서 순수 딕셔너리를 여기저기 넘기는 대신, 고객 계정을 일관되게 표현하는 작은 내부 라이브러리를 만들고 있다.

**To do (할 일 목록):**
- [x] Define a `Customer` class with `__init__`, and a `__repr__` for clean debugging output.  
`__init__`과 깔끔한 디버깅 출력을 위한 `__repr__`을 가진 `Customer` 클래스를 정의한다.
- [x] Add a `@classmethod` that builds a `Customer` from a raw CSV-style row.  
원본 CSV 형태의 행으로부터 `Customer`를 만드는 `@classmethod`를 추가한다.
- [x] Add a `@property` that computes the customer's tier from their total spend, without storing it directly.  
직접 저장하지 않고 총 지출로부터 고객 등급을 계산하는 `@property`를 추가한다.
- [x] Implement `__lt__` so a list of customers can be sorted by spend with plain `sorted()`.  
순수 `sorted()`로 고객 리스트를 지출 기준 정렬할 수 있도록 `__lt__`를 구현한다.

````python
class Customer:
    def __init__(self, name, total_spend):
        self.name = name
        self.total_spend = total_spend

    def __repr__(self):
        return f"Customer({self.name!r}, {self.total_spend})"

    @classmethod
    def from_row(cls, row):
        """Build a Customer from a 'name,total_spend' CSV-style string."""
        name, spend = row.split(",")
        return cls(name, float(spend))

    @property
    def tier(self):
        """Computed on the fly from total_spend — never stored directly."""
        if self.total_spend >= 1000:
            return "Gold"
        elif self.total_spend >= 500:
            return "Silver"
        return "Bronze"

    def __lt__(self, other):
        return self.total_spend < other.total_spend

raw_rows = ["booki,1200", "Kim,450", "Lee,780", "Park,50"]
customers = [Customer.from_row(row) for row in raw_rows]

for c in customers:
    print(f"{c.name:<6} spend=${c.total_spend:<8.2f} tier={c.tier}")

print("-" * 40)
print("Sorted by spend, highest first (지출 기준 정렬, 높은 순):")
for c in sorted(customers, reverse=True):
    print(f"  {c.name}: {c.tier}")
````

# 📝 Summary (요약)
*Write today's concept in 3~5 sentences. 오늘 배운 개념을 3~5문장으로 정리해보세요.*

A Python class bundles data (via `self.attribute` inside `__init__`) and behavior (methods) into a single reusable blueprint, with `self` as the explicit stand-in for what other languages call `this`. Magic methods — `__str__`, `__repr__`, `__eq__`, `__lt__`, `__len__`, and more — are how a custom class plugs directly into Python's built-in operators and functions, so implementing `__lt__` once makes `sorted()`, `min()`, and `max()` all work correctly without a `key=` lambda. `@classmethod` provides alternative constructors that receive the class itself (`cls`), `@staticmethod` is a plain function that simply lives inside the class namespace, and `@property` makes a method callable like a plain attribute — useful for validation on write or values computed on the fly. Inheritance (`class Dog(Animal):`) reuses a parent's behavior via `super()` and can override any method by redefining it, and `isinstance()` correctly recognizes both the exact class and any subclass. `@dataclass` automates the most common boilerplate (`__init__`, `__repr__`, `__eq__`) for simple data-holding classes from nothing more than a field list.

Python 클래스는 데이터(`__init__` 안의 `self.attribute`를 통해)와 동작(메서드)을 하나의 재사용 가능한 청사진으로 묶으며, `self`는 다른 언어의 `this`에 해당하는 명시적인 자리 표시자다. 매직 메서드 — `__str__`, `__repr__`, `__eq__`, `__lt__`, `__len__` 등 — 는 커스텀 클래스가 Python의 내장 연산자·함수에 직접 연결되는 방법이라, `__lt__`를 한 번 구현하면 `key=` 람다 없이도 `sorted()`, `min()`, `max()`가 모두 올바르게 동작한다. `@classmethod`는 클래스 자체(`cls`)를 받는 대안 생성자를 제공하고, `@staticmethod`는 클래스 네임스페이스에 있는 평범한 함수이며, `@property`는 메서드를 평범한 속성처럼 호출 가능하게 만든다 — 쓰기 시 검증하거나 즉석에서 계산되는 값에 유용하다. 상속(`class Dog(Animal):`)은 `super()`를 통해 부모의 동작을 재사용하고 재정의로 어떤 메서드든 오버라이드할 수 있으며, `isinstance()`는 정확한 클래스와 모든 서브클래스를 모두 올바르게 인식한다. `@dataclass`는 필드 목록만으로 단순한 데이터 보관용 클래스에서 가장 흔한 보일러플레이트(`__init__`, `__repr__`, `__eq__`)를 자동화한다.

# ❓ Review Questions (복습 문제)

**Q1.** Why does every instance method need `self` as its first parameter, even though it is never explicitly passed when calling `obj.method()`?  
`obj.method()`를 호출할 때 명시적으로 전달하지 않는데도, 왜 모든 인스턴스 메서드는 첫 번째 매개변수로 `self`가 필요한가요?

`self` is the reference to the current instance. Python automatically passes the object as the first argument when calling an instance method.  
`self`는 현재 객체(instance)를 가리키는 참조입니다. `obj.method()`를 호출하면 Python이 자동으로 객체를 첫 번째 인자로 전달합니다.

**Q2.** What does implementing `__lt__` on a class make possible, without any other code changes?  
클래스에 `__lt__`를 구현하면 다른 코드 변경 없이 무엇이 가능해지나요?

`__lt__` allows objects to be compared with < and enables sorting with `sorted()`.  
`__lt__`를 구현하면 객체끼리 < 비교가 가능해지고 `sorted()`를 사용할 수 있습니다.

**Q3.** What is the difference between `@classmethod` and `@staticmethod`?  
`@classmethod`와 `@staticmethod`는 어떻게 다른가요?

@classmethod receives the class itself as the first argument (cls) and is useful for alternative constructors or class-level operations. @staticmethod does not receive self or cls and is used for utility functions related to the class.  
@classmethod는 첫 번째 인자로 클래스 자체(cls)를 받으며, 다른 방식의 객체 생성(factory method)이나 클래스 단위 작업에 사용합니다. @staticmethod는 self나 cls를 받지 않으며, 클래스와 관련은 있지만 객체나 클래스 데이터가 필요 없는 유틸리티 함수에 사용합니다.

**Q4.** Why does defining `__eq__` on a class make its instances unhashable unless `__hash__` is also defined?  
클래스에 `__eq__`를 정의하면 왜 `__hash__`도 함께 정의하지 않는 한 인스턴스가 해시 불가능해지나요?

Defining __eq__ changes how objects are compared by value. Python disables hashing by default because mutable objects with custom equality could break dictionary and set behavior.  
__eq__를 정의하면 객체 비교 방식이 변경됩니다. Python은 값 비교 기준과 hash 값이 달라져 딕셔너리와 셋의 동작이 깨지는 것을 막기 위해 기본적으로 hash 사용을 비활성화합니다.

**Q5.** What three things does `@dataclass` generate automatically from a field list?  
`@dataclass`는 필드 목록으로부터 자동으로 어떤 세 가지를 생성하나요?

@dataclass automatically generates methods such as __init__, __repr__, and __eq__.  
@dataclass는 필드 목록을 기반으로 __init__, __repr__, __eq__ 같은 메서드를 자동으로 생성합니다.
