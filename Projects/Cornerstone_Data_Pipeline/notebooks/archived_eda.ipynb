```python
import pandas as pd

df = pd.read_excel("../data/online_retail.xlsx")

display(df.head())
```

```python
# 1. StockCode and Description have different unique value counts (StockCode와 Description의 unique 개수가 다름)
# Why don't the number of StockCodes and Descriptions match? (매칭이 안되는 이유가 뭘까?)
desc_count = df.groupby("StockCode")["Description"].nunique()
codes = desc_count[desc_count >= 2].index

display(
    df[df["StockCode"].isin(codes)]
    .assign(StockCode=lambda x: x["StockCode"].astype(str))
    .drop_duplicates(subset=["StockCode", "Description"])  # Keep only one row per (StockCode, Description) combination (같은 조합은 1행만)
    .set_index(["StockCode", "Description"])
    .sort_index()
)
# Non-product values found in Description: (Description에 들어있는 비상품 값들)
# check, NaN, damaged, ebay, Dotcomgiftshop Gift Voucher £20.00,
# 'to push order through as stock was', 'wrongly marked' etc.
```

```python
# 2. WHITE HANGING HEART T-LIGHT HOLDER appears the most in the raw count
# However, other products may have higher total quantities sold, 
# so frequency alone doesn't confirm best-seller
# Total quantity sold by Description
qty_by_desc = df.groupby("Description")["Quantity"].sum()

# Top 10 best-selling products by quantity (가장 많이 팔린 top 10)
print(qty_by_desc.nlargest(10))
# WORLD WAR 2 GLIDERS ASSTD DESIGNS has the highest total quantity sold (얘가 제일 많이 팔림)

# Bottom 10 by total quantity — includes negatives (가장 적게 팔린 bottom 10, 음수 포함)
print(qty_by_desc.nsmallest(10))
# Non-product entries show up at the bottom (비상품 항목들이 하단에 나타남):
# Unsaleable, destroyed, thrown away, Damaged, damages, incorrect stock entry, throw away, 'wrongly marked'
# Similar meanings but stored as different Description strings (비슷한 의미인데 다른 표현으로 기록돼있음)
```

```python
# Revenue per row (행별 매출)
df["Revenue"] = df["Quantity"] * df["UnitPrice"]

# Total revenue by Description (Description별 총 매출)
revenue_by_desc = df.groupby("Description")["Revenue"].sum()

print("Top 10 products by total revenue (가장 많이 팔린 금액 top 10)")
print(revenue_by_desc.nlargest(10))
```

```python
# 3. Quantity min is -80995 and max is 80995 — identical magnitude suggests a matched order/cancellation pair
#    (min/max 값이 동일한 숫자 → 주문과 취소 쌍일 가능성)
print(df["Quantity"].describe())
print(df["Quantity"].value_counts().head(20))
```

```python
# Rows with min/max Quantity (min/max 행)
display(df[df["Quantity"] == df["Quantity"].min()])
display(df[df["Quantity"] == df["Quantity"].max()])

# Alternatively: sort by absolute value to surface the most extreme rows (또는 절대값 큰 순으로 정렬)
display(
    df.reindex(df["Quantity"].abs().sort_values(ascending=False).index)
    .head(10)[["InvoiceNo", "StockCode", "Description", "Quantity", "UnitPrice", "CustomerID", "InvoiceDate"]]
)
```

```python
# When StockCode and CustomerID match but InvoiceNo differs and starts with 'C' → likely a cancellation
#   (StockCode와 CustomerID가 같은데 InvoiceNo가 다르고 앞에 C가 있으면 취소 주문)
# Negative Quantity + InvoiceNo starting with 'C' → cancelled order (Quantity 음수 + C 인보이스 → 취소)
# Negative Quantity + CustomerID NaN → non-product Description, InvoiceNo does NOT start with 'C'
#   (Quantity 음수 + CustomerID NaN → Description이 이상한 값, InvoiceNo에 C 없음)
```

```python
# C536757 appears as a cancellation, but the original order is not found in the dataset (원본 주문 기록이 안보임) — let's investigate
display(df[( df["StockCode"] == 84347) & (df["CustomerID"] == 15838.0)])
# The original order likely existed before the date range covered by this dataset
#   (현재 데이터 범위 전에 주문이 있었던 것으로 추정)
```

```python
# InvoiceNo 542504 has a NaN Description — what is this record? (description이 NaN인데 얘는 뭘까?)
display(df[(df["StockCode"] == 37413) & (df["Description"].isna())])
display(df[(df["StockCode"] == 37413) & (df["CustomerID"].isna())])
```

```python
desc_nan = df[df["Description"].isna()]
desc_cust_nan = df[(df["Description"].isna()) & (df["CustomerID"].isna())]

print(f"Description NaN: {len(desc_nan)} rows (행)")
print(f"Description NaN and CustomerID NaN (둘 다 NaN): {len(desc_cust_nan)} rows (행)")

display(desc_nan)
# When Description is NaN, CustomerID is also always NaN (description이 NaN이면 CustomerID도 NaN)
```

```python
# 5. CustomerID is missing in 135,080 rows — what do these rows have in common?
#    (CustomerID가 135,080개 없음 — 어떤 공통점이 있을까?)
cust_nan = df[df["CustomerID"].isna()]

# Rows where CustomerID is NaN but Description exists (CustomerID NaN인데 Description 있는 것들)
# display(cust_nan[cust_nan["Description"].notna()])

# Rows where CustomerID is NaN but UnitPrice is positive (CustomerID NaN인데 UnitPrice 양수인 것들)
display(cust_nan[cust_nan["UnitPrice"] > 0])

# display(
#     cust_nan[(cust_nan["Description"].notna()) & (cust_nan["UnitPrice"] <= 0)][["Description", "Quantity", "UnitPrice", "InvoiceNo"]]
#     .astype({"Description": str})  # Cast Description to str so NaN values don't break string-based sorting
#   (Description을 str로 변환해 NaN이 정렬을 망가뜨리지 않도록)
#     .set_index("Description")
#     .sort_index()
# )
# print(
#     cust_nan[(cust_nan["Description"].notna()) & (cust_nan["UnitPrice"] <= 0)]["Description"].value_counts()
# )
# Check which rows have InvoiceNo starting with 'C' (cancellations) vs. not
#   (InvoiceNo가 C로 시작하는/안하는 것들 구분)
is_cancelled = cust_nan["InvoiceNo"].astype(str).str.startswith("C", na=False)
display(cust_nan[is_cancelled])   # Cancellations — InvoiceNo starts with 'C' (취소 건)
# Check if any C-invoices have positive Quantity (C 인보이스 중 Quantity 양수인 것 있는지)
display(cust_nan[~is_cancelled])  # Not cancelled (C 없음)
# Rows where CustomerID is NaN but Quantity is positive (CustomerID NaN인데 Quantity 양수)

# Rows where both CustomerID and Description are NaN (둘 다 NaN)

# Rows where CustomerID is NaN and UnitPrice is 0 (UnitPrice 0)

# Rows where CustomerID is NaN and UnitPrice is negative (UnitPrice 음수)

# Rows where CustomerID is NaN and Quantity is negative (Quantity 음수)
```

```python
# 6. UnitPrice mean is 4.61, but min is -11062.06 and max is 38970.00 — extreme spread worth investigating
#    (mean 값은 4.61인데 min/max 차이가 너무 큼)
print(df["UnitPrice"].describe())

# 구간별 개수
print("UnitPrice <= 0:", (df["UnitPrice"] <= 0).sum())   # 2,517건
print("0 < UnitPrice < 100:", ((df["UnitPrice"] > 0) & (df["UnitPrice"] < 100)).sum())
print("UnitPrice >= 100:", (df["UnitPrice"] >= 100).sum())
```

```python
# Rows with min/max UnitPrice (min/max 행)
display(df[df["UnitPrice"] == df["UnitPrice"].min()])
display(df[df["UnitPrice"] == df["UnitPrice"].max()])

# Top 10 highest UnitPrice rows (UnitPrice 상위 10)
display(
    df.nlargest(10, "UnitPrice")[
        ["InvoiceNo", "StockCode", "Description", "Quantity", "UnitPrice", "CustomerID"]
    ]
)
```

```python
zero_price = df[df["UnitPrice"] == 0]
print(zero_price["Description"].value_counts(dropna=False).head(15))
print("CustomerID NaN 비율:", zero_price["CustomerID"].isna().mean())
print("Quantity < 0 비율:", (zero_price["Quantity"] < 0).mean())
```
